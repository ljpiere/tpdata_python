```python
import requests
import time
import pandas as pd
```

```python
#@title *Conjunto de Datos a Analizar - JSON - Pandas*

JSON_URL = "https://www.datos.gov.co/resource/uzcf-b9dh.json" # Portal Nacional - SODA2

def obtener_dataframe(url, limite = 1000, max_total = None):
    all_rows = []
    offset = 0

    while True:
        params = {"$limit": limite, "$offset": offset}  # SODA2
        try:
            r = requests.get(url, params = params)
            r.raise_for_status()
            data = r.json()
        except Exception as e:
            print(f"Error en petición: {e}")
            break

        if not data:
            break

        all_rows.extend(data)

        if max_total and len(all_rows) >= max_total:
            all_rows = all_rows[:max_total]
            break

        offset += limite
        time.sleep(0.1)

    return pd.DataFrame(all_rows)


    
df = obtener_dataframe(JSON_URL, limite = 1000, max_total = 1000000) # Filas Total Conjunto
df.head()
```
