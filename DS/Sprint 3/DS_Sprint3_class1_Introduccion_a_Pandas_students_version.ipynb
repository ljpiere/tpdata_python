## <span style="color: #2826DE;">Selección de columnas y filas (20 mins)

Una de las operaciones más comunes al trabajar con datos es **seleccionar partes específicas del dataset**.

Esto puede incluir:

- seleccionar **una o varias columnas**
- seleccionar **filas específicas**
- seleccionar **filas y columnas al mismo tiempo**

Pandas permite hacer esto de forma muy sencilla utilizando **indexación**.



### <span style="color:#2772F2;">Selección de columnas</span>

Las columnas de un DataFrame pueden seleccionarse usando **el nombre de la columna**.

### Seleccionar una sola columna

````python
# seleccionar columna
df["species"]
````

### Seleccionar múltiples columnas

Si queremos seleccionar varias columnas debemos usar doble corchete.

````python
df[["species", "island"]]
````

Esto devuelve un nuevo DataFrame con únicamente esas columnas.


### <span style="color:#2772F2;">Selección de filas</span>

Para seleccionar filas por posición numérica, Pandas utiliza iloc.

### Seleccionar una fila

````python
df.iloc[0]
````

Esto devuelve la primera fila del dataset.

### Seleccionar múltiples filas

````python
df.iloc[0:5]
````

Esto devuelve las primeras 5 filas.

Pregunta

### Selección de filas y columnas al mismo tiempo

Podemos usar iloc para seleccionar filas y columnas simultáneamente.

La estructura es:
```
df.iloc[filas, columnas]
```

````python
df.iloc[0:5, 0:3]
````

Esto selecciona:

- filas 0 a 4

- columnas 0 a 2

## <span style="color: #2826DE;">Filtros de datos (15 mins)

En análisis de datos es muy común filtrar registros que cumplen ciertas condiciones.
Esto permite enfocarnos únicamente en los datos relevantes.

En pandas, los filtros se realizan utilizando condiciones booleanas.

### <span style="color:#2772F2;">Filtrar por una condición</span>

Podemos seleccionar solo las filas que cumplen una condición.

Por ejemplo: pingüinos de la especie Adelie

````python
df[df["species"] == "Adelie"]
````

Aquí:

- `df["species"] == "Adelie"` genera una serie booleana

- `True` indica las filas que cumplen la condición

- pandas devuelve solo esas filas

### <span style="color:#2772F2;">Filtrar valores mayores o menores</span>

También podemos filtrar con operadores de comparación.

Ejemplo: pingüinos con longitud de pico mayor a 50 mm

````python
df[df["bill_length_mm"] > 50]
````

#### Otros operadores comunes:

| Operador | Significado   |
| -------- | ------------- |
| `>`      | mayor que     |
| `<`      | menor que     |
| `>=`     | mayor o igual |
| `<=`     | menor o igual |
| `==`     | igual         |
| `!=`     | diferente     |


### <span style="color:#2772F2;">Filtrar con múltiples condiciones</span>

Podemos combinar condiciones usando:

& → AND

| → OR

Ejemplo: pingüinos Adelie con masa corporal mayor a 4000 g

````python
df[(df["species"] == "Adelie") & (df["body_mass_g"] > 4000)]
````

> Importante: cada condición debe ir entre paréntesis.

### <span style="color:#2772F2;">Filtrar valores nulos</span>

En datasets reales es común tener valores faltantes (NaN).

Podemos detectarlos con:

````python
df[df["bill_length_mm"].isna()]
````

### <span style="color:#2772F2;">Filtrar con múltiples valores</span>

Si queremos filtrar varias categorías usamos isin().

Ejemplo: pingüinos de las especies Adelie o Gentoo

````python
df[df["species"].isin(["Adelie", "Gentoo"])]
````

Esto crea un objeto agrupado donde cada grupo representa una especie.

### <span style="color:#2772F2;">Calcular estadísticas por grupo</span>

Podemos calcular métricas como:

- promedio

- suma

- conteo

- máximo

- mínimo

Ejemplo: promedio de masa corporal por especie

````python
df.groupby("species")["body_mass_g"].mean()
````

### <span style="color:#2772F2;">Contar registros por grupo</span>

Para saber cuántos pingüinos hay de cada especie:

````python
df.groupby("species").size()
````

### <span style="color:#2772F2;">Aplicar múltiples agregaciones</span>

Podemos calcular varias métricas al mismo tiempo usando agg().

````python
df.groupby("species")["body_mass_g"].agg(["mean", "min", "max"])
````

## <span style="color: #2826DE;">Cálculos agrupados (15 mins)

Una de las operaciones más importantes en análisis de datos es agrupar datos y calcular estadísticas.

En pandas esto se hace con `groupby()`

### <span style="color:#2772F2;">Agrupar por una columna</span>

Podemos agrupar los datos por especie.

````python
df.groupby("species")
````
