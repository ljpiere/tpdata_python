## <span style="color: #2826DE;">Estadística Descriptiva (25 min)

La estadística descriptiva nos permite **resumir y entender** un conjunto de datos. Antes de cualquier modelo de ML, siempre empezamos aquí.

### Medidas de tendencia central
- **Media (μ)**: promedio aritmético — sensible a outliers
- **Mediana**: valor central — robusta ante outliers
- **Moda**: valor más frecuente — útil para categóricas

### Medidas de dispersión
- **Varianza (σ²)**: promedio de las desviaciones al cuadrado
- **Desviación estándar (σ)**: raíz de la varianza — mismas unidades que los datos
- **IQR**: Q3 - Q1 — rango intercuartílico, robusto ante outliers
- **Rango**: max - min — sensible a extremos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de ejemplo: salarios de un equipo de DS
salarios = pd.Series([35000, 42000, 45000, 48000, 50000, 52000, 55000, 60000, 120000])

print('=== Medidas de Tendencia Central ===')
print(f'Media:   ${salarios.mean():,.0f}')
print(f'Mediana: ${salarios.median():,.0f}')
print(f'Moda:    ${salarios.mode().values[0]:,.0f}')

print('\n=== Medidas de Dispersión ===')
print(f'Desv. Estándar: ${salarios.std():,.0f}')
print(f'IQR:            ${salarios.quantile(0.75) - salarios.quantile(0.25):,.0f}')
print(f'Rango:          ${salarios.max() - salarios.min():,.0f}')

# Visualización
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(salarios, bins=8, edgecolor='black', alpha=0.7)
axes[0].axvline(salarios.mean(), color='red', linestyle='--', label=f'Media: ${salarios.mean():,.0f}')
axes[0].axvline(salarios.median(), color='green', linestyle='-', label=f'Mediana: ${salarios.median():,.0f}')
axes[0].legend()
axes[0].set_title('Distribución de Salarios')
axes[1].boxplot(salarios, vert=False)
axes[1].set_title('Boxplot de Salarios')
plt.tight_layout()
plt.show()

print('\n💡 Nota: La media ($56,333) es mayor que la mediana ($50,000)')
print('   Esto indica que hay un outlier alto que "jala" la media hacia arriba.')
```

## <span style="color: #2826DE;">Distribuciones de Probabilidad (25 min)

<img src="../../Images/S5_DS_lotery.jpeg" width="600">

Una **distribución de probabilidad** describe cómo se reparten los valores posibles de una variable aleatoria.

### Tipos de variables aleatorias
- **Discretas**: valores contables (número de clientes, dados)
- **Continuas**: valores en un rango (temperatura, salario)

### Distribuciones más importantes en DS
| Distribución | Tipo | Uso en DS |
|---|---|---|
| Normal (Gaussiana) | Continua | Errores, alturas, muchos fenómenos naturales |
| Binomial | Discreta | Éxito/fracaso repetido (clicks, conversiones) |
| Poisson | Discreta | Eventos raros en un intervalo (fraudes, llamadas) |
| Uniforme | Ambas | Generación de números aleatorios |
| Exponencial | Continua | Tiempo entre eventos (tiempo entre compras) |

```python
from scipy import stats

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Normal
x = np.linspace(-4, 4, 100)
axes[0,0].plot(x, stats.norm.pdf(x), 'b-', lw=2)
axes[0,0].fill_between(x, stats.norm.pdf(x), alpha=0.3)
axes[0,0].set_title('Normal (μ=0, σ=1)')
axes[0,0].axvline(0, color='red', linestyle='--', alpha=0.5)

# Binomial
n, p = 20, 0.3
x_binom = np.arange(0, 21)
axes[0,1].bar(x_binom, stats.binom.pmf(x_binom, n, p), color='green', alpha=0.7)
axes[0,1].set_title(f'Binomial (n={n}, p={p})')

# Poisson
lam = 3
x_pois = np.arange(0, 15)
axes[1,0].bar(x_pois, stats.poisson.pmf(x_pois, lam), color='orange', alpha=0.7)
axes[1,0].set_title(f'Poisson (λ={lam})')

# Exponencial
x_exp = np.linspace(0, 5, 100)
axes[1,1].plot(x_exp, stats.expon.pdf(x_exp), 'r-', lw=2)
axes[1,1].fill_between(x_exp, stats.expon.pdf(x_exp), alpha=0.3, color='red')
axes[1,1].set_title('Exponencial (λ=1)')

plt.suptitle('Distribuciones de Probabilidad más usadas en DS', fontsize=14)
plt.tight_layout()
plt.show()
```

## <span style="color: #2826DE;">Ley de Grandes Números y Teorema Central del Límite (20 min)

Estos dos teoremas son la **base matemática** de por qué la estadística funciona:

### Ley de Grandes Números (LLN)
> A medida que aumentamos el tamaño de la muestra, la media muestral se acerca a la media poblacional.

<img src="../../Images/S5_DS_LGN.png" width="1200">

### Teorema Central del Límite (CLT)
> La distribución de las medias muestrales tiende a una distribución normal, sin importar la forma de la distribución original, siempre que n sea suficientemente grande (n ≥ 30).

**¿Por qué importa?** Porque nos permite usar tests estadísticos basados en la normal incluso cuando los datos originales no son normales.

<img src="../../Images/S5_DS_TCL.png" width="1200">

```python
# Demostración visual del Teorema Central del Límite
np.random.seed(42)

# Distribución original: UNIFORME (no es normal)
poblacion = np.random.uniform(0, 100, size=100000)

# Tomamos muchas muestras y calculamos sus medias
tamanos = [1, 5, 10, 30, 50, 100]
fig, axes = plt.subplots(2, 3, figsize=(14, 8))

for idx, n in enumerate(tamanos):
    ax = axes[idx // 3, idx % 3]
    # Tomar 1000 muestras de tamaño n y calcular la media de cada una
    medias = [np.random.choice(poblacion, size=n).mean() for _ in range(1000)]
    ax.hist(medias, bins=30, edgecolor='black', alpha=0.7, density=True)
    ax.set_title(f'n = {n} (std = {np.std(medias):.2f})')
    ax.set_xlim(0, 100)

plt.suptitle('CLT: Distribución de medias muestrales\n'
             '(Población original: Uniforme [0, 100])', fontsize=14)
plt.tight_layout()
plt.show()

print('💡 Observa cómo a medida que n crece:')
print('   1. La distribución de medias se vuelve NORMAL (campana)')
print('   2. La dispersión se reduce (std disminuye)')
print('   3. Se centra alrededor de la media poblacional (50)')
```
