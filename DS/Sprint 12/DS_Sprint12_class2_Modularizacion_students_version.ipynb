### <span style="color: #2772F2;">Del caos del notebook al orden profesional

Un proyecto de DS profesional separa las responsabilidades en carpetas con propósitos claros. Esta es una estructura recomendada:

```
mi_proyecto/
├── README.md              ← Documentación del proyecto
├── params.py              ← Parámetros y configuración centralizada
├── requirements.txt       ← Dependencias del proyecto
├── run.py                 ← Script principal de ejecución
│
├── input/                 ← Datos crudos (NUNCA se modifican)
│   ├── train.csv
│   └── test.csv
│
├── preprocessing/         ← Módulos de limpieza y transformación
│   ├── __init__.py
│   ├── clean_data.py
│   └── feature_engineering.py
│
├── models/                ← Módulos de entrenamiento y evaluación
│   ├── __init__.py
│   ├── train_model.py
│   └── evaluate.py
│
├── output/                ← Resultados (predicciones, reportes)
│   ├── predictions.csv
│   └── metrics.json
│
├── saved_models/          ← Modelos entrenados serializados
│   └── model_v1.joblib
│
└── notebooks/             ← Análisis exploratorio (NO producción)
    └── EDA.ipynb
```

### <span style="color: #2772F2;">Principios clave:

1. **Separación de concerns**: cada carpeta/archivo tiene UNA responsabilidad
2. **Datos inmutables**: `input/` nunca se modifica; los resultados van a `output/`
3. **Reproducibilidad**: con params.py + requirements.txt, cualquiera puede replicar
4. **Sin notebooks en producción**: los notebooks son para explorar, no para ejecutar

Veamos cómo se ve esta estructura en código. Crearemos una versión simplificada para entender cada pieza.

````python
# ====== params.py ======
# Archivo de configuración centralizada

"""
Parámetros del proyecto.
Cambiar aquí afecta todo el pipeline sin tocar otros archivos.
"""

# Rutas
INPUT_PATH = "input/train.csv"
TEST_PATH = "input/test.csv"
MODEL_PATH = "saved_models/model_v1.joblib"
OUTPUT_PATH = "output/predictions.csv"

# Parámetros del modelo
MODEL_TYPE = "random_forest"
N_ESTIMATORS = 200
MAX_DEPTH = 10
RANDOM_STATE = 42

# Parámetros de preprocesamiento
TEST_SIZE = 0.2
TARGET_COL = "target"
DROP_COLS = ["id", "fecha_registro"]
NUMERIC_COLS = ["edad", "ingreso", "antiguedad"]
CATEGORICAL_COLS = ["region", "tipo_contrato"]

# Parámetros de evaluación
METRIC = "f1"
CV_FOLDS = 5

print("=== params.py ===")
print(f"Modelo: {MODEL_TYPE} con {N_ESTIMATORS} árboles")
print(f"Columnas numéricas: {NUMERIC_COLS}")
print(f"Columnas categóricas: {CATEGORICAL_COLS}")
print(f"Métrica de evaluación: {METRIC}")
print("\n→ Todos los parámetros en UN solo lugar")
````

````python
# ====== preprocessing/clean_data.py ======
"""
Funciones de limpieza de datos.
"""
import pandas as pd
import numpy as np

def load_data(path):
    """Carga el dataset y muestra información básica."""
    df = pd.read_csv(path) if path.endswith('.csv') else pd.read_parquet(path)
    print(f"  Cargado: {path} ({df.shape[0]} filas, {df.shape[1]} columnas)")
    return df

def clean_data(df, drop_cols=None, target_col=None):
    """Limpia el dataset: elimina columnas, maneja NaN."""
    df = df.copy()
    
    # Eliminar columnas innecesarias
    if drop_cols:
        existing_cols = [c for c in drop_cols if c in df.columns]
        df = df.drop(columns=existing_cols)
    
    # Eliminar filas con target nulo
    if target_col and target_col in df.columns:
        df = df.dropna(subset=[target_col])
    
    # Reportar NaN restantes
    nan_count = df.isna().sum().sum()
    if nan_count > 0:
        print(f"  ⚠️ {nan_count} valores NaN restantes (se manejarán en pipeline)")
    
    return df

# Demo
print("=== preprocessing/clean_data.py ===")
print("Funciones disponibles: load_data(), clean_data()")
print("\nCada función tiene docstring y hace UNA cosa bien.")
````

````python
# ====== models/train_model.py ======
"""
Funciones de entrenamiento y evaluación del modelo.
"""
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

def build_pipeline(numeric_cols, categorical_cols, model_type='random_forest', **model_params):
    """Construye el pipeline completo de preprocesamiento + modelo."""
    
    # Preprocesamiento numérico
    numeric_transformer = Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])
    
    # Preprocesamiento categórico
    categorical_transformer = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
    ])
    
    # Combinar
    preprocessor = ColumnTransformer([
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
    
    # Seleccionar modelo
    if model_type == 'random_forest':
        model = RandomForestClassifier(**model_params)
    elif model_type == 'gradient_boosting':
        model = GradientBoostingClassifier(**model_params)
    
    # Pipeline completo
    pipeline = Pipeline([
        ('preprocessing', preprocessor),
        ('model', model)
    ])
    
    return pipeline

def evaluate_model(pipeline, X, y, cv=5, metric='f1'):
    """Evalúa el pipeline con cross-validation."""
    scores = cross_val_score(pipeline, X, y, cv=cv, scoring=metric)
    print(f"  {metric}: {scores.mean():.4f} ± {scores.std():.4f}")
    return scores

# Demo
print("=== models/train_model.py ===")
print("Funciones: build_pipeline(), evaluate_model()")
print("\nEl pipeline COMPLETO se construye con una función.")
print("Cambiar modelo o preprocesamiento = cambiar params.py")
````

### <span style="color: #2772F2;">Imports entre módulos

Cuando tu proyecto tiene múltiples archivos .py, necesitas importar funciones de un módulo a otro:

```python
# En run.py:
from preprocessing.clean_data import load_data, clean_data
from models.train_model import build_pipeline, evaluate_model
import params
```

El archivo `__init__.py` (puede estar vacío) convierte una carpeta en un "paquete" de Python, permitiendo estos imports.

````python
# Ejemplo de cómo se verían los imports en run.py
example_run_py = """
# ====== run.py ======
import params
from preprocessing.clean_data import load_data, clean_data
from preprocessing.feature_eng import create_features
from models.train_model import build_pipeline, evaluate_model
import joblib

def main(mode='train'):
    if mode == 'train':
        # Cargar y limpiar
        df = load_data(params.INPUT_PATH)
        df = clean_data(df, drop_cols=params.DROP_COLS)
        df = create_features(df)
        
        # Separar X, y
        X = df.drop(params.TARGET_COL, axis=1)
        y = df[params.TARGET_COL]
        
        # Construir y evaluar
        pipeline = build_pipeline(
            numeric_cols=params.NUMERIC_COLS,
            categorical_cols=params.CATEGORICAL_COLS,
            model_type=params.MODEL_TYPE,
            n_estimators=params.N_ESTIMATORS
        )
        evaluate_model(pipeline, X, y, cv=params.CV_FOLDS)
        
        # Entrenar con todos los datos y guardar
        pipeline.fit(X, y)
        joblib.dump(pipeline, params.MODEL_PATH)
        print(f"Modelo guardado en {params.MODEL_PATH}")

if __name__ == '__main__':
    main(mode='train')
"""
print(example_run_py)
````

## <span style="color: #2826DE;">Guardar y cargar modelos con joblib (15 mins)

### <span style="color: #2772F2;">¿Por qué guardar modelos?

Entrenar un modelo puede tomar minutos u horas. No queremos re-entrenar cada vez que necesitamos hacer predicciones. La solución es **serializar** (guardar) el modelo entrenado en un archivo.

### <span style="color: #2772F2;">joblib vs pickle

Ambos serializan objetos Python, pero **joblib** es preferido para ML porque:
- Más eficiente con arrays de NumPy (que abundan en modelos sklearn)
- Mejor compresión
- Más rápido para objetos grandes

### <span style="color: #2772F2;">Flujo en producción:

```
ENTRENAMIENTO (una vez):
  datos → pipeline.fit(X_train, y_train) → joblib.dump(pipeline, 'model.joblib')

PREDICCIÓN (muchas veces):
  pipeline = joblib.load('model.joblib') → pipeline.predict(X_nuevo)
```

**Importante:** Guardamos el **Pipeline completo**, no solo el modelo. Así incluimos todo el preprocesamiento y no tenemos que replicarlo manualmente.

Demostremos el flujo completo: entrenar, guardar, cargar y predecir con joblib.

````python
import joblib
import tempfile
import os

# Crear un dataset de ejemplo
np.random.seed(42)
n = 400
df_demo = pd.DataFrame({
    'edad': np.random.randint(18, 65, n).astype(float),
    'ingreso': np.random.exponential(35000, n) + 15000,
    'antiguedad': np.random.exponential(3, n),
    'region': np.random.choice(['norte', 'sur', 'centro'], n),
    'tipo_contrato': np.random.choice(['temporal', 'indefinido'], n),
})
y_demo = (df_demo['ingreso'] > 40000).astype(int)

# Construir pipeline
numeric_cols = ['edad', 'ingreso', 'antiguedad']
categorical_cols = ['region', 'tipo_contrato']

pipeline_demo = build_pipeline(
    numeric_cols=numeric_cols,
    categorical_cols=categorical_cols,
    model_type='random_forest',
    n_estimators=100,
    random_state=42
)

# Entrenar
from sklearn.model_selection import train_test_split
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    df_demo, y_demo, test_size=0.2, random_state=42)

pipeline_demo.fit(X_train_d, y_train_d)
print(f"Modelo entrenado. Accuracy en test: {pipeline_demo.score(X_test_d, y_test_d):.4f}")
````

````python
# --- GUARDAR el modelo ---
model_path = os.path.join(tempfile.gettempdir(), 'modelo_demo_v1.joblib')
joblib.dump(pipeline_demo, model_path)
file_size = os.path.getsize(model_path) / 1024  # KB

print(f"=== Modelo GUARDADO ===")
print(f"  Ruta: {model_path}")
print(f"  Tamaño: {file_size:.1f} KB")
print(f"  Incluye: preprocesamiento + modelo entrenado")
````

````python
# --- CARGAR el modelo (simular producción) ---
pipeline_cargado = joblib.load(model_path)

# Simular datos nuevos (como llegarían en producción)
nuevos_datos = pd.DataFrame({
    'edad': [28, 45, 33],
    'ingreso': [55000, 30000, 42000],
    'antiguedad': [2.5, 8.0, 1.0],
    'region': ['norte', 'sur', 'centro'],
    'tipo_contrato': ['indefinido', 'temporal', 'indefinido']
})

# Predecir SIN re-entrenar
predicciones = pipeline_cargado.predict(nuevos_datos)
probabilidades = pipeline_cargado.predict_proba(nuevos_datos)[:, 1]

print("=== Modelo CARGADO - Predicciones en producción ===\n")
for i, (_, row) in enumerate(nuevos_datos.iterrows()):
    pred_label = "Ingreso alto" if predicciones[i] == 1 else "Ingreso normal"
    print(f"  Cliente {i+1}: edad={row['edad']}, ingreso={row['ingreso']:.0f}")
    print(f"    → Predicción: {pred_label} (probabilidad: {probabilidades[i]:.2%})\n")

# Limpiar archivo temporal
os.remove(model_path)
print("✅ Flujo completo: entrenar → guardar → cargar → predecir")
````

4️⃣ ¿Qué se guarda con joblib.dump(pipeline, path)?

- Solo los pesos del modelo
- Solo el preprocesamiento
- El Pipeline completo (preprocesamiento + modelo entrenado) 
- Solo los hiperparámetros


5️⃣ ¿Por qué NO se usan notebooks en producción?

- Porque son muy lentos
- Porque no se pueden ejecutar sin interfaz gráfica de forma confiable 
- Porque no soportan pandas
- Porque son gratuitos


6️⃣ ¿Cuál es el beneficio principal de modularizar?

- El código se ejecuta más rápido
- Cada archivo tiene una responsabilidad clara, es reutilizable y testeable 
- Se usan menos librerías
- No se necesita documentación
