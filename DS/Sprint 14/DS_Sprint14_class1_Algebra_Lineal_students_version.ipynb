## <span style="color: #2826DE;">Vectores: Representación y Operaciones (25 mins)

Un **vector** es simplemente una lista ordenada de números. En ML, cada observación (fila) es un vector.

**Representaciones:**
- Geométrica: una flecha con dirección y magnitud.
- Algebraica: `v = [v₁, v₂, ..., vₙ]`
- En Python: un array de NumPy.

**Operaciones fundamentales:**
- **Suma:** `u + v` → se suman componente a componente.
- **Producto por escalar:** `c × v` → estira o encoge el vector.
- **Magnitud (norma):** `||v|| = √(v₁² + v₂² + ... + vₙ²)` → la "longitud" del vector.
- **Producto punto:** `u · v = u₁v₁ + u₂v₂ + ... + uₙvₙ` → mide "alineación".

La intuición clave: el producto punto es alto cuando dos vectores apuntan en la misma dirección, cero si son perpendiculares, y negativo si van en sentidos opuestos.

Visualicemos vectores en 2D para construir intuición geométrica. Cada vector es una flecha desde el origen.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos vectores
u = np.array([3, 2])
v = np.array([1, 4])
w = u + v  # suma de vectores

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# --- Subplot 1: Vectores individuales ---
ax = axes[0]
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f'u = {u}')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label=f'v = {v}')
ax.set_xlim(-1, 6); ax.set_ylim(-1, 6)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.legend(); ax.set_title('Vectores u y v')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# --- Subplot 2: Suma de vectores (regla del paralelogramo) ---
ax = axes[1]
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5)
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5)
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='green', linewidth=2, label=f'u+v = {w}')
# Paralelogramo
ax.plot([u[0], w[0]], [u[1], w[1]], 'r--', alpha=0.4)
ax.plot([v[0], w[0]], [v[1], w[1]], 'b--', alpha=0.4)
ax.set_xlim(-1, 6); ax.set_ylim(-1, 7)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.legend(); ax.set_title('Suma: u + v')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

# --- Subplot 3: Escalar un vector ---
ax = axes[2]
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f'u = {u}')
ax.quiver(0, 0, 2*u[0], 2*u[1], angles='xy', scale_units='xy', scale=1, color='purple', label=f'2u = {2*u}')
ax.quiver(0, 0, -u[0], -u[1], angles='xy', scale_units='xy', scale=1, color='orange', label=f'-u = {-u}')
ax.set_xlim(-5, 8); ax.set_ylim(-4, 6)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.legend(); ax.set_title('Producto por escalar')
ax.axhline(y=0, color='k', linewidth=0.5); ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()

# Operaciones numéricas
print(f"u = {u}, v = {v}")
print(f"u + v = {u + v}")
print(f"||u|| (magnitud) = {np.linalg.norm(u):.3f}")
print(f"u · v (producto punto) = {np.dot(u, v)}")
```

### Producto Punto: La operación más importante en ML

El **producto punto** merece atención especial porque aparece en TODOS los algoritmos:

| Algoritmo | Uso del producto punto |
|---|---|
| Regresión Lineal | `ŷ = X·β` (predicción = producto punto de features × pesos) |
| KNN | Internamente usa normas (basadas en producto punto) |
| SVM | Kernel trick se basa en productos punto en espacios de alta dimensión |
| Redes Neuronales | Cada neurona calcula un producto punto: `z = w·x + b` |
| Cosine Similarity | Es un producto punto normalizado |

Cuando entrenas un modelo lineal, literalmente estás buscando el vector de pesos `β` cuyo producto punto con cada observación `x` se acerque más al target `y`.

```python
# El producto punto en acción: predicción de regresión lineal
# Si β = [2, -1, 3] y x = [1, 4, 2], la predicción es:
beta = np.array([2, -1, 3])
x = np.array([1, 4, 2])

prediccion = np.dot(beta, x)
print(f"Pesos β = {beta}")
print(f"Features x = {x}")
print(f"Predicción ŷ = β·x = {beta[0]}×{x[0]} + {beta[1]}×{x[1]} + {beta[2]}×{x[2]} = {prediccion}")
print(f"\n💡 Esto es EXACTAMENTE lo que hace sklearn internamente en LinearRegression")
```

## <span style="color: #2826DE;">Distancias Vectoriales: L1, L2, Coseno (20 mins)

En Machine Learning, **medir la distancia entre puntos** es fundamental. KNN clasifica según los vecinos más cercanos. Clustering agrupa puntos cercanos. Pero "cercano" depende de cómo midas.

**Distancia L1 (Manhattan):**
- `d₁(u,v) = |u₁-v₁| + |u₂-v₂| + ... + |uₙ-vₙ|`
- Como caminar por cuadras en una ciudad (solo horizontal/vertical).
- Robusta a outliers.

**Distancia L2 (Euclidiana):**
- `d₂(u,v) = √((u₁-v₁)² + (u₂-v₂)² + ... + (uₙ-vₙ)²)`
- Línea recta entre dos puntos.
- La más usada, pero sensible a diferencias grandes en una dimensión.

**Similitud Coseno:**
- `cos(θ) = (u · v) / (||u|| × ||v||)`
- Mide el ángulo entre vectores (ignora magnitud, solo dirección).
- Rango: -1 (opuestos) a 1 (misma dirección). 0 = perpendiculares.
- Ideal para textos (TF-IDF) y recomendaciones.

**¿Cuándo usar cada una?**
- L2: datos numéricos escalados, KNN.
- L1: datos con outliers o en alta dimensión.
- Coseno: cuando importa la dirección, no la magnitud (NLP, embeddings).

Calculemos las tres distancias entre pares de vectores y visualicemos la diferencia entre L1 y L2.

```python
# Distancias entre vectores
a = np.array([1, 2])
b = np.array([4, 6])

# L1 (Manhattan)
d_l1 = np.sum(np.abs(a - b))

# L2 (Euclidiana)
d_l2 = np.linalg.norm(a - b)

# Coseno (similitud y distancia)
cos_sim = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
cos_dist = 1 - cos_sim

print(f"Vectores: a = {a}, b = {b}")
print(f"Distancia L1 (Manhattan): {d_l1}")
print(f"Distancia L2 (Euclidiana): {d_l2:.3f}")
print(f"Similitud Coseno: {cos_sim:.4f}")
print(f"Distancia Coseno: {cos_dist:.4f}")

# Visualización: L1 vs L2
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(*a, 'bo', markersize=10, label=f'a = {a}')
ax.plot(*b, 'ro', markersize=10, label=f'b = {b}')

# L2: línea recta
ax.plot([a[0], b[0]], [a[1], b[1]], 'g-', linewidth=2, label=f'L2 = {d_l2:.2f}')

# L1: camino por cuadras
ax.plot([a[0], b[0]], [a[1], a[1]], 'purple', linewidth=2, linestyle='--')
ax.plot([b[0], b[0]], [a[1], b[1]], 'purple', linewidth=2, linestyle='--', label=f'L1 = {d_l1}')

ax.set_xlim(0, 6); ax.set_ylim(0, 8)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.legend(fontsize=11); ax.set_title('L1 (Manhattan) vs L2 (Euclidiana)', fontweight='bold')
plt.tight_layout()
plt.show()
```

Veamos cómo la similitud coseno se enfoca en la dirección, no en la magnitud. Dos vectores pueden tener magnitudes muy diferentes pero apuntar en la misma dirección.

```python
# Similitud coseno: dirección vs magnitud
v1 = np.array([1, 2])
v2 = np.array([2, 4])   # misma dirección que v1, doble magnitud
v3 = np.array([2, -1])  # dirección diferente

pairs = [(v1, v2, 'v1 vs v2 (misma dirección)'),
         (v1, v3, 'v1 vs v3 (diferente dirección)')]

for va, vb, label in pairs:
    cos = np.dot(va, vb) / (np.linalg.norm(va) * np.linalg.norm(vb))
    l2 = np.linalg.norm(va - vb)
    print(f"{label}:")
    print(f"  Coseno = {cos:.4f}, L2 = {l2:.4f}")
    print()

print("💡 v1 y v2 tienen coseno=1.0 (misma dirección) aunque L2 > 0")
print("   Esto es clave en NLP: documentos largos y cortos pueden ser 'similares' en tema")
```

## <span style="color: #2826DE;">Matrices como Transformaciones Lineales (20 mins)

Una **matriz** no es solo una tabla de números. Cuando multiplicas un vector por una matriz, estás **transformando** ese vector: rotándolo, estirándolo, reflejándolo o combinando estas operaciones.

**Transformaciones comunes en 2D:**
- **Escala:** Estira o encoge cada eje independientemente.
- **Rotación:** Gira el vector por un ángulo θ.
- **Reflexión:** Espeja respecto a un eje.
- **Shear:** Desplaza una dimensión proporcionalmente a otra.

**¿Por qué importa?**
Cuando `sklearn` ajusta un modelo lineal, internamente está encontrando la matriz de transformación que mejor mapea features → predicciones. PCA encuentra la rotación que maximiza la varianza.

La intuición: una matriz de 2×2 transforma TODOS los puntos del plano simultáneamente, preservando líneas rectas y el origen.

Visualicemos cómo diferentes matrices transforman un conjunto de puntos. Usamos un cuadrado unitario como referencia.

```python
# Transformaciones lineales visualizadas
# Puntos originales: cuadrado unitario + puntos interiores
theta = np.linspace(0, 2*np.pi, 50)
square = np.array([[0,0],[1,0],[1,1],[0,1],[0,0]]).T  # cuadrado
circle = np.vstack([0.5 + 0.3*np.cos(theta), 0.5 + 0.3*np.sin(theta)])

# Matrices de transformación
transformaciones = {
    'Escala (2x, 0.5y)': np.array([[2, 0], [0, 0.5]]),
    'Rotación 45°': np.array([[np.cos(np.pi/4), -np.sin(np.pi/4)],
                               [np.sin(np.pi/4),  np.cos(np.pi/4)]]),
    'Shear': np.array([[1, 0.5], [0, 1]]),
    'Reflexión (eje x)': np.array([[1, 0], [0, -1]])
}

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, (nombre, M) in zip(axes, transformaciones.items()):
    # Original
    ax.plot(square[0], square[1], 'b-', alpha=0.3, linewidth=2, label='Original')
    ax.plot(circle[0], circle[1], 'b-', alpha=0.3)
    # Transformado
    sq_t = M @ square
    ci_t = M @ circle
    ax.plot(sq_t[0], sq_t[1], 'r-', linewidth=2, label='Transformado')
    ax.plot(ci_t[0], ci_t[1], 'r-')
    ax.set_xlim(-2, 3); ax.set_ylim(-2, 2.5)
    ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
    ax.set_title(nombre, fontsize=10, fontweight='bold')
    ax.axhline(y=0, color='k', linewidth=0.3); ax.axvline(x=0, color='k', linewidth=0.3)
    ax.legend(fontsize=8)

plt.suptitle('Transformaciones Lineales: Efecto sobre figuras geométricas', fontweight='bold', fontsize=12)
plt.tight_layout()
plt.show()
```

## <span style="color: #2826DE;">Conexión con Machine Learning (15 mins)

El álgebra lineal es el **lenguaje interno** de Machine Learning. Cada algoritmo que usamos tiene álgebra lineal por dentro:

**1. KNN usa distancias:**
- Para clasificar un punto nuevo, calcula la distancia L2 a todos los puntos de entrenamiento.
- Los K más cercanos "votan" por la clase.

**2. Regresión Lineal = ecuación matricial:**
- `y = Xβ + ε` donde X es la matriz de features y β son los coeficientes.
- La solución analítica es: `β = (XᵀX)⁻¹Xᵀy` (ecuación normal).

**3. PCA = eigenvectores de la covarianza:**
- PCA encuentra las direcciones (eigenvectores) donde los datos tienen más varianza.
- Proyectar sobre estos ejes reduce dimensionalidad preservando información.

**4. Redes Neuronales:**
- Cada capa es una multiplicación matricial seguida de una función no lineal.
- `output = f(Wx + b)` donde W es la matriz de pesos.

No necesitas ser experto en álgebra lineal para usar ML, pero entenderla te permite **diagnosticar problemas** y tomar mejores decisiones.

```python
# Ejemplo: KNN es literalmente "encontrar los vectores más cercanos"
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_moons

# Datos en forma de lunas
X, y = make_moons(n_samples=200, noise=0.25, random_state=42)

# Entrenar KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)

# Visualizar decisiones (basadas en distancias L2)
xx, yy = np.meshgrid(np.linspace(-2, 3, 200), np.linspace(-1.5, 2, 200))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(8, 5))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='black', s=50)
plt.title('KNN: Clasificación basada en distancias L2', fontweight='bold')
plt.xlabel('Feature 1 (dimensión 1 del vector)')
plt.ylabel('Feature 2 (dimensión 2 del vector)')
plt.colorbar(label='Clase')
plt.show()

print("💡 Cada punto es un VECTOR en 2D.")
print("   KNN clasifica midiendo distancias euclidianas a los K vecinos más cercanos.")
```

### La Maldición de la Dimensionalidad

Un concepto crucial al trabajar con vectores de alta dimensión: la **maldición de la dimensionalidad**.

Cuando tienes muchas features (muchas dimensiones):
- Todos los puntos están "lejos" entre sí (las distancias se homogenizan).
- KNN pierde efectividad porque los "vecinos cercanos" ya no son tan cercanos.
- El volumen del espacio crece exponencialmente → necesitas exponencialmente más datos.

**Ejemplo:** En 2D, 100 puntos cubren bien un cuadrado. En 100D, necesitarías 10^100 puntos para la misma densidad.

**Soluciones:**
- Reducir dimensionalidad (PCA, que veremos en la siguiente clase).
- Selección de features (eliminar las irrelevantes).
- Regularización (penalizar complejidad).

Esto conecta directamente con por qué PCA es tan importante: nos permite trabajar en un espacio de dimensiones manejable.

```python
# Demostración: distancias se homogenizan en alta dimensión
dimensiones = [2, 5, 10, 50, 100, 500]
n_points = 200

print("Distancias promedio entre puntos aleatorios por dimensión:")
print("=" * 55)
for d in dimensiones:
    points = np.random.uniform(0, 1, (n_points, d))
    # Calcular distancias entre todos los pares
    from scipy.spatial.distance import pdist
    dists = pdist(points, metric='euclidean')
    print(f"  Dim={d:>3}: media={dists.mean():.3f}, std={dists.std():.3f}, "
          f"ratio std/mean={dists.std()/dists.mean():.4f}")

print("\n💡 A medida que crece la dimensión, la variabilidad relativa de")
print("   las distancias DISMINUYE → todos los puntos están 'igual de lejos'")
print("   → KNN pierde capacidad de discriminar vecinos")
```
