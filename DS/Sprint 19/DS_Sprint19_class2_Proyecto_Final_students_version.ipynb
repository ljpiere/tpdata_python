## <span style="color: #2826DE;">Peer Review Cruzado (15 mins)

El **peer review** es una práctica profesional fundamental. En la industria, tu código y tus análisis SIEMPRE son revisados por colegas.

**Rúbrica de evaluación entre equipos:**

| Criterio | 1 (Bajo) | 2 (Medio) | 3 (Alto) |
|---|---|---|---|
| Claridad del problema | No se entiende | Se entiende parcialmente | Muy claro y enfocado |
| Calidad del análisis | Sin EDA | EDA básico | EDA completo con insights |
| Modelado | Sin modelo | Un modelo sin tuning | Múltiples modelos comparados |
| Presentación | Confusa | Correcta pero técnica | Clara, concisa, orientada a negocio |
| Recomendaciones | No hay | Genéricas | Específicas y accionables |

**Instrucciones:**
1. Cada equipo evalúa al equipo de su derecha.
2. Sean constructivos: lo que hicieron bien + lo que pueden mejorar.
3. Escriban 2-3 oraciones de feedback por criterio.
4. Compartan el feedback directamente (respetuosamente).

**Recuerden:** El peer review no es para destruir, es para ayudar a crecer.

```python
# Rúbrica de evaluación para peer review
rubrica = {
    "EDA y limpieza": {"peso": "20%", "criterios": ["Distribuciones exploradas", "Nulos tratados", "Outliers identificados"]},
    "Feature Engineering": {"peso": "15%", "criterios": ["≥2 features nuevas", "Justificación clara", "Encoding correcto"]},
    "Modelado": {"peso": "25%", "criterios": ["≥2 modelos comparados", "Split correcto", "Tuning básico"]},
    "Evaluación": {"peso": "15%", "criterios": ["Métrica correcta (F1)", "Interpretación de negocio", "Feature importance"]},
    "Documentación": {"peso": "15%", "criterios": ["Código comentado", "Markdown explicativo", "Conclusiones claras"]},
    "Presentación": {"peso": "10%", "criterios": ["5-7 min", "Orientada a negocio", "Visualizaciones claras"]}
}

print("📋 RÚBRICA DE EVALUACIÓN DEL PROYECTO")
print("=" * 55)
for seccion, info in rubrica.items():
    print(f"\n{seccion} ({info['peso']})")
    for criterio in info['criterios']:
        print(f"  □ {criterio}")
print(f"\n{'=' * 55}")
print("Nota mínima aprobatoria: 7/10")
```
