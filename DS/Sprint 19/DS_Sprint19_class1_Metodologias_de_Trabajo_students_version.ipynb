## <span style="color: #2826DE;">CRISP-DM: La Metodología Estándar (25 mins)

**CRISP-DM** (Cross-Industry Standard Process for Data Mining) es la metodología más usada en la industria para proyectos de Data Science. Define 6 fases iterativas:

**1. 📋 Business Understanding (Entendimiento del Negocio)**
- ¿Cuál es el problema de negocio?
- ¿Qué métrica de éxito define el cliente?
- ¿Cómo se usará la solución en la práctica?

**2. 📊 Data Understanding (Entendimiento de los Datos)**
- ¿Qué datos están disponibles?
- ¿Calidad? ¿Volumen? ¿Frecuencia?
- EDA exploratorio inicial.

**3. 🧹 Data Preparation (Preparación de Datos)**
- Limpieza, transformaciones, feature engineering.
- 60-80% del tiempo de un proyecto real se gasta aquí.

**4. 🤖 Modeling (Modelado)**
- Selección y entrenamiento de modelos.
- Iteración de hiperparámetros.

**5.  Evaluation (Evaluación)**
- ¿El modelo cumple el objetivo de negocio?
- ¿Es mejor que la solución actual (baseline)?

**6. 🚀 Deployment (Despliegue)**
- Poner el modelo en producción.
- Monitoreo y mantenimiento.

**Clave:** Es **iterativo**, no lineal. Puedes (y debes) volver a fases anteriores cuando descubres algo nuevo.

Veamos cómo se mapea un proyecto real a las fases de CRISP-DM con un ejemplo concreto.

````python
import pandas as pd

# Ejemplo: Proyecto de predicción de churn en telecomunicaciones
crisp_dm = pd.DataFrame({
    'Fase': ['Business Understanding', 'Data Understanding', 'Data Preparation',
             'Modeling', 'Evaluation', 'Deployment'],
    'Pregunta Clave': [
        '¿Qué clientes cancelarán en 30 días?',
        '¿Qué datos tenemos? ¿Calidad?',
        '¿Cómo limpiamos y creamos features?',
        '¿Qué modelo predice mejor?',
        '¿Supera la solución actual?',
        '¿Cómo lo ponemos en producción?'
    ],
    'Entregable': [
        'Definición del problema + KPI (recall > 0.8)',
        'EDA report + data quality assessment',
        'Dataset limpio + feature engineering',
        'Modelo entrenado + métricas',
        'Report de evaluación + recomendación',
        'API/pipeline en producción + monitoreo'
    ],
    'Tiempo Típico (%)': [5, 10, 40, 20, 15, 10]
})

print("CRISP-DM aplicado a un proyecto de Churn:")
print("=" * 70)
print(crisp_dm.to_string(index=False))
print("\n💡 Nota: ¡la preparación de datos toma el 40% del tiempo!")
print("   Esto es normal en la industria. No te frustres si limpiar datos toma mucho.")
````

## <span style="color: #2826DE;">Agile para Data Science (20 mins)

**Agile** nació en desarrollo de software, pero se adapta bien a Data Science con algunos ajustes:

**Principios Agile aplicados a DS:**
- **Iteraciones cortas (sprints):** En vez de un modelo perfecto en 3 meses, entrega un modelo básico en 2 semanas y mejóralo.
- **MVP de modelo:** Empieza con un baseline simple (LogisticRegression, media) y mejora solo si es necesario.
- **Feedback temprano:** Muestra resultados parciales al stakeholder. Mejor corregir rumbo pronto.
- **Documentación viva:** El README y los notebooks se actualizan constantemente.

**Kanban para DS:**
| Backlog | In Progress | Review | Done |
|---|---|---|---|
| Feature engineering | Modelo XGBoost | Report EDA | Limpieza datos |
| SARIMA para ventas | | | Split temporal |

**Diferencias con software:**
- DS tiene más **incertidumbre**: no sabes si un modelo va a funcionar hasta probarlo.
- Los **experimentos** (probar modelos) son el equivalente a los sprints de desarrollo.
- La **reproducibilidad** es más difícil (depende de datos que cambian).

**Herramientas:**
- Trello/Jira para gestión de tareas.
- Git para versionamiento de código.
- MLflow/Weights & Biases para tracking de experimentos.

## <span style="color: #2826DE;">Estructura del Proyecto Final (20 mins)

Un proyecto profesional de DS debe estar organizado de forma que **cualquier persona** pueda entenderlo y reproducirlo sin preguntarte nada.

**Estructura recomendada:**
```
mi_proyecto/
├── README.md              ← Descripción del proyecto (LO MÁS IMPORTANTE)
├── requirements.txt       ← Dependencias exactas
├── data/
│   ├── raw/              ← Datos originales (NO modificar)
│   └── processed/        ← Datos limpios y transformados
├── notebooks/
│   ├── 01_eda.ipynb      ← Análisis exploratorio
│   ├── 02_modeling.ipynb ← Entrenamiento y evaluación
│   └── 03_results.ipynb  ← Resultados finales
├── src/                   ← Código reutilizable (funciones, clases)
│   ├── data_prep.py
│   └── model.py
└── reports/
    └── presentacion.pdf   ← Presentación para stakeholders
```

**El README debe contener:**
1. Título y descripción del problema.
2. Datos utilizados (fuente, tamaño, periodo).
3. Metodología (pasos seguidos).
4. Resultados principales (métricas, gráficas clave).
5. Cómo reproducir (instrucciones paso a paso).
6. Conclusiones y recomendaciones.

Veamos cómo crear la estructura base de un proyecto y un buen README con Python.

````python
# Creación automática de estructura de proyecto
import os

def crear_estructura_proyecto(nombre_proyecto):
    """Crea la estructura de carpetas para un proyecto de DS."""
    carpetas = [
        f'{nombre_proyecto}/data/raw',
        f'{nombre_proyecto}/data/processed',
        f'{nombre_proyecto}/notebooks',
        f'{nombre_proyecto}/src',
        f'{nombre_proyecto}/reports',
    ]

    for carpeta in carpetas:
        os.makedirs(carpeta, exist_ok=True)

    # README template
    readme = f"""# {nombre_proyecto}

## Descripción
[Describe el problema de negocio que resuelves]

## Datos
- **Fuente:** [De dónde vienen los datos]
- **Periodo:** [Rango de fechas]
- **Tamaño:** [Filas x columnas]

## Metodología
1. EDA y limpieza de datos
2. Feature Engineering
3. Modelado ([modelos usados])
4. Evaluación ([métricas])

## Resultados
- Métrica principal: [valor]
- [Insight clave 1]
- [Insight clave 2]

## Cómo reproducir
```bash
pip install -r requirements.txt
jupyter notebook notebooks/01_eda.ipynb
```

## Autor
[Tu nombre] - [Fecha]
"""
    with open(f'{nombre_proyecto}/README.md', 'w') as f:
        f.write(readme)

    # requirements.txt
    requirements = """pandas==2.1.0
numpy==1.24.0
scikit-learn==1.3.0
matplotlib==3.7.0
seaborn==0.12.0
"""
    with open(f'{nombre_proyecto}/requirements.txt', 'w') as f:
        f.write(requirements)

    print(f"✅ Proyecto '{nombre_proyecto}' creado con éxito!")
    print(f"\nEstructura:")
    for carpeta in carpetas:
        print(f"  📁 {carpeta}/")
    print(f"  📄 {nombre_proyecto}/README.md")
    print(f"  📄 {nombre_proyecto}/requirements.txt")

# Ejemplo (NO ejecutar si no quieres crear carpetas)
# crear_estructura_proyecto('prediccion_churn')
print("Función lista. Descomenta la última línea para crear tu proyecto.")
print("\n💡 Tip: En tu proyecto final, usa esta estructura desde el inicio.")
````

## <span style="color: #2826DE;">Documentación y Reproducibilidad (20 mins)

**Reproducibilidad** = que alguien más (o tú en 6 meses) pueda obtener los mismos resultados ejecutando tu código.

**Buenas prácticas:**

1. **Seeds fijas:** `np.random.seed(42)`, `random_state=42` en todo.
2. **Requirements congelados:** `pip freeze > requirements.txt` con versiones exactas.
3. **Datos versionados:** No modifiques los datos raw. Crea versiones procesadas.
4. **Comentarios útiles:** No "incrementa i en 1", sino "calcula el promedio móvil de 7 días".
5. **Notebooks narrativos:** Cada celda de código precedida por explicación del por qué.
6. **Git:** Commits pequeños y frecuentes con mensajes descriptivos.

**Anti-patrones (lo que NO hacer):**
- ❌ `from utils import *` (¿qué importaste?)
- ❌ Variables tipo `df2`, `model_final_v3_REAL` (usa nombres descriptivos).
- ❌ Código que solo funciona en tu máquina (rutas absolutas, etc.).
- ❌ Notebooks con celdas ejecutadas fuera de orden.
- ❌ "Funciona pero no sé por qué" (entiende antes de entregar).

## <span style="color: #2826DE;">Ejemplo: De la Idea al Código Reproducible

Veamos un ejemplo completo de cómo un proyecto bien organizado se ve en la práctica: código limpio, documentado y reproducible.

````python
# Ejemplo de código bien documentado vs mal documentado

# ❌ MAL: Sin contexto, sin nombres claros, sin documentación
# df2 = df[df.x > 3]
# m = lr.fit(df2[['a','b']], df2.c)
# print(m.score(df2[['a','b']], df2.c))

# ✅ BIEN: Claro, documentado, reproducible
def entrenar_modelo_churn(datos, features, target='is_canceled', test_size=0.2):
    """
    Entrena un modelo de predicción de churn con LogisticRegression.

    Parameters
    ----------
    datos : pd.DataFrame
        Dataset completo con features y target.
    features : list
        Lista de nombres de columnas a usar como features.
    target : str
        Nombre de la columna objetivo.
    test_size : float
        Proporción de datos para test (default 0.2).

    Returns
    -------
    dict : Diccionario con modelo entrenado, métricas y metadata.
    """
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LogisticRegression
    from sklearn.metrics import f1_score
    import datetime

    # Separar features y target
    X = datos[features]
    y = datos[target]

    # Split estratificado
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=y
    )

    # Entrenar modelo
    modelo = LogisticRegression(max_iter=1000, random_state=42)
    modelo.fit(X_train, y_train)

    # Evaluar
    y_pred = modelo.predict(X_test)
    f1 = f1_score(y_test, y_pred)

    return {
        'modelo': modelo,
        'f1_score': f1,
        'n_train': len(X_train),
        'n_test': len(X_test),
        'features': features,
        'fecha_entrenamiento': datetime.datetime.now().isoformat()
    }

# Uso
print("Función documentada lista para usar.")
print("Nota cómo incluye docstring, parámetros tipados y metadata del entrenamiento.")
print("\nEsto facilita la reproducibilidad y el debugging.")
````

**Regla:** Si en 6 meses lees tu historial de commits, ¿podrías entender qué hiciste y por qué?

````python
# Ejemplo: Tracking de experimentos con un diccionario simple
import datetime

# En un proyecto real usarías MLflow, pero para empezar esto funciona
experimentos = []

def registrar_experimento(nombre, modelo, params, metricas):
    """Registra un experimento con metadata útil."""
    exp = {
        'nombre': nombre,
        'modelo': modelo,
        'params': params,
        'metricas': metricas,
        'fecha': datetime.datetime.now().strftime('%Y-%m-%d %H:%M'),
    }
    experimentos.append(exp)
    print(f"✅ Experimento registrado: {nombre}")
    return exp

# Simulación de registro de experimentos
registrar_experimento(
    nombre='baseline_lr',
    modelo='LogisticRegression',
    params={'max_iter': 1000, 'C': 1.0},
    metricas={'f1': 0.72, 'accuracy': 0.78}
)

registrar_experimento(
    nombre='rf_tuned',
    modelo='RandomForestClassifier',
    params={'n_estimators': 200, 'max_depth': 10},
    metricas={'f1': 0.81, 'accuracy': 0.84}
)

import pandas as pd
print("\n📊 Historial de experimentos:")
print(pd.DataFrame(experimentos).to_string(index=False))
print("\n💡 Esto te permite comparar modelos y reproducir el mejor.")
````
