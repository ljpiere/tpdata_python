## <span style="color: #2826DE;">Estrategias de Branching Profesionales (25 mins)

<img src="../../Images/S7_DS_branching.png" width="500">

En un equipo real, **nunca** todos trabajan directamente en `main`. Existen estrategias formales para organizar el trabajo en paralelo sin pisarse el código. La elección de estrategia depende del tamaño del equipo, la frecuencia de releases y la madurez del CI/CD.

### GitFlow: Para equipos grandes con releases programados

GitFlow fue propuesto por Vincent Driessen en 2010 y es el estándar en muchas empresas. Define roles claros para cada tipo de rama:

```
main ─────────────────────────────────── (producción estable)
  └── develop ────────────────────────── (integración continua)
        ├── feature/analisis-eda ──────── (tu trabajo actual)
        ├── feature/modelo-rf ─────────── (trabajo de un compañero)
        └── release/v1.0 ──────────────── (preparar siguiente versión)
  └── hotfix/fix-bug-produccion ──────── (emergencia en producción)
```

**Reglas de GitFlow:**
- `main` → solo código que está en producción. Solo recibe merges de `release` o `hotfix`
- `develop` → rama de integración. Aquí se juntan todas las features terminadas
- `feature/*` → una rama por cada tarea. Se crea desde `develop`, se mergea a `develop`
- `release/*` → se crea cuando `develop` está listo para producción. Se hacen últimos ajustes
- `hotfix/*` → para bugs críticos en producción. Se crea desde `main`, se mergea a `main` Y `develop`

**¿Cuándo usar GitFlow?**
- Equipos de 5+ personas
- Releases programados (cada 2 semanas, cada mes)
- Productos con versiones (v1.0, v1.1, v2.0)
- Cuando necesitas mantener múltiples versiones en producción

### Trunk-Based Development: Para equipos ágiles con CI/CD

Usado por Google, Facebook, Netflix y la mayoría de startups modernas. La idea es simple: **ramas muy cortas y merge frecuente**.

```
main ─────────────────────────────────── (siempre deployable)
  ├── feature/short-lived-1 ──────────── (máximo 1-2 días)
  └── feature/short-lived-2 ──────────── (merge rápido a main)
```

**Reglas de Trunk-Based:**
- Las ramas viven **máximo 1-2 días** (idealmente horas)
- Merge a `main` varias veces al día
- `main` siempre debe estar en estado deployable
- Requiere **CI/CD robusto** (tests automáticos en cada push)
- Feature flags para código incompleto que ya está en main

**¿Cuándo usar Trunk-Based?**
- Equipos pequeños y ágiles (2-5 personas)
- Deployment continuo (varias veces al día)
- Buena cobertura de tests automatizados
- Cultura de code review rápido

### Comparación directa

| Aspecto | GitFlow | Trunk-Based |
|---|---|---|
| Complejidad | Alta (muchas ramas) | Baja (pocas ramas) |
| Velocidad de deploy | Lenta (releases) | Rápida (continuo) |
| Riesgo de conflictos | Alto (ramas largas) | Bajo (ramas cortas) |
| Requiere CI/CD | No necesariamente | Sí, obligatorio |
| Ideal para | Productos enterprise | Startups, SaaS |

## <span style="color: #2826DE;">Rebase Interactivo y Squash (20 mins)

<img src="../../Images/S7_DS_git_rebase.jpg" width="600">

El **rebase** es una de las herramientas más poderosas (y temidas) de Git. Permite **reescribir el historial** de commits para que sea más limpio y legible.

### ¿Por qué reescribir el historial?

Imagina que estás trabajando en una feature y haces estos commits:
```
abc1234 - WIP: empezando análisis
def5678 - fix typo
ghi9012 - ahora sí funciona
jkl3456 - olvidé importar pandas
mno7890 - análisis exploratorio completo
```

Antes de hacer un Pull Request, quieres que tu historial se vea así:
```
xyz1234 - feat: análisis exploratorio del dataset de ventas
```

Eso es exactamente lo que hace el **squash** con rebase interactivo.

### Merge vs Rebase: ¿Cuándo usar cada uno?

| Aspecto | `git merge` | `git rebase` |
|---|---|---|
| Historial | Preserva todo (crea merge commits) | Lineal y limpio |
| Seguridad | No reescribe historial | ⚠️ Reescribe historial |
| Cuándo usar | Ramas compartidas, PRs finales | Ramas locales, antes de PR |
| Resultado visual | Ramas que se juntan | Línea recta |
| Comando | `git merge feature` | `git rebase main` |

### Rebase interactivo paso a paso

```bash
# Rebase interactivo de los últimos 3 commits
git rebase -i HEAD~3
```

Se abre un editor con algo así:
```
pick abc1234 WIP: empezando análisis
pick def5678 fix typo
pick ghi9012 análisis completo
```

Puedes cambiar `pick` por:
| Comando | Qué hace | Cuándo usarlo |
|---|---|---|
| `pick` | Mantener el commit tal cual | Commits que quieres conservar |
| `squash` (o `s`) | Combinar con el commit anterior | Unificar commits relacionados |
| `reword` (o `r`) | Cambiar el mensaje | Mejorar mensajes poco descriptivos |
| `edit` (o `e`) | Pausar para modificar | Separar un commit en dos |
| `drop` (o `d`) | Eliminar el commit | Commits que no deberían existir |
| `fixup` (o `f`) | Como squash pero descarta el mensaje | Fixes menores |

### ⚠️ La regla de oro del rebase

> **NUNCA hagas rebase de commits que ya están en una rama compartida (pushed).**

Si reescribes historial que otros ya descargaron, crearás un caos de commits duplicados. Solo haz rebase de tu trabajo **local** antes de hacer push por primera vez.

## <span style="color: #2826DE;">Resolución Avanzada de Conflictos (20 mins)

Los conflictos de merge son **inevitables** cuando trabajas en equipo. No son un error ni algo que debas temer — son simplemente Git diciéndote: "dos personas cambiaron lo mismo y necesito que tú decidas cuál versión conservar."

### ¿Cuándo ocurren los conflictos?

Un conflicto ocurre cuando:
1. Dos ramas modifican la **misma línea** del **mismo archivo**
2. Una rama elimina un archivo que la otra modificó
3. Dos ramas agregan un archivo con el mismo nombre pero contenido diferente

**No** causan conflicto:
- Dos ramas que modifican archivos diferentes
- Dos ramas que modifican líneas diferentes del mismo archivo
- Agregar archivos nuevos con nombres diferentes

### Anatomía de un conflicto

Cuando Git detecta un conflicto, marca el archivo así:
```python
# Código sin conflicto (arriba)
import pandas as pd
df = pd.read_csv("datos.csv")

<<<<<<< HEAD
# Tu versión (la rama donde estás)
modelo = RandomForestClassifier(n_estimators=100)
=======
# Versión del otro (la rama que intentas mergear)
modelo = GradientBoostingClassifier(n_estimators=200)
>>>>>>> feature/nuevo-modelo

# Código sin conflicto (abajo)
modelo.fit(X_train, y_train)
```

### Estrategia sistemática para resolver conflictos

| Paso | Acción | Comando/Herramienta |
|---|---|---|
| 1 | Ver qué archivos tienen conflictos | `git status` o `git diff --name-only --diff-filter=U` |
| 2 | Entender qué cambió cada persona | Leer ambas versiones, preguntar si es necesario |
| 3 | Decidir la resolución | Elegir una versión, combinar ambas, o reescribir |
| 4 | Editar el archivo | Eliminar marcadores `<<<<`, `====`, `>>>>` |
| 5 | Verificar que funciona | Ejecutar el código, correr tests |
| 6 | Marcar como resuelto | `git add archivo_resuelto.py` |
| 7 | Completar el merge | `git commit` (Git genera el mensaje automáticamente) |

### Herramientas para resolver conflictos

- **VSCode** — La mejor opción. Muestra los conflictos con colores y botones:
  - "Accept Current Change" (tu versión)
  - "Accept Incoming Change" (la otra versión)
  - "Accept Both Changes" (combinar ambas)
  - "Compare Changes" (ver lado a lado)
- **`git mergetool`** — Abre una herramienta visual de 3 paneles
- **`git merge --abort`** — Cancelar el merge y volver al estado anterior (si te pierdes)

### Cómo prevenir conflictos

1. **Pull frecuente** — `git pull` antes de empezar a trabajar cada día
2. **Ramas cortas** — Merge tu feature lo antes posible
3. **Comunicación** — Avisa al equipo si vas a modificar un archivo compartido
4. **Archivos pequeños** — Divide archivos grandes en módulos

## <span style="color: #2826DE;">Cherry-pick, Stash y Bisect (20 mins)

Estas tres herramientas resuelven problemas específicos que encontrarás frecuentemente en tu trabajo diario como Data Scientist.

### `git cherry-pick` — Copiar un commit específico a otra rama

**Situación:** Tu compañero hizo un fix en su rama que tú necesitas, pero su rama todavía no está lista para merge. No quieres hacer merge de toda su rama, solo de ese commit.

```bash
# Copiar el commit abc1234 a tu rama actual
git cherry-pick abc1234

# Si hay conflicto:
git cherry-pick --abort    # Cancelar
git cherry-pick --continue # Después de resolver
```

**Casos de uso en DS:**
- Traer un fix de preprocesamiento de otra rama de experimento
- Copiar la configuración de un modelo que funcionó bien
- Recuperar un commit que eliminaste por error

---

### `git stash` — Guardar trabajo temporal sin hacer commit

**Situación:** Estás a mitad de un análisis, pero tu jefe te pide revisar algo urgente en otra rama. No quieres hacer commit de código incompleto, pero tampoco quieres perder tu trabajo.

```bash
# Guardar cambios actuales en el stash
git stash push -m "WIP: feature engineering para modelo de churn"

# Ver la lista de stashes guardados
git stash list
# stash@{0}: On feature/churn: WIP: feature engineering para modelo de churn
# stash@{1}: On main: experimento con nuevos hiperparámetros

# Cambiar de rama, hacer lo urgente...
git checkout main
# ... trabajo urgente ...

# Volver a tu rama y recuperar el stash
git checkout feature/churn
git stash pop              # Recupera Y elimina del stash
# o
git stash apply stash@{0}  # Recupera pero NO elimina (por si acaso)
```

**¿Por qué es mejor que un commit "WIP"?**
- No ensucia el historial con commits sin significado
- Puedes tener múltiples stashes organizados con mensajes
- Es temporal por diseño — no se sube a GitHub

---

### `git bisect` — Encontrar qué commit introdujo un bug

**Situación:** Tu pipeline de ML funcionaba perfecto hace una semana, pero hoy falla. Hubo 50 commits en ese tiempo. ¿Cuál rompió todo? En lugar de revisar uno por uno, `bisect` usa **búsqueda binaria** para encontrar el culpable en ~6 pasos (log₂(50) ≈ 6).

```bash
# Iniciar la búsqueda
git bisect start

# Marcar el estado actual como "malo" (tiene el bug)
git bisect bad

# Marcar un commit antiguo donde todo funcionaba
git bisect good abc1234

# Git te lleva a un commit intermedio. Tú pruebas y dices:
git bisect good   # Si funciona aquí
git bisect bad    # Si el bug ya existe aquí

# Repite hasta que Git te diga cuál es el primer commit "malo"
# Resultado: "abc5678 is the first bad commit"

# Terminar la búsqueda
git bisect reset
```

**¿Por qué es útil para DS?**
- Encontrar qué cambio en el preprocesamiento rompió el modelo
- Identificar cuándo se introdujo un data leak
- Detectar qué commit cambió los resultados de un experimento
