## <span style="color: #2826DE;">¿Sabías que ya usas OOP todos los días? (10 mins)

Cada vez que escribes código en Python para Data Science, estás usando Programación Orientada a Objetos sin darte cuenta. Veamos algunos ejemplos:

```python
# Esto es OOP: df es un OBJETO (instancia de la CLASE DataFrame)
df = pd.DataFrame({"col1": [1, 2, 3]})
df.describe()    # .describe() es un MÉTODO del objeto df
df.shape         # .shape es un ATRIBUTO del objeto df

# Esto también es OOP: model es un OBJETO
model = RandomForestClassifier(n_estimators=100)
model.fit(X, y)      # .fit() es un MÉTODO
model.feature_importances_  # es un ATRIBUTO
```

**La OOP es simplemente una forma de organizar código** agrupando:
- **Datos** (atributos) — la información que el objeto almacena
- **Funciones** (métodos) — las acciones que el objeto puede realizar

...en una misma estructura llamada **clase**.

### ¿Por qué importa para un Data Scientist?

| Sin OOP | Con OOP |
|---|---|
| Funciones sueltas por todos lados | Código organizado en clases |
| Variables globales peligrosas | Cada objeto tiene su propio estado |
| Difícil de reutilizar | Creas una vez, usas muchas veces |
| Difícil de testear | Cada clase se testea independientemente |
| Código espagueti | Estructura clara y predecible |

### La analogía fundamental: Clase = Plano, Instancia = Casa

Imagina que eres un arquitecto:
- El **plano** (clase) define la estructura: cuántas habitaciones, dónde va la cocina, etc.
- Cada **casa construida** (instancia) es un objeto real basado en ese plano
- Puedes construir muchas casas con el mismo plano, cada una con sus propios colores y muebles

En Python:
- La **clase** define qué atributos y métodos tendrá cada objeto
- La **instancia** es un objeto concreto creado a partir de esa clase
- Puedes crear muchas instancias de la misma clase, cada una con sus propios datos

### Vocabulario esencial

| Término | Significado | Ejemplo |
|---|---|---|
| **Clase** | El plano/molde/plantilla | `class Estudiante:` |
| **Instancia** | Un objeto concreto creado | `ana = Estudiante("Ana", 7)` |
| **Atributo** | Un dato que pertenece al objeto | `ana.nombre` → "Ana" |
| **Método** | Una función que pertenece al objeto | `ana.avanzar_sprint()` |
| **Constructor** | Método que se ejecuta al crear | `__init__(self, ...)` |
| **`self`** | Referencia a la instancia actual | Siempre es el primer parámetro |

### Convenciones de nombres
- Clases: **PascalCase** → `DataPipeline`, `ModeloPrediccion`, `Estudiante`
- Atributos y métodos: **snake_case** → `nombre_completo`, `calcular_promedio()`
- Métodos privados: **_prefijo** → `_validar_datos()` (no se debe llamar desde fuera)

````python
# === DEFINIR UNA CLASE ===

class Estudiante:
    """Representa a un estudiante del bootcamp de Data Science.

    Atributos:
        nombre (str): Nombre del estudiante
        sprint_actual (int): Sprint en el que se encuentra
        proyectos_completados (list): Lista de proyectos terminados
    """

    # ATRIBUTO DE CLASE: compartido por TODAS las instancias
    bootcamp = 'TripleTen Data Science'

    def __init__(self, nombre, sprint_actual):
        """Constructor: se ejecuta automáticamente al crear una instancia.

        Args:
            nombre: Nombre del estudiante
            sprint_actual: Número del sprint actual
        """
        # ATRIBUTOS DE INSTANCIA: únicos para cada objeto
        self.nombre = nombre
        self.sprint_actual = sprint_actual
        self.proyectos_completados = []  # Lista vacía al inicio

    def __repr__(self):
        """Representación legible del objeto (para debugging y print)."""
        return f'Estudiante({self.nombre}, Sprint {self.sprint_actual})'

    def completar_proyecto(self, nombre_proyecto):
        """Registra un proyecto completado.

        Args:
            nombre_proyecto: Nombre del proyecto terminado
        """
        self.proyectos_completados.append(nombre_proyecto)
        print(f'  ✅ {self.nombre} completó: {nombre_proyecto}')

    def avanzar_sprint(self):
        """Avanza al siguiente sprint del bootcamp."""
        self.sprint_actual += 1
        print(f'  🚀 {self.nombre} avanzó al Sprint {self.sprint_actual}')

    def resumen(self):
        """Muestra un resumen del progreso del estudiante."""
        print(f'  📊 {self.nombre} | Sprint {self.sprint_actual} | '
              f'{len(self.proyectos_completados)} proyectos')


# === CREAR INSTANCIAS (objetos concretos) ===
ana = Estudiante('Ana García', 7)
luis = Estudiante('Luis Pérez', 5)

# Usar __repr__
print('Instancias creadas:')
print(f"  {ana}")
print(f"  {luis}")

# Acceder a atributos
print(f'\nAtributo de clase: {ana.bootcamp}')
print(f'Atributo de instancia: {ana.nombre}')

# Usar métodos
print('\nUsando métodos:')
ana.completar_proyecto('EDA de ventas')
ana.completar_proyecto('Modelo de clasificación')
ana.avanzar_sprint()
ana.resumen()
````

## <span style="color: #2826DE;">Herencia y los 4 Principios OOP (25 mins)

La **herencia** es el mecanismo que permite crear clases nuevas basadas en clases existentes. La clase hija **hereda** todos los atributos y métodos del padre, y puede agregar nuevos o modificar los existentes.

### ¿Por qué es útil la herencia?

Imagina que tienes un `DataPipeline` básico que carga y limpia datos. Ahora necesitas un pipeline para ML que además separe features y entrene modelos. En lugar de reescribir todo desde cero, **heredas** del pipeline básico y solo agregas lo nuevo.

### Los 4 principios fundamentales de OOP

| Principio | Qué significa | Ejemplo en Data Science |
|---|---|---|
| **Encapsulación** | Ocultar detalles internos, exponer solo lo necesario | `model.fit()` oculta todo el algoritmo de optimización |
| **Abstracción** | Interfaz simple para operaciones complejas | `pd.read_csv()` maneja encoding, tipos, separadores, nulos... |
| **Herencia** | Reutilizar código de clases padre | Todos los modelos de sklearn heredan de `BaseEstimator` |
| **Polimorfismo** | Misma interfaz, diferente comportamiento | `.fit()` y `.predict()` funcionan en CUALQUIER modelo |

**El polimorfismo** es la razón por la que puedes escribir:
```python
for modelo in [RandomForest(), GradientBoosting(), LogisticRegression()]:
    modelo.fit(X_train, y_train)      # Misma interfaz
    score = modelo.score(X_test, y_test)  # Diferente implementación interna
```
Sin importar qué modelo sea, la interfaz es la misma. Eso es polimorfismo.

Veamos un ejemplo práctico: construir un sistema de pipelines con herencia.

````python
import pandas as pd
import numpy as np

# === CLASE BASE: Pipeline genérico de datos ===
class DataPipeline:
    """Pipeline base para procesamiento de datos.

    Proporciona funcionalidad básica de carga, limpieza y descripción.
    Las clases hijas pueden extender con funcionalidad específica.
    """

    def __init__(self, nombre):
        self.nombre = nombre
        self.datos = None
        self.log = []  # Registro de operaciones

    def cargar(self, datos):
        """Carga datos en el pipeline (siempre trabaja con copia)."""
        self.datos = datos.copy()
        self._registrar(f'Datos cargados: {self.datos.shape}')
        return self  # Permite encadenar: pipe.cargar(df).limpiar()

    def limpiar(self):
        """Limpieza básica: elimina duplicados y filas con nulos."""
        antes = len(self.datos)
        self.datos = self.datos.drop_duplicates()
        self.datos = self.datos.dropna()
        despues = len(self.datos)
        self._registrar(f'Limpieza: {antes} → {despues} filas ({antes-despues} eliminadas)')
        return self

    def describir(self):
        """Muestra un resumen estadístico de los datos."""
        print(f'\n{'='*50}')
        print(f'📊 Pipeline: {self.nombre}')
        print(f'   Shape: {self.datos.shape}')
        print(f'   Columnas: {list(self.datos.columns)}')
        print(f'   Nulos: {self.datos.isnull().sum().sum()}')
        print(f'{'='*50}')
        return self

    def _registrar(self, mensaje):
        """Método PRIVADO para logging interno (prefijo _ = no llamar desde fuera)."""
        self.log.append(mensaje)
        print(f'  [{self.nombre}] {mensaje}')


# === CLASE HIJA: Extiende DataPipeline para ML ===
class MLPipeline(DataPipeline):
    """Pipeline extendido para Machine Learning.

    Hereda carga y limpieza del padre, agrega separación de features.
    """

    def __init__(self, nombre, target):
        # super() llama al __init__ del padre (DataPipeline)
        super().__init__(nombre)
        self.target = target  # Atributo nuevo, solo de MLPipeline
        self.X = None
        self.y = None

    def separar_features_target(self):
        """Separa el DataFrame en features (X) y target (y)."""
        self.X = self.datos.drop(columns=[self.target])
        self.y = self.datos[self.target]
        self._registrar(f'Separado: {self.X.shape[1]} features, target={self.target}')
        return self


# === USO: Encadenamiento de métodos ===
datos = pd.DataFrame({
    'edad': [25, 30, 35, 40, 25, np.nan, 45, 30],
    'salario': [40000, 55000, 60000, 75000, 40000, 50000, 80000, 55000],
    'compra': [0, 1, 1, 1, 0, 0, 1, 1]
})

print('=== Ejecutando MLPipeline ===')
pipe = MLPipeline('Predicción de compra', target='compra')
pipe.cargar(datos).limpiar().separar_features_target().describir()

print(f'\n📝 Log completo del pipeline:')
for i, entry in enumerate(pipe.log, 1):
    print(f'   {i}. {entry}')
````

## <span style="color: #2826DE;">Tips y buenas prácticas

- **Una clase, una responsabilidad** — no metas todo en una sola clase gigante
- **Nombres descriptivos** — `DataPipeline` es mejor que `DP` o `MiClase`
- **Docstrings siempre** — documenta qué hace cada clase y método
- **`return self`** — permite encadenar métodos (`pipe.cargar().limpiar().describir()`)
- **Prefijo `_`** — para métodos internos que no deben llamarse desde fuera
- **Herencia con moderación** — no crees jerarquías de más de 2-3 niveles
