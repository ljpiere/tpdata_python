## <span style="color: #2826DE;">HTML Básico: Estructura y Etiquetas (15 mins)

Antes de poder extraer datos de una página web, necesitamos entender **cómo está construida**. Todas las páginas web están escritas en **HTML** (HyperText Markup Language), un lenguaje de etiquetas que define la estructura y el contenido de la página.

### ¿Qué es HTML?

HTML organiza el contenido en **elementos** delimitados por etiquetas de apertura y cierre. Piensa en HTML como un árbol: cada elemento puede contener otros elementos dentro de sí.

### Estructura básica de una página web

```html
<html>
  <head>
    <title>Título de la pestaña</title>
  </head>
  <body>
    <h1>Título principal visible</h1>
    <p>Un párrafo de texto.</p>
    <div class="productos">
      <div class="producto">
        <span class="nombre">Laptop HP</span>
        <span class="precio">$15,000</span>
        <a href="/laptop-hp">Ver detalle</a>
      </div>
    </div>
  </body>
</html>
```

### Etiquetas más importantes para scraping

| Etiqueta | Qué representa | Ejemplo de uso en scraping |
|---|---|---|
| `<div>` | Contenedor genérico | Agrupa productos, secciones |
| `<span>` | Texto inline | Precios, nombres, etiquetas |
| `<a>` | Enlace (link) | URLs de detalle, paginación |
| `<table>`, `<tr>`, `<td>` | Tabla | Datos tabulares (estadísticas, rankings) |
| `<ul>`, `<li>` | Lista | Categorías, menús, características |
| `<img>` | Imagen | URLs de imágenes de productos |
| `<h1>`-`<h6>` | Títulos | Nombres de productos, secciones |
| `<p>` | Párrafo | Descripciones, reseñas |

### Atributos clave

Los elementos HTML tienen **atributos** que nos ayudan a identificarlos:
- `class="nombre-clase"` — identifica un grupo de elementos similares (el más usado en scraping)
- `id="identificador"` — identifica un elemento único en la página
- `href="url"` — la URL de destino en un enlace `<a>`
- `src="url"` — la URL de una imagen `<img>`

### Herramienta clave: DevTools (F12)

Todos los navegadores tienen **Developer Tools** (F12 o clic derecho → Inspeccionar). Con esta herramienta puedes:
1. Ver la estructura HTML completa de cualquier página
2. Identificar las clases y atributos de los elementos que quieres extraer
3. Probar selectores CSS antes de escribir código
4. Ver las peticiones de red que hace la página

## <span style="color: #2826DE;">Requests: Obtener Páginas Web con Python (15 mins)

La librería **Requests** es la herramienta estándar en Python para hacer peticiones HTTP. Cuando visitas una página web en tu navegador, este hace una petición HTTP GET al servidor y recibe el HTML como respuesta. Con Requests, hacemos exactamente lo mismo pero desde código.

### ¿Cómo funciona HTTP?

HTTP (HyperText Transfer Protocol) es el protocolo de comunicación de la web:

```
Tu código (cliente)  →  Petición GET  →  Servidor web
Tu código (cliente)  ←  Respuesta HTML ←  Servidor web
```

### Tipos de peticiones HTTP
| Método | Qué hace | Cuándo se usa en scraping |
|---|---|---|
| `GET` | Solicita datos | **El más común** — obtener páginas |
| `POST` | Envía datos | Formularios, login, búsquedas |
| `HEAD` | Solo headers | Verificar si una URL existe |

### Status codes: ¿Qué respondió el servidor?

Cada respuesta viene con un **código de estado** que indica si la petición fue exitosa:

| Código | Significado | Qué hacer |
|---|---|---|
| **200** | OK — todo bien | Proceder a parsear el HTML |
| **301/302** | Redirección | Requests lo maneja automáticamente |
| **403** | Forbidden — acceso denegado | Agregar headers (User-Agent) |
| **404** | Not Found — no existe | Verificar la URL |
| **429** | Too Many Requests | Esperar más tiempo entre peticiones |
| **500** | Error del servidor | No es tu culpa, reintentar después |

**Regla:** Siempre verifica el status code antes de intentar parsear el HTML.

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# === HACER UNA PETICIÓN HTTP GET ===
url = 'https://books.toscrape.com/'
response = requests.get(url)

# Verificar que la petición fue exitosa
print(f'Status code: {response.status_code}')  # 200 = OK
print(f'Encoding: {response.encoding}')
print(f'Tamaño del HTML: {len(response.text):,} caracteres')
print(f'Headers del servidor:')
for key in ['Content-Type', 'Server']:
    print(f'  {key}: {response.headers.get(key, "N/A")}')

# Ver los primeros 300 caracteres del HTML recibido
print('\n--- Primeros 300 chars del HTML ---')
print(response.text[:300])
````

### Headers personalizados

Algunos sitios bloquean peticiones que no parecen venir de un navegador real. Para evitarlo, puedes agregar un **User-Agent**:

```python
headers = {

```

Esto le dice al servidor que tu petición viene de un navegador Chrome en Windows.

## <span style="color: #2826DE;">BeautifulSoup: Parsear y Extraer Datos (30 mins)

**BeautifulSoup** es una librería que convierte el HTML crudo (un string largo y desordenado) en un **árbol de objetos** que puedes navegar y buscar fácilmente. Es como tener un mapa del HTML en lugar de leerlo línea por línea.

### El flujo de trabajo de scraping

```
1. requests.get(url)     → Obtener el HTML crudo (string)
2. BeautifulSoup(html)   → Convertir a árbol navegable
3. soup.find_all(...)    → Buscar los elementos que necesitas
4. extraer datos         → .get_text(), ["href"], etc.
5. pd.DataFrame(datos)   → Organizar en tabla para análisis
```

### Métodos principales

| Método | Qué hace | Cuándo usarlo |
|---|---|---|
| `find(tag, attrs)` | Encuentra el **primer** elemento | Cuando solo hay uno (título, precio único) |
| `find_all(tag, attrs)` | Encuentra **todos** los elementos | Cuando hay muchos (lista de productos) |
| `select(css)` | Usa **selectores CSS** | Cuando la estructura es compleja |
| `.get_text()` | Extrae el **texto** del elemento | Para obtener el contenido visible |
| `["atributo"]` | Extrae un **atributo** | Para URLs (`href`), clases, IDs |
| `.parent` | Elemento padre | Para navegar hacia arriba en el árbol |
| `.children` | Elementos hijos | Para navegar hacia abajo |

### Selectores CSS vs find()

```python
# Equivalentes:
soup.find('div', class_='producto')           # find con atributo
soup.select_one('div.producto')               # CSS selector

soup.find_all('span', class_='precio')        # find_all
soup.select('span.precio')                    # CSS selector

# CSS selectors son más potentes para estructuras complejas:
soup.select('div.producto > span.precio')     # hijo directo
soup.select('table tr td:nth-child(2)')       # segunda columna
```

Veamos un ejemplo completo. Vamos a extraer todos los libros de la página principal de books.toscrape.com. Cada libro tiene: título, precio, rating y disponibilidad.

````python
# === PARSEAR HTML CON BEAUTIFULSOUP ===

# Crear el objeto soup a partir del HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Primero, exploremos la estructura
print('Título de la página:', soup.title.get_text())
print('Número de links:', len(soup.find_all('a')))
print('Número de artículos:', len(soup.find_all('article')))

# Cada libro está dentro de un <article class="product_pod">
articulos = soup.find_all('article', class_='product_pod')
print(f'\nLibros encontrados: {len(articulos)}')

# Veamos la estructura de UN artículo para entenderla
print('\n--- Estructura del primer artículo ---')
print(articulos[0].prettify()[:500])
````

Ahora que entendemos la estructura, extraigamos los datos de cada libro de forma sistemática:

````python
# === EXTRAER DATOS DE TODOS LOS LIBROS ===

libros = []

for articulo in articulos:
    # 1. Título: está en el atributo "title" del link dentro de <h3>
    titulo = articulo.h3.a['title']

    # 2. Precio: está en <p class="price_color">
    precio_texto = articulo.find('p', class_='price_color').get_text()

    # 3. Rating: está codificado en la clase CSS del <p class="star-rating">
    #    Ejemplo: <p class="star-rating Three"> → rating = "Three"
    rating_texto = articulo.find('p', class_='star-rating')['class'][1]

    # 4. Disponibilidad: si existe <p class="instock">
    disponible = articulo.find('p', class_='instock') is not None

    libros.append({
        'titulo': titulo,
        'precio': precio_texto,
        'rating': rating_texto,
        'disponible': disponible
    })

# Convertir a DataFrame para análisis
df_libros = pd.DataFrame(libros)
print(f'Libros extraídos: {len(df_libros)}')
print(f'Columnas: {list(df_libros.columns)}')
print()
df_libros.head(10)
````

### Limpieza de datos scrapeados

Los datos extraídos casi siempre necesitan limpieza. En nuestro caso:
- El precio viene como string con símbolo de moneda (`Â£51.77`)
- El rating viene como texto (`Three`) en lugar de número

Esta limpieza es parte fundamental del proceso de scraping.

````python
# === LIMPIAR LOS DATOS SCRAPEADOS ===

# Convertir precio: quitar símbolo £ y convertir a float
df_libros['precio_num'] = (df_libros['precio']
    .str.replace('Â£', '', regex=False)
    .astype(float))

# Convertir rating de texto a número
rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
df_libros['rating_num'] = df_libros['rating'].map(rating_map)

print('Datos limpios:')
print(df_libros[['titulo', 'precio_num', 'rating_num']].head())
print(f'\nPrecio promedio: £{df_libros["precio_num"].mean():.2f}')
print(f'Rating promedio: {df_libros["rating_num"].mean():.1f} estrellas')
````

## <span style="color: #2826DE;">Paginación: Scraping de Múltiples Páginas (15 mins)

En la mayoría de sitios web, los datos están distribuidos en **múltiples páginas**. Por ejemplo, books.toscrape.com tiene 1000 libros distribuidos en 50 páginas de 20 libros cada una.

### ¿Cómo funciona la paginación?

Los sitios web usan patrones de URL predecibles para sus páginas:
- `sitio.com/page/1`, `sitio.com/page/2`, `sitio.com/page/3`...
- `sitio.com/productos?page=1`, `sitio.com/productos?page=2`...
- `sitio.com/catalogue/page-1.html`, `sitio.com/catalogue/page-2.html`...

**Estrategia:** Identificar el patrón de URL y usar un bucle `for` para recorrer todas las páginas.

### Consideraciones importantes

1. **`time.sleep()`** — Esperar entre peticiones para no sobrecargar el servidor
2. **Verificar status code** — Si una página devuelve error, detener el scraping
3. **Detectar la última página** — No todas las páginas existen; verificar si hay contenido
4. **Guardar progreso** — En scraping largo, guardar datos parciales por si falla

````python
# === SCRAPING DE MÚLTIPLES PÁGINAS ===

todos_los_libros = []
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'

# Scrapear las primeras 3 páginas (para la demo)
for pagina in range(1, 4):
    # Construir la URL de esta página
    url = base_url.format(pagina)

    # Hacer la petición
    response = requests.get(url)

    # Verificar que fue exitosa
    if response.status_code != 200:
        print(f'⚠️ Error en página {pagina}: status {response.status_code}')
        break

    # Parsear el HTML
    soup = BeautifulSoup(response.text, 'html.parser')
    articulos = soup.find_all('article', class_='product_pod')

    # Extraer datos de cada libro
    for art in articulos:
        todos_los_libros.append({
            'titulo': art.h3.a['title'],
            'precio': art.find('p', class_='price_color').get_text(),
            'rating': art.find('p', class_='star-rating')['class'][1]
        })

    print(f'  ✅ Página {pagina}: {len(articulos)} libros extraídos')

    # IMPORTANTE: esperar 1 segundo entre peticiones
    # Esto es respetuoso con el servidor y evita que te bloqueen
    time.sleep(1)

# Resultado final
df_todos = pd.DataFrame(todos_los_libros)
print(f'\n📊 Total libros scrapeados: {len(df_todos)}')
print(f'   Páginas procesadas: 3')
print(f'   Tiempo aproximado: 3 segundos (1s por página)')
df_todos.head()
````

## <span style="color: #2826DE;">Ética y Buenas Prácticas del Web Scraping (10 mins)

El web scraping es una herramienta poderosa, pero con poder viene responsabilidad. No todo lo que **puedes** scrapear **debes** scrapearlo.

### Reglas del scraping responsable

1. **Revisa `robots.txt`** — Este archivo (en `sitio.com/robots.txt`) indica qué partes del sitio el dueño permite o prohíbe scrapear. Respétalo siempre.

2. **Respeta los rate limits** — Usa `time.sleep()` entre peticiones. Una regla general: no más de 1 petición por segundo.

3. **No scrapees datos privados** — Solo información que es públicamente visible. Nunca intentes acceder a datos detrás de login sin autorización.

4. **Revisa los Términos de Servicio** — Algunos sitios prohíben explícitamente el scraping.

5. **Usa APIs cuando existan** — Si el sitio ofrece una API, úsala. Es más estable, legal y eficiente.

6. **Identifícate** — Usa un User-Agent descriptivo que incluya tu email, para que el dueño del sitio pueda contactarte si hay problemas.

7. **No sobrecargues el servidor** — Si tu scraping causa problemas al sitio, detente.

### ¿Cuándo NO hacer scraping?

| Situación | Alternativa |
|---|---|
| El sitio tiene API pública | Usar la API |
| Los datos están en un dataset público | Descargar el dataset |
| Los ToS lo prohíben explícitamente | Buscar otra fuente |
| Los datos son privados/sensibles | No hacerlo |
| El sitio es muy pequeño y tu scraping lo afectaría | Ser muy cuidadoso |

````python
# === VERIFICAR robots.txt ===
# Siempre revisa esto ANTES de scrapear un sitio nuevo

robots_url = 'https://books.toscrape.com/robots.txt'
robots_response = requests.get(robots_url)

if robots_response.status_code == 200:
    print('Contenido de robots.txt:')
    print(robots_response.text if robots_response.text.strip() else '(vacío - todo permitido)')
else:
    print(f'No se encontró robots.txt (status: {robots_response.status_code})')
    print('Esto generalmente significa que no hay restricciones explícitas.')
````
