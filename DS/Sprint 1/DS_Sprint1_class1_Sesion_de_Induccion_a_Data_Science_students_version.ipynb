## <span style="color: #2826DE;">Visualizaciones en python (15 mins)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Distribución de Edad**

**Distribución de Horas de Sueño**

Discusión:

**Sueño vs Bebida**

```python
plt.figure()
sns.boxplot(x="Bebida", y="Horas_sueno", data=df)
plt.title("Horas de sueño según bebida favorita")
plt.show()
```

Discusión:

¿El café parece relacionarse con menos sueño?

**Participación por País**

> **⚠️ Nota sobre visualización:** Los gráficos de torta (pie charts) no se recomiendan porque es difícil comparar ángulos entre sí. Un gráfico de barras permite comparar alturas de forma mucho más intuitiva y precisa. Como alternativa al área, también se puede usar un **Tree Map**.

```python
conteo = df['Pais'].value_counts()

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(conteo.index, conteo.values, color=sns.color_palette('pastel', len(conteo)))
ax.set_title('Participación por País')
ax.set_xlabel('País')
ax.set_ylabel('Cantidad de estudiantes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Relación Edad vs Sueño**

```python
plt.figure()
sns.scatterplot(x="Edad", y="Horas_sueno", data=df)
plt.title("Edad vs Horas de sueño")
plt.show()
```

## <span style="color: #2826DE;">Mini IA – Clustering (15 mins)

<img src="../../Images/S1_DS_projectVShardware.jpeg" width="800">

### <span style="color: #2772F2;">Vamos a usar IA para ver si existen tipos de estudiantes según edad y sueño.

```python
from sklearn.cluster import KMeans

X = df[["Edad", "Horas_sueno"]]

kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)

df.head()
```

```python
plt.figure()
sns.scatterplot(x="Edad", y="Horas_sueno", hue="Cluster", data=df)
plt.title("IA detectando patrones en la clase")
plt.show()
```

Discusión guiada:

- ¿Qué grupos aparecen?

- ¿La IA predijo algo o solo agrupó?
