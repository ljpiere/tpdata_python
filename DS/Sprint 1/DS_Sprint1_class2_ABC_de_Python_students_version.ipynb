## <span style="color: #2826DE;">Variables: tipos y convenciones (15 mins)

### <span style="color: #2772F2;">¿Qué es una variable?

<img src="../../Images/S1_DS_variable.gif" width="600">

Una variable es un **contenedor con nombre** que almacena un valor en memoria.

Piénsala como una **caja etiquetada**: la etiqueta es el nombre, el contenido es el valor.

En Python, crear una variable es tan simple como escribir `nombre = valor`. No necesitas declarar el tipo — Python lo detecta automáticamente.

Los **4 tipos básicos** que usarás todo el tiempo en Data Science son:
- `str` → texto (nombres, categorías, etiquetas)
- `int` → números enteros (edad, conteos)
- `float` → números decimales (precios, medidas)
- `bool` → verdadero o falso (flags, condiciones)

Ejecuta el siguiente código y observa cómo Python identifica el tipo de cada variable:

```python
# Tipos básicos de variables
nombre = "Ana"          # str (texto) — siempre entre comillas
edad = 28               # int (entero) — sin comillas, sin punto decimal
altura = 1.65           # float (decimal) — con punto decimal
es_estudiante = True    # bool — solo True o False (con mayúscula)

# type() nos dice qué tipo de dato es cada variable
print(type(nombre))        # <class 'str'>
print(type(edad))          # <class 'int'>
print(type(altura))        # <class 'float'>
print(type(es_estudiante)) # <class 'bool'>
```

### <span style="color: #2772F2;">Convenciones de nombres

En Python usamos **snake_case** para variables:

|  Correcto | Opcional |
|---|---|
| `nombre_completo` | `NombreCompleto` |
| `horas_sueno` | `horasSueno` |
| `precio_total` | `PrecioTotal` |

**Reglas:**
- Solo letras, números y guiones bajos
- No puede empezar con número
- No usar palabras reservadas (`for`, `if`, `class`, etc.)

```python
# Asignación múltiple
x = y = z = 0
print(x, y, z)

# Asignación en una línea
a, b, c = 1, 2, 3
print(a, b, c)

# Intercambiar valores (Pythonic!)
a, b = b, a
print(a, b)
```

Ahora que ya sabemos crear variables, vamos a explorar los **strings** con más detalle.

Los strings son cadenas de texto y tienen muchos métodos útiles que usarás constantemente en Data Science para limpiar nombres de columnas, extraer información de texto, y estandarizar datos.

### En esta ocasión, quien compartirá resultados será:

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random

rules = [
    "El más joven",
    "El que vive más lejos",
    "El más alto",
    "El más deportista",
    "Más experiencia programando",
    "El que durmió más tarde",
    "Más café hoy",
    "Más cursos terminados",
    "Username más largo",
    "Más mascotas"
]

n = len(rules)

angles = np.linspace(0, 2*np.pi, n, endpoint=False)

fig, ax = plt.subplots(figsize=(6,6))

ax.set_aspect("equal")
ax.axis("off")

colors = plt.cm.tab20(np.linspace(0,1,n))

wedges = []

for i, angle in enumerate(angles):
    wedge = plt.matplotlib.patches.Wedge(
        (0,0),
        1,
        np.degrees(angle),
        np.degrees(angle + 2*np.pi/n),
        facecolor=colors[i]
    )
    ax.add_patch(wedge)

    x = 0.7*np.cos(angle + np.pi/n)
    y = 0.7*np.sin(angle + np.pi/n)

    ax.text(
        x, y,
        rules[i],
        ha="center",
        va="center",
        fontsize=9,
        rotation=np.degrees(angle + np.pi/n)
    )

pointer, = ax.plot([0,0],[0,1.1], lw=4)

spin_speed = 0.3
angle = 0

chosen_index = random.randint(0, n-1)

target_angle = 2*np.pi*(5) + angles[chosen_index]

def update(frame):
    global angle

    if angle < target_angle:
        angle += spin_speed

    pointer.set_data(
        [0, np.cos(angle)],
        [0, np.sin(angle)]
    )

    return pointer,

ani = animation.FuncAnimation(
    fig,
    update,
    frames=200,
    interval=30,
    blit=True
)

plt.show()

print("Elegido:", rules[chosen_index])
```
