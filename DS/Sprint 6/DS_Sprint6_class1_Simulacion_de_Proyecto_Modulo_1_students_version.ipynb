## <span style="color: #2826DE;">Dataset: Ventas de una Tienda Online (10 mins)

Trabajaremos con un dataset simulado de ventas de una tienda online. El dataset tiene problemas realistas que deberán resolver: valores nulos, duplicados y tipos incorrectos.

**Pregunta principal de negocio:**
> ¿Existe una diferencia significativa en el monto de compra entre clientes nuevos y recurrentes?

**Preguntas secundarias para explorar:**
- ¿Cuál es la categoría de producto más vendida?
- ¿Hay patrones temporales en las ventas?
- ¿Qué canal de adquisición genera más ingresos?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Generar dataset simulado de ventas
np.random.seed(42)
n = 500

datos = pd.DataFrame({
    'cliente_id': np.random.randint(1000, 9999, n),
    'tipo_cliente': np.random.choice(['nuevo', 'recurrente'], n, p=[0.4, 0.6]),
    'categoria': np.random.choice(['electrónica', 'ropa', 'hogar', 'deportes'], n),
    'canal': np.random.choice(['orgánico', 'paid', 'referido', 'email'], n),
    'monto': np.round(np.random.exponential(80, n) + 20, 2),
    'fecha': pd.date_range('2024-01-01', periods=n, freq='4h')
})

# Introducir problemas realistas
datos.loc[np.random.choice(n, 15), 'monto'] = np.nan
datos.loc[np.random.choice(n, 10), 'categoria'] = None
datos = pd.concat([datos, datos.sample(8)])  # duplicados

print(f'Shape: {datos.shape}')
print(f'Nulos por columna:')
print(datos.isnull().sum())
print(f'\nDuplicados: {datos.duplicated().sum()}')
datos.head(10)
```

## <span style="color: #2826DE;">Retroalimentación y Errores Comunes

**Errores que observamos frecuentemente:**
- No verificar nulos antes de hacer tests estadísticos
- Usar pie charts en lugar de barras para comparar categorías
- No formular H₀ y H₁ antes de ejecutar el test
