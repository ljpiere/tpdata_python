# <span style="color: #2826DE;">Introducción (10 mins)


En proyectos reales de programación, repetir código es una mala práctica porque:

- Hace el código más difícil de leer

- Aumenta la probabilidad de errores

- Hace más difícil modificar el programa

Pero ante esta situación las funciones nos ayudan a resolver este problema.



# <span style="color: #2826DE;">Concepto de funciones (20 mins)

### <span style="color: #2772F2;">Definición

Una función es un bloque de código reutilizable que realiza una tarea específica.

Permite:

- Organizar mejor el código

- Reutilizar lógica

- Hacer programas más claros y mantenibles


### <span style="color: #2772F2;">Ejemplo de la vida real

Una cafetera funciona como una función:

- Entrada (input):

    - Agua

    - Café

    - Botón presionado

- Proceso:

    - Calienta agua

    - Filtra café

- Salida (output):

    - Café listo 

Discusión:

- ¿Has tenido la necesidad de duplicar código en python?

- ¿En qué casos podrías sacar provecho de las funciones?

**Una funcion es como crear una herramienta específica para realizar una tarea monótona**

# <span style="color: #2826DE;">Sintaxis básica de una función (20 mins)

En Python usamos la palabra clave `def`

````python
def nombre_de_funcion():
    # código que ejecuta la función
````

Ejemplo:

````python
def saludar():
    print("Hola estudiantes del bootcamp!")
````

¿Cómo se ejecuta entonces una función?

````python
saludar()
````

### <span style="color: #2772F2;">¿Por qué usar funciones?

Sin funciones:

````python
print("Hola Juan espero que te encuentres bien hoy en tu cumpleaños número 23")
print("Hola Maria espero que te encuentres bien hoy en tu cumpleaños número 45")
print("Hola Pedro espero que te encuentres bien hoy en tu cumpleaños número 18")
````

Con funciones:

````python
def saludar(nombre, n_cumples):
    print("Hola", nombre, "espero que te encuentres bien hoy en tu cumpleaños número", n_cumples)
````

````python
saludar("Juan",23)
saludar("Maria",45)
saludar("Pedro",18)
````

El código es:

- Más limpio

- Reutilizable

- Fácil de mantener (corregir o cambiar)

# <span style="color: #2826DE;">Parámetros y valores de retorno (20 mins)

Los parámetros permiten que la función reciba información.

````python
def saludar(nombre):
    print("Hola", nombre)
````

Uso:

````python
saludar("Ana")
saludar("Carlos")
````

### <span style="color: #2772F2;">Funciones con múltiples parámetros

````python
def sumaRara(a, b):
    print(a + a + b + b)
````

````python
sumaRara(5, 3)
````

### <span style="color: #2772F2;">Funciones con múltiples parámetros

Una función también puede devolver un resultado.

````python
def sumaMasRara(a, b):
    return a + b + a*b + b + a
````

````python
resultado = sumaMasRara(4, 6)
````

````python
print(resultado)
````

### Esto es muy importante en Data Science, porque muchas funciones devuelven datos que luego usamos en análisis.
