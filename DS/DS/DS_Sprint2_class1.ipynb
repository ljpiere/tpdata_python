# <span style="color: #2826DE;">¿Qué es el Pseudocódigo? (15 mins)

### Definición

El pseudocódigo es una forma de describir un algoritmo utilizando lenguaje sencillo parecido al lenguaje humano.

### Características
- No es código ejecutable.
- Es un plan para resolver un problema antes de programar.

```

```

### Ejercicio 1.1: Calcular la mediana de una lista de números

**Pseudocódigo**

1. Ordenar la lista de números.
2. Calcular la longitud de la lista y guardarla en la variable `n`.
3. Si `n` es impar:
   - la mediana es el elemento en la posición `(n // 2)`
4. Si `n` es par:
   - la mediana es el promedio de los elementos en las posiciones `(n//2 - 1)` y `(n//2)`
5. Mostrar o devolver la mediana.


### Ejercicio 1.2: Verificar si un número es primo

**Pseudocódigo**

1. Si el número n es menor o igual que 1, entonces NO es primo.

2. Para cada divisor d desde 2 hasta la raíz cuadrada de n:

3. Si n es divisible por d, entonces NO es primo.

4. Si no se encuentra ningún divisor, entonces el número es primo.

5. Mostrar el resultado.

### Ejercicio 1.3: Encontrar el número mayor en una lista

**Pseudocódigo**

Leer una lista de números.

Inicializar la variable mayor con el primer elemento de la lista.

Para cada número x en la lista:

Si x es mayor que mayor, entonces actualizar mayor = x.

Mostrar el valor de mayor.

### Ejercicio 2.1: Calcular el total de una compra con descuento

**Problema**
Dada una lista de precios de productos, calcular el total de todos los productos y aplicar un descuento D% si la suma supera cierto valor (por ejemplo, 500).

**Pseudocódigo**

Leer la lista de precios.

Inicializar la variable total = 0.

Para cada precio p en la lista:

total = total + p

Si total > 500:

aplicar descuento: total = total - (total * D / 100)

Mostrar el total final.
