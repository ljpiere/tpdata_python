## <span style="color: #2826DE;">Variables: tipos y convenciones (20 mins)

### <span style="color: #2772F2;">¿Qué es una variable?

Una variable es un **contenedor con nombre** que almacena un valor en memoria.

Piénsala como una **caja etiquetada**: la etiqueta es el nombre, el contenido es el valor.

En Python, crear una variable es tan simple como escribir `nombre = valor`. No necesitas declarar el tipo — Python lo detecta automáticamente.

Los **4 tipos básicos** que usarás todo el tiempo en Data Science son:
- `str` → texto (nombres, categorías, etiquetas)
- `int` → números enteros (edad, conteos)
- `float` → números decimales (precios, medidas)
- `bool` → verdadero o falso (flags, condiciones)

Ejecuta el siguiente código y observa cómo Python identifica el tipo de cada variable:

```python
# Tipos básicos de variables
nombre = "Ana"          # str (texto) — siempre entre comillas
edad = 28               # int (entero) — sin comillas, sin punto decimal
altura = 1.65           # float (decimal) — con punto decimal
es_estudiante = True    # bool — solo True o False (con mayúscula)

# type() nos dice qué tipo de dato es cada variable
print(type(nombre))        # <class 'str'>
print(type(edad))          # <class 'int'>
print(type(altura))        # <class 'float'>
print(type(es_estudiante)) # <class 'bool'>
```

### <span style="color: #2772F2;">Formateo de strings (f-strings)

Los **f-strings** son la forma moderna y recomendada de combinar texto con variables:

Los **f-strings** (disponibles desde Python 3.6) son la forma más limpia de combinar texto con variables. Nota la `f` antes de las comillas — eso le dice a Python que evalúe las expresiones entre `{}`. Son mucho más legibles que concatenar strings con `+`.

### <span style="color: #2772F2;">F-strings: la forma moderna de formatear texto

Los **f-strings** (format strings) son la forma m?s limpia y legible de combinar texto con variables en Python.

Simplemente pones una `f` antes de las comillas y usas `{}` para insertar variables o expresiones.

Esto es mucho mejor que concatenar con `+` porque:
- Es más legible
- Puedes hacer operaciones dentro de `{}`
- Puedes formatear números directamente

Prueba el siguiente ejemplo:

```python
nombre = "Carlos"
edad = 32
ciudad = "México"

# f-string (recomendado)
mensaje = f"Hola, me llamo {nombre}, tengo {edad} años y vivo en {ciudad}."
print(mensaje)

# También puedes hacer operaciones dentro
print(f"En 10 años tendré {edad + 10} años.")

# Formatear números
precio = 1234.5678
print(f"El precio es: ${precio:.2f}")
```

### <span style="color: #2772F2;">Indexación de strings

Python usa **indexación basada en 0**:

### <span style="color: #2772F2;">Indexación: acceder a partes de un string

En Python, **todo empieza en 0**. El primer car?cter de un string está en la posición 0, no en la 1.

También puedes usar índices negativos para contar desde el final: `-1` es el último carácter.

El **slicing** (`texto[inicio:fin]`) te permite extraer una parte del string. Nota que el índice final **no se incluye**.

```python
texto = "Python"

print(texto[0])    # P (primer carácter)
print(texto[-1])   # n (último carácter)
print(texto[0:3])  # Pyt (slicing)
print(texto[::2])  # Pto (cada 2 caracteres)
```

### <span style="color: #2772F2;">Listas de listas

Una lista de listas es como una **tabla de datos** — la base de un DataFrame:

Una lista de listas es esencialmente una **tabla de datos**. Cada lista interna es una fila, y cada posición dentro de esa lista es una columna. ¿Te suena familiar? Así es exactamente como funciona un DataFrame de pandas internamente — y por eso es tan fácil convertir una lista de listas en un DataFrame.

Una **lista de listas** es exactamente lo que parece: una lista donde cada elemento es otra lista.

Esto es muy poderoso porque nos permite representar **tablas de datos** antes de usar pandas.

Pi?nsalo as?: cada lista interna es una **fila** de tu tabla, y cada posici?n dentro de esa lista es una **columna**.

Cuando cargues datos con `pd.DataFrame(datos)`, pandas convierte exactamente esta estructura en un DataFrame.

```python
# Lista de listas = tabla de datos
datos = [
    ["Ana", 28, "México", 7.0],
    ["Luis", 35, "España", 6.0],
    ["María", 22, "Colombia", 8.0],
    ["Carlos", 41, "Argentina", 5.0]
]

# Acceder a una fila
print(datos[0])          # Primera fila

# Acceder a un elemento específico
print(datos[0][0])       # Nombre de la primera persona
print(datos[2][2])       # País de la tercera persona

# Recorrer la tabla
for fila in datos:
    print(f"{fila[0]} tiene {fila[1]} años")
```

### <span style="color: #2772F2;">try/except: compatibilidad entre entornos

Cuando trabajas en la plataforma del bootcamp **y** en tu máquina local, las rutas de archivos pueden ser diferentes.

`try/except` te permite manejar esto elegantemente:

Cuando trabajas en la plataforma del bootcamp **y** en tu máquina local, las rutas de archivos son diferentes. En lugar de tener dos versiones del código, `try/except` te permite manejar ambos casos elegantemente. El bloque `try` intenta ejecutar el código; si falla, `except` captura el error y ejecuta el código alternativo.

Uno de los problemas m?s comunes cuando trabajas en el bootcamp es que el c?digo funciona en la plataforma pero falla en tu m?quina local (o viceversa) porque las rutas de archivos son diferentes.

`try/except` es la soluci?n elegante: **intenta** hacer algo, y si falla, **maneja el error** de forma controlada.

En el siguiente ejemplo, intentamos cargar un archivo local. Si no existe, cargamos desde internet:

```python
import pandas as pd

# Intentar cargar desde ruta local, si falla usar URL
try:
    df = pd.read_csv("datos/penguins.csv")
    print("Cargado desde archivo local")
except FileNotFoundError:
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
    df = pd.read_csv(url)
    print("Cargado desde URL")

print(df.shape)
```

### <span style="color: #2772F2;">Tipos de errores comunes en Python

## <span style="color: #2826DE;">Actividad práctica – Breakout Rooms (25 mins)

### <span style="color: #aa0808;">En esta ocasión, quien compartirá resultados será: (Giremos la ruleta)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random

rules = [
    "El más joven",
    "El que vive más lejos",
    "El más alto",
    "El más deportista",
    "Más experiencia programando",
    "El que durmió más tarde",
    "Más café hoy",
    "Más cursos terminados",
    "Username más largo",
    "Más mascotas"
]

n = len(rules)
angles = np.linspace(0, 2*np.pi, n, endpoint=False)
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect("equal")
ax.axis("off")
colors = plt.cm.tab20(np.linspace(0,1,n))

for i, angle in enumerate(angles):
    wedge = plt.matplotlib.patches.Wedge(
        (0,0), 1,
        np.degrees(angle),
        np.degrees(angle + 2*np.pi/n),
        facecolor=colors[i]
    )
    ax.add_patch(wedge)
    x = 0.7*np.cos(angle + np.pi/n)
    y = 0.7*np.sin(angle + np.pi/n)
    ax.text(x, y, rules[i], ha="center", va="center", fontsize=9,
            rotation=np.degrees(angle + np.pi/n))

pointer, = ax.plot([0,0],[0,1.1], lw=4, color="black")
chosen_index = random.randint(0, n-1)
target_angle = 2*np.pi*(5) + angles[chosen_index]
angle = 0

def update(frame):
    global angle
    if angle < target_angle:
        angle += 0.3
    pointer.set_data([0, np.cos(angle)], [0, np.sin(angle)])
    return pointer,

ani = animation.FuncAnimation(fig, update, frames=200, interval=30, blit=True)
plt.show()
print("Elegido:", rules[chosen_index])
```
