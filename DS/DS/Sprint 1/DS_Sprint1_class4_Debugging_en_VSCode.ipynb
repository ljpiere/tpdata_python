## <span style="color: #2826DE;">¿Qué es debugging? (20 mins)

### <span style="color: #2772F2;">El origen del término

En 1947, Grace Hopper encontró un **insecto real** (una polilla) atascado en un relé de la computadora Harvard Mark II.

Desde entonces, "debugging" significa encontrar y eliminar errores en el código.

### <span style="color: #2772F2;">¿Cuándo aplicar debugging?

- El código **no produce el resultado esperado**
- El código **lanza un error** que no entiendes
- El código **funciona a veces** pero no siempre
- Quieres **entender cómo funciona** un código ajeno

Antes de ver el debugger de VSCode, practiquemos la técnica más simple y más usada: **agregar prints temporales**. Esta técnica funciona en cualquier entorno y es perfecta para entender qué está pasando dentro de una función paso a paso.

### <span style="color: #2772F2;">Estrategias de debugging

**1. Lee el error completo**
El mensaje de error te dice exactamente qué falló y en qué línea.

**2. Usa `print()` para inspeccionar**
Agrega prints temporales para ver el valor de las variables en cada paso.

**3. Reproduce el bug**
Crea el caso más simple posible que reproduzca el error.

**4. Busca en internet**
Copia el mensaje de error en Google + "python" + librería.

**5. Usa el debugger**
Pausa la ejecución en puntos específicos y examina el estado.

**6. Pregunta a un LLM**
Comparte el código y el error con ChatGPT o Claude.

### <span style="color: #2772F2;">Debugging con print() — el método más simple

```python
# Código con un bug
def calcular_promedio(numeros):
    total = 0
    for n in numeros:
        total = total + n
    promedio = total / len(numeros)
    return promedio

datos = [10, 20, 30, 40]
resultado = calcular_promedio(datos)
print(f"Promedio: {resultado}")
```

```python
# Debugging con print()
def calcular_promedio_debug(numeros):
    print(f"DEBUG: numeros = {numeros}")
    total = 0
    for i, n in enumerate(numeros):
        total = total + n
        print(f"DEBUG: paso {i}, n={n}, total={total}")
    promedio = total / len(numeros)
    print(f"DEBUG: promedio final = {promedio}")
    return promedio

calcular_promedio_debug([10, 20, 30, 40])
```

## <span style="color: #2826DE;">Debugger de VSCode: breakpoints y watch (25 mins)

### <span style="color: #2772F2;">¿Qué es un breakpoint?

Un **breakpoint** es un punto de pausa en tu código. Cuando el debugger llega a esa línea, **detiene la ejecución** y te permite inspeccionar el estado del programa.

### <span style="color: #2772F2;">Cómo usar el debugger en VSCode

**Paso 1:** Haz clic en el margen izquierdo de una línea → aparece un punto rojo (breakpoint)

**Paso 2:** Presiona `F5` o el botón ▶ en la sección de Debug

**Paso 3:** El código se pausa en el breakpoint

**Paso 4:** Usa los controles:
- **F10** → Step Over (ejecutar línea actual, no entrar en funciones)
- **F11** → Step Into (entrar en la función)
- **F5** → Continue (continuar hasta el siguiente breakpoint)

**Paso 5:** En el panel **Variables** ves el valor de todas las variables en ese momento

**Paso 6:** En el panel **Watch** puedes escribir expresiones para evaluar

El debugger de VSCode es como tener un "modo pausa" para tu código. Puedes detener la ejecución en cualquier línea y examinar el valor de todas las variables en ese momento exacto. Esto es especialmente útil cuando el bug ocurre dentro de un bucle o una función anidada.

### <span style="color: #2772F2;">Tipos de breakpoints

| Tipo | Cómo activarlo | Cuándo usarlo |
|---|---|---|
| **Simple** | Clic en el margen | Siempre pausar aquí |
| **Condicional** | Clic derecho → Add Conditional Breakpoint | Pausar solo si `x > 10` |
| **Hit Count** | Clic derecho → Add Hit Count Breakpoint | Pausar en la 5ta iteración |
| **Log Message** | Clic derecho → Add Logpoint | Imprimir sin pausar |

## <span style="color: #2826DE;">IntelliSense, linting y autocompletado (20 mins)

### <span style="color: #2772F2;">¿Qué significan los símbolos de IntelliSense?

Cuando escribes código en VSCode, aparecen sugerencias con iconos de colores:

| Símbolo | Color | Significa |
|---|---|---|
| 🟣 Círculo | Morado | **Método** (función de un objeto) |
| 🔵 Cuadrado | Azul | **Propiedad** (atributo de un objeto) |
| 🟡 Triángulo | Amarillo | **Variable** local |
| 🟢 Cuadrado | Verde | **Función** |
| 🔷 Diamante | Azul | **Clase** |
| 📦 Caja | Azul | **Módulo** / librería |
| 🔑 Llave | Gris | **Keyword** (palabra reservada) |

**Atajos útiles:**
- `Ctrl+Space` → Activar sugerencias manualmente
- `F12` → Ir a la definición
- `Shift+F12` → Ver todas las referencias
- `F2` → Renombrar símbolo en todo el proyecto

### <span style="color: #2772F2;">Linting: errores antes de ejecutar

El **linter** analiza tu código mientras escribes y marca problemas:

- **Línea roja ondulada** → Error de sintaxis

## <span style="color: #2826DE;">Actividad práctica – Breakout Rooms (15 mins)

### <span style="color: #aa0808;">En esta ocasión, quien compartirá resultados será: (Giremos la ruleta)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random

rules = [
    "El más joven", "El que vive más lejos", "El más alto",
    "El más deportista", "Más experiencia programando",
    "El que durmió más tarde", "Más café hoy",
    "Más cursos terminados", "Username más largo", "Más mascotas"
]

n = len(rules)
angles = np.linspace(0, 2*np.pi, n, endpoint=False)
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect("equal")
ax.axis("off")
colors = plt.cm.tab20(np.linspace(0,1,n))

for i, angle in enumerate(angles):
    wedge = plt.matplotlib.patches.Wedge(
        (0,0), 1, np.degrees(angle), np.degrees(angle + 2*np.pi/n), facecolor=colors[i])
    ax.add_patch(wedge)
    x = 0.7*np.cos(angle + np.pi/n)
    y = 0.7*np.sin(angle + np.pi/n)
    ax.text(x, y, rules[i], ha="center", va="center", fontsize=9,
            rotation=np.degrees(angle + np.pi/n))

pointer, = ax.plot([0,0],[0,1.1], lw=4, color="black")
chosen_index = random.randint(0, n-1)
target_angle = 2*np.pi*(5) + angles[chosen_index]
angle = 0

def update(frame):
    global angle
    if angle < target_angle:
        angle += 0.3
    pointer.set_data([0, np.cos(angle)], [0, np.sin(angle)])
    return pointer,

ani = animation.FuncAnimation(fig, update, frames=200, interval=30, blit=True)
plt.show()
print("Elegido:", rules[chosen_index])
```
