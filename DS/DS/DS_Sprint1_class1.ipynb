## <span style="color: #2826DE;">Visualizaciones en python (15 mins)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Distribución de Edad**

```python
plt.figure()
sns.histplot(df["Edad"], kde=True)
plt.title("Distribución de edades del grupo")
plt.show()
```

Discusión:

- ¿Clase joven? ¿Diversa?

**Distribución de Horas de Sueño**

```python
plt.figure()
sns.histplot(df["Horas_sueno"], kde=True)
plt.title("¿Dormimos bien como grupo?")
plt.show()
```

Discusión:

¿La mayoría duerme poco o suficiente?

**Sueño vs Bebida**

```python
plt.figure()
sns.boxplot(x="Bebida", y="Horas_sueno", data=df)
plt.title("Horas de sueño según bebida favorita")
plt.show()
```

Discusión:

¿El café parece relacionarse con menos sueño?

**Participación por País**

```python

conteo = df['Pais'].value_counts()

colors = sns.color_palette('pastel')

fig, ax = plt.subplots()

wedges, texts, autotexts = ax.pie(
    conteo.values,
    labels=conteo.index,
    colors=colors,
    startangle=90,
    autopct='%.0f%%',
    wedgeprops=dict(width=0.5)
)

ax.legend(wedges, conteo.index, title="Distribución geográfica del grupo", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

**Relación Edad vs Sueño**

```python
plt.figure()
sns.scatterplot(x="Edad", y="Horas_sueno", data=df)
plt.title("Edad vs Horas de sueño")
plt.show()
```

## <span style="color: #2826DE;">Mini IA – Clustering (15 mins)

<img src="../Images/S1_DS_projectVShardware.jpeg" width="800">

### <span style="color: #2772F2;">Vamos a usar IA para ver si existen tipos de estudiantes según edad y sueño.

```python
from sklearn.cluster import KMeans

X = df[["Edad", "Horas_sueno"]]

kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)

df.head()
```

```python
plt.figure()
sns.scatterplot(x="Edad", y="Horas_sueno", hue="Cluster", data=df)
plt.title("IA detectando patrones en la clase")
plt.show()
```

Discusión guiada:

- ¿Qué grupos aparecen?

- ¿La IA predijo algo o solo agrupó?
