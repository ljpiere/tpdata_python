## <span style="color: #2826DE;">Branching strategies: GitFlow (20 mins)

### <span style="color: #2772F2;">¿Qué es GitFlow?

GitFlow es una estrategia de ramas para proyectos con releases:

```
main ──────────────────────────────────────────────► (producción)
  │                                        ▲
  │ git flow init                          │ release
  ▼                                        │
develop ──────────────────────────────────────────► (desarrollo)
  │              ▲         │              ▲
  │ feature      │ merge   │ hotfix       │
  ▼              │         ▼              │
feature/X ───────          hotfix/Y ──────
```

### <span style="color: #2772F2;">Comandos GitFlow

```bash
# Inicializar
git flow init

# Crear feature
git flow feature start analisis-outliers
# ... trabajar ...
git flow feature finish analisis-outliers

# Crear release
git flow release start 1.0.0
git flow release finish 1.0.0

# Hotfix urgente
git flow hotfix start fix-bug-critico
git flow hotfix finish fix-bug-critico
```

## <span style="color: #2826DE;">Conflictos: merge vs rebase (20 mins)

### <span style="color: #2772F2;">¿Cuándo ocurren conflictos?

Cuando dos personas modifican la **misma línea** del mismo archivo en ramas diferentes.

### <span style="color: #2772F2;">Cómo se ven los conflictos

```python
<<<<<<< HEAD
# Tu versión
def calcular_promedio(datos):
    return sum(datos) / len(datos)
=======
# Versión de la otra rama
def calcular_promedio(datos):
    return np.mean(datos)
>>>>>>> feature/usar-numpy
```

**Resolución:** Edita el archivo, elimina los marcadores, elige qué código conservar.

### <span style="color: #2772F2;">Merge vs Rebase

| | Merge | Rebase |
|---|---|---|
| **Historial** | Preserva todo (no lineal) | Reescribe (lineal) |
| **Cuándo usar** | Ramas públicas/compartidas | Ramas locales/privadas |
| **Conflictos** | Una vez | Uno por commit |

```bash
# Merge (preserva historial)
git checkout main
git merge feature/analisis

# Rebase (historial lineal)
git checkout feature/analisis
git rebase main

# Abortar si hay problemas
git merge --abort
git rebase --abort
```

## <span style="color: #2826DE;">Stash y amend (15 mins)

### <span style="color: #2772F2;">Git stash: guardar trabajo en progreso

```bash
# Guardar cambios sin commitear
git stash
git stash push -m "WIP: análisis de outliers"

# Ver stashes guardados
git stash list

# Recuperar el último stash
git stash pop

# Recuperar un stash específico
git stash apply stash@{2}

# Eliminar un stash
git stash drop stash@{0}

# Eliminar todos
git stash clear
```

### <span style="color: #2772F2;">Git commit --amend: corregir el último commit

```bash
# Corregir el mensaje del último commit
git commit --amend -m "Mensaje correcto"

# Agregar archivos olvidados al último commit
git add archivo_olvidado.py
git commit --amend --no-edit
```

⚠️ **Solo usa amend en commits que NO hayas subido a GitHub**

## <span style="color: #2826DE;">Forks y Pull Requests (10 mins)

### <span style="color: #2772F2;">¿Qué es un fork?

Un **fork** es una copia de un repositorio en tu cuenta de GitHub. Úsalo para:
- Contribuir a proyectos open source
- Experimentar sin afectar el original

```bash
# Flujo de contribución con fork
# 1. Fork en GitHub (botón Fork)
# 2. Clonar tu fork
git clone https://github.com/TU-USUARIO/proyecto.git

# 3. Agregar el original como remoto
git remote add upstream https://github.com/ORIGINAL/proyecto.git

# 4. Crear rama, hacer cambios, push
git checkout -b fix/mi-mejora
git push origin fix/mi-mejora

# 5. Crear PR desde tu fork al original

# 6. Mantener tu fork actualizado
git fetch upstream
git merge upstream/main
```

## <span style="color: #2826DE;">Ejercicio 6 – Proyecto completo con GitFlow

### <span style="color: #aa0808;">En esta ocasión, quien compartirá resultados será: (Giremos la ruleta)

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random

rules = ['El más joven','El que vive más lejos','El más alto','El más deportista','Más experiencia programando','El que durmió más tarde','Más café hoy','Más cursos terminados','Username más largo','Más mascotas']
n = len(rules)
angles = np.linspace(0, 2*np.pi, n, endpoint=False)
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect('equal')
ax.axis('off')
colors = plt.cm.tab20(np.linspace(0,1,n))
for i, angle in enumerate(angles):
    wedge = plt.matplotlib.patches.Wedge((0,0), 1, np.degrees(angle), np.degrees(angle + 2*np.pi/n), facecolor=colors[i])
    ax.add_patch(wedge)
    x = 0.7*np.cos(angle + np.pi/n)
    y = 0.7*np.sin(angle + np.pi/n)
    ax.text(x, y, rules[i], ha='center', va='center', fontsize=9, rotation=np.degrees(angle + np.pi/n))
pointer, = ax.plot([0,0],[0,1.1], lw=4, color='black')
chosen_index = random.randint(0, n-1)
target_angle = 2*np.pi*(5) + angles[chosen_index]
angle = 0
def update(frame):
    global angle
    if angle < target_angle:
        angle += 0.3
    pointer.set_data([0, np.cos(angle)], [0, np.sin(angle)])
    return pointer,
ani = animation.FuncAnimation(fig, update, frames=200, interval=30, blit=True)
plt.show()
print('Elegido:', rules[chosen_index])
````
