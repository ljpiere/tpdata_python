## <span style="color: #2826DE;">Configuración y aliases (15 mins)

### <span style="color: #2772F2;">Configuración global

```bash
# Identidad
git config --global user.name "Tu Nombre"
git config --global user.email "tu@email.com"

# Editor por defecto
git config --global core.editor "code --wait"  # VSCode

# Ver configuración
cat ~/.gitconfig
```

### <span style="color: #2772F2;">Aliases útiles

```bash
# Historial visual con ramas
git config --global alias.lg "log --graph --pretty=format:'%Cred%h%Creset -%C(yellow)%d%Creset %s %Cgreen(%cr) %C(bold blue)<%an>%Creset' --abbrev-commit --date=relative"

# Shortcuts
git config --global alias.co checkout
git config --global alias.br branch
git config --global alias.st status
git config --global alias.cm 'commit -m'

# Uso
git lg          # Historial visual
git co main     # Cambiar a main
git st          # Ver estado
```

## <span style="color: #2826DE;">Branching strategies: GitFlow (20 mins)

### <span style="color: #2772F2;">¿Qué es GitFlow?

GitFlow es una estrategia de ramas para proyectos con releases:

```
main ──────────────────────────────────────────────► (producción)
  │                                        ▲
  │ git flow init                          │ release
  ▼                                        │
develop ──────────────────────────────────────────► (desarrollo)
  │              ▲         │              ▲
  │ feature      │ merge   │ hotfix       │
  ▼              │         ▼              │
feature/X ───────          hotfix/Y ──────
```

### <span style="color: #2772F2;">Comandos GitFlow

```bash
# Inicializar
git flow init

# Crear feature
git flow feature start analisis-outliers
# ... trabajar ...
git flow feature finish analisis-outliers

# Crear release
git flow release start 1.0.0
git flow release finish 1.0.0

# Hotfix urgente
git flow hotfix start fix-bug-critico
git flow hotfix finish fix-bug-critico
```

## <span style="color: #2826DE;">Stash y amend (15 mins)

### <span style="color: #2772F2;">Git stash: guardar trabajo en progreso

```bash
# Guardar cambios sin commitear
git stash
git stash push -m "WIP: análisis de outliers"

# Ver stashes guardados
git stash list

# Recuperar el último stash
git stash pop

# Recuperar un stash específico
git stash apply stash@{2}

# Eliminar un stash
git stash drop stash@{0}

# Eliminar todos
git stash clear
```

### <span style="color: #2772F2;">Git commit --amend: corregir el último commit

```bash
# Corregir el mensaje del último commit
git commit --amend -m "Mensaje correcto"

# Agregar archivos olvidados al último commit
git add archivo_olvidado.py
git commit --amend --no-edit
```

⚠️ **Solo usa amend en commits que NO hayas subido a GitHub**

## <span style="color: #2826DE;">Forks y Pull Requests (10 mins)

### <span style="color: #2772F2;">¿Qué es un fork?

Un **fork** es una copia de un repositorio en tu cuenta de GitHub. Úsalo para:
- Contribuir a proyectos open source
- Experimentar sin afectar el original

```bash
# Flujo de contribución con fork
# 1. Fork en GitHub (botón Fork)
# 2. Clonar tu fork
git clone https://github.com/TU-USUARIO/proyecto.git

# 3. Agregar el original como remoto
git remote add upstream https://github.com/ORIGINAL/proyecto.git

# 4. Crear rama, hacer cambios, push
git checkout -b fix/mi-mejora
git push origin fix/mi-mejora

# 5. Crear PR desde tu fork al original

# 6. Mantener tu fork actualizado
git fetch upstream
git merge upstream/main
```

## <span style="color: #2826DE;">Tips y buenas prácticas

- **Usa aliases** — ahorran tiempo y errores de tipeo
- **Stash antes de cambiar de rama** — no pierdas trabajo
- **Amend solo en commits locales** — nunca en commits ya subidos
- **Merge para ramas públicas, rebase para locales**
- **Escribe buenos mensajes de commit** — tu yo del futuro te lo agradecerá

## <span style="color: #2826DE;">Cierre y Kahoot (5 mins)

Kahoot - Preguntas sugeridas

1️⃣ ¿Que hace git stash?

- Elimina los cambios no guardados
- Hace un commit automatico
- Sube los cambios a GitHub
- **Guarda cambios sin commitear temporalmente** ✅

2️⃣ ¿Cuando usar rebase en lugar de merge?

- Siempre, rebase es mejor
- Cuando trabajas en ramas publicas compartidas
- Cuando hay muchos conflictos
- **En ramas locales para mantener un historial lineal** ✅

3️⃣ ¿Que hace git commit --amend?

- Crea un nuevo commit vacio
- Elimina el ultimo commit
- Sube el commit a GitHub
- ***Modifica el ultimo commit (mensaje o archivos)** ✅

4️⃣ ¿Que es un fork en GitHub?

- Una rama del repositorio
- Un tipo de commit especial
- Un alias de Git
- **Una copia del repositorio en tu cuenta de GitHub** ✅

5️⃣ ¿Para que sirve GitFlow?

- Para hacer commits mas rapidos
- Para sincronizar con GitHub
- Para resolver conflictos automaticamente
- **Para organizar el desarrollo con ramas estructuradas** ✅
 
6️⃣ ¿Que hace git stash pop?

- Elimina todos los stashes
- Crea un nuevo stash
- Muestra la lista de stashes
- **Recupera el ultimo stash y lo elimina de la lista** ✅

7️⃣ ¿Cuando ocurre un conflicto de merge?

- Cuando hay demasiados commits
- Cuando el repositorio es muy grande
- Cuando no tienes permisos de escritura
- **Cuando dos ramas modifican la misma linea del mismo archivo** ✅

8️⃣ ¿Que hace git remote add upstream [URL]?

- Crea una nueva rama remota
- Elimina el remoto origin
- Sube los cambios al remoto
- **Agrega el repositorio original como remoto para mantener el fork actualizad** ✅
