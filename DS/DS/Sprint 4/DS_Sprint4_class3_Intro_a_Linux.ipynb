### <span style="color: #2772F2;">Estructura de carpetas en Linux

```
/                    # Raíz del sistema
├── bin/             # Programas esenciales (ls, cp, mv...)
├── usr/bin/         # Programas de usuario
├── etc/             # Configuraciones del sistema
├── home/            # Carpetas de usuarios
│   └── usuario/     # Tu carpeta personal (~)
│       ├── .bashrc  # Configuración del shell
│       ├── .ssh/    # Claves SSH
│       └── .gitconfig
├── var/             # Logs y datos variables
├── tmp/             # Archivos temporales
└── opt/             # Software opcional
```

**Windows:** Usa **WSL** (Windows Subsystem for Linux) o **Git Bash**

## <span style="color: #2826DE;">Comandos básicos de terminal (30 mins)

### <span style="color: #2772F2;">Navegación

```bash
pwd                  # Mostrar directorio actual
ls                   # Listar archivos
ls -la               # Listar con detalles y ocultos
cd carpeta           # Entrar a carpeta
cd ..                # Subir un nivel
cd ~                 # Ir a home
```

### <span style="color: #2772F2;">Archivos y carpetas

```bash
mkdir mi_proyecto    # Crear carpeta
touch archivo.py     # Crear archivo vacío
cp origen destino    # Copiar
mv origen destino    # Mover/renombrar
rm archivo.py        # Eliminar archivo
rm -rf carpeta/      # Eliminar carpeta (¡cuidado!)
cat archivo.txt      # Ver contenido
head -n 5 archivo    # Ver primeras 5 líneas
```

### <span style="color: #2772F2;">Python desde terminal

```bash
python3 script.py    # Ejecutar script
pip install pandas   # Instalar librería
pip list             # Ver librerías instaladas
python3 -m venv .venv  # Crear ambiente virtual
source .venv/bin/activate  # Activar (Linux/Mac)
.venv\Scripts\activate     # Activar (Windows)
```

### <span style="color: #2772F2;">Permisos

```bash
chmod +x script.py   # Hacer ejecutable
chown usuario archivo # Cambiar propietario
sudo comando         # Ejecutar como administrador
```

## <span style="color: #2826DE;">Variables de entorno y PATH (20 mins)

### <span style="color: #2772F2;">¿Qué son las variables de entorno?

Son variables del sistema que los programas pueden leer. Úsalas para:
- Almacenar **API keys** y contraseñas
- Configurar el **comportamiento** de programas
- Definir **rutas** de instalación

```bash
# Ver todas las variables de entorno
env

# Ver una variable específica
echo $HOME
echo $PATH

# Crear una variable temporal (solo esta sesión)
export MI_API_KEY="abc123"

# Agregar al PATH
export PATH="$PATH:/nueva/ruta"
```

### <span style="color: #2772F2;">Archivos .env para secretos

Nunca hardcodees contraseñas en el código. Usa archivos `.env`:

```bash
# Archivo .env (NUNCA subir a Git)
API_KEY=mi_clave_secreta
DB_PASSWORD=mi_password
DB_HOST=localhost
```

```python
# En Python, cargar con python-dotenv
from dotenv import load_dotenv
import os

load_dotenv()  # Carga el archivo .env
api_key = os.getenv('API_KEY')
print(f"API Key cargada: {api_key[:4]}...")
```

⚠️ **Siempre agrega `.env` a tu `.gitignore`**

## <span style="color: #2826DE;">Alias, funciones y .bashrc (15 mins)

### <span style="color: #2772F2;">Crear alias para productividad

Edita `~/.bashrc` (Linux/Mac) o `~/.bash_profile` (Mac):

```bash
# Alias útiles para DS
alias ll='ls -la'
alias gs='git status'
alias ga='git add .'
alias gc='git commit -m'
alias gp='git push'
alias activate='source .venv/bin/activate'
alias jn='jupyter notebook'

# Función para crear proyecto DS
new_ds_project() {
    mkdir -p $1/{data,notebooks,src,tests}
    cd $1
    git init
    python3 -m venv .venv
    echo '.venv/' > .gitignore
    echo '*.pyc' >> .gitignore
    echo '.env' >> .gitignore
    echo "# $1" > README.md
    echo "Proyecto $1 creado!"
}
```

Después de editar: `source ~/.bashrc`

### <span style="color: #2772F2;">Crontab: programar scripts

```bash
# Editar crontab
crontab -e

# Formato: minuto hora dia mes dia_semana comando
# Ejecutar script cada día a las 8am
0 8 * * * python3 /home/usuario/scripts/reporte.py

# Ejecutar cada hora
0 * * * * python3 /home/usuario/scripts/actualizar_datos.py

# Ver crontabs activos
crontab -l
```
