## <span style="color: #2826DE;">Alias, funciones y .bashrc (15 mins)

### <span style="color: #2772F2;">Crear alias para productividad

Edita `~/.bashrc` (Linux/Mac) o `~/.bash_profile` (Mac):

```bash

    mkdir -p $1/{data,notebooks,src,tests}

    echo '*.pyc' >> .gitignore
```

## <span style="color: #2826DE;">Actividad prĂˇctica â€“ Breakout Rooms (15 mins)

### <span style="color: #aa0808;">En esta ocasiĂłn, quien compartirĂˇ resultados serĂˇ: (Giremos la ruleta)

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random

rules = ['El mĂˇs joven','El que vive mĂˇs lejos','El mĂˇs alto','El mĂˇs deportista','MĂˇs experiencia programando','El que durmiĂł mĂˇs tarde','MĂˇs cafĂ© hoy','MĂˇs cursos terminados','Username mĂˇs largo','MĂˇs mascotas']
n = len(rules)
angles = np.linspace(0, 2*np.pi, n, endpoint=False)
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect('equal')
ax.axis('off')
colors = plt.cm.tab20(np.linspace(0,1,n))
for i, angle in enumerate(angles):
    wedge = plt.matplotlib.patches.Wedge((0,0), 1, np.degrees(angle), np.degrees(angle + 2*np.pi/n), facecolor=colors[i])
    ax.add_patch(wedge)
    x = 0.7*np.cos(angle + np.pi/n)
    y = 0.7*np.sin(angle + np.pi/n)
    ax.text(x, y, rules[i], ha='center', va='center', fontsize=9, rotation=np.degrees(angle + np.pi/n))
pointer, = ax.plot([0,0],[0,1.1], lw=4, color='black')
chosen_index = random.randint(0, n-1)
target_angle = 2*np.pi*(5) + angles[chosen_index]
angle = 0
def update(frame):
    global angle
    if angle < target_angle:
        angle += 0.3
    pointer.set_data([0, np.cos(angle)], [0, np.sin(angle)])
    return pointer,
ani = animation.FuncAnimation(fig, update, frames=200, interval=30, blit=True)
plt.show()
print('Elegido:', rules[chosen_index])
````
