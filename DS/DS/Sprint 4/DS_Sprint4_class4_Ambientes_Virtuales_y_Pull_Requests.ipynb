## <span style="color: #2826DE;">Ambientes virtuales (25 mins)

### <span style="color: #2772F2;">¿Por qué son necesarios?

Sin ambientes virtuales:
- Proyecto A necesita `pandas==1.3`
- Proyecto B necesita `pandas==2.0`
- **Conflicto** — no pueden coexistir globalmente

Con ambientes virtuales:
- Cada proyecto tiene sus **propias versiones** de librerías
- **Aislamiento total** entre proyectos

### <span style="color: #2772F2;">Crear y gestionar ambientes virtuales

```bash
# Crear ambiente virtual
python3 -m venv .venv

# Activar
source .venv/bin/activate          # Linux/Mac
.venv\Scripts\activate             # Windows

# Verificar que está activo
which python                        # Debe mostrar ruta del .venv

# Instalar librerías
pip install pandas numpy matplotlib seaborn scikit-learn

# Desactivar
deactivate
```

### <span style="color: #2772F2;">Seleccionar el venv en VSCode

1. `Ctrl+Shift+P` → "Python: Select Interpreter"
2. Selecciona el intérprete de `.venv`
3. VSCode usará ese ambiente para ejecutar código

## <span style="color: #2826DE;">pip freeze y requirements.txt (15 mins)

### <span style="color: #2772F2;">Generar requirements.txt

```bash
# Guardar todas las librerías instaladas
pip freeze > requirements.txt

# Ver el contenido
cat requirements.txt
```

Ejemplo de `requirements.txt`:
```
numpy==1.24.3
pandas==2.0.1
matplotlib==3.7.1
seaborn==0.12.2
scikit-learn==1.2.2
```

### <span style="color: #2772F2;">Reproducir un proyecto

```bash
# Clonar el repositorio
git clone https://github.com/usuario/proyecto.git
cd proyecto

# Crear ambiente virtual
python3 -m venv .venv
source .venv/bin/activate

# Instalar exactamente las mismas versiones
pip install -r requirements.txt

# ¡Listo! El proyecto funciona igual que en la máquina original
```

> **Esto es reproducibilidad** — cualquier persona puede ejecutar tu código y obtener los mismos resultados.

## <span style="color: #2826DE;">Pull Requests y trabajo colaborativo (20 mins)

### <span style="color: #2772F2;">¿Qué es un Pull Request?

Un **Pull Request (PR)** es una solicitud para fusionar cambios de una rama a otra.

Flujo típico en equipos:

```
main ──────────────────────────────────────────────────►
         │                              ▲
         │ git checkout -b feature      │ merge (PR aprobado)
         ▼                              │
feature ─────────────────────────────────
         (trabajas aquí)    (creas PR)
```

### <span style="color: #2772F2;">Crear un Pull Request

```bash
# 1. Crear rama para tu feature
git checkout -b feature/analisis-outliers

# 2. Hacer cambios y commits
git add .
git commit -m "Agregar análisis de outliers con IQR"

# 3. Subir la rama a GitHub
git push -u origin feature/analisis-outliers

# 4. En GitHub: New Pull Request
# - Base: main
# - Compare: feature/analisis-outliers
# - Agregar descripción clara
# - Asignar reviewers
```

### <span style="color: #2772F2;">Revisar un Pull Request

1. Lee el código cambiado
2. Deja comentarios en líneas específicas
3. Aprueba o solicita cambios
4. Merge cuando esté listo

## <span style="color: #2826DE;">Actividad práctica – Breakout Rooms (20 mins)

### <span style="color: #aa0808;">En esta ocasión, quien compartirá resultados será: (Giremos la ruleta)

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random

rules = ['El más joven','El que vive más lejos','El más alto','El más deportista','Más experiencia programando','El que durmió más tarde','Más café hoy','Más cursos terminados','Username más largo','Más mascotas']
n = len(rules)
angles = np.linspace(0, 2*np.pi, n, endpoint=False)
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect('equal')
ax.axis('off')
colors = plt.cm.tab20(np.linspace(0,1,n))
for i, angle in enumerate(angles):
    wedge = plt.matplotlib.patches.Wedge((0,0), 1, np.degrees(angle), np.degrees(angle + 2*np.pi/n), facecolor=colors[i])
    ax.add_patch(wedge)
    x = 0.7*np.cos(angle + np.pi/n)
    y = 0.7*np.sin(angle + np.pi/n)
    ax.text(x, y, rules[i], ha='center', va='center', fontsize=9, rotation=np.degrees(angle + np.pi/n))
pointer, = ax.plot([0,0],[0,1.1], lw=4, color='black')
chosen_index = random.randint(0, n-1)
target_angle = 2*np.pi*(5) + angles[chosen_index]
angle = 0
def update(frame):
    global angle
    if angle < target_angle:
        angle += 0.3
    pointer.set_data([0, np.cos(angle)], [0, np.sin(angle)])
    return pointer,
ani = animation.FuncAnimation(fig, update, frames=200, interval=30, blit=True)
plt.show()
print('Elegido:', rules[chosen_index])
````
