## <span style="color: #2826DE;">Comandos de trabajo remoto (30 mins)

### <span style="color: #2772F2;">El flujo completo con GitHub

```
Tu máquina                    GitHub
─────────────────────────────────────────
Working Dir → Staging → Local Repo ←→ Remote Repo
              git add    git commit   git push/pull
```

### <span style="color: #2772F2;">Comandos remotos esenciales

```bash
# Conectar con GitHub
git remote add origin https://github.com/usuario/repo.git
git remote -v                    # Ver remotos configurados

# Subir cambios
git push -u origin main          # Primera vez (establece tracking)
git push                         # Veces siguientes

# Descargar cambios
git fetch                        # Descarga sin aplicar
git pull                         # Descarga y aplica (fetch + merge)

# Clonar un repositorio existente
git clone https://github.com/usuario/repo.git
```

### Preguntas:

- ¿Cual es la diferencia entre git fetch y git pull?

- ¿Por que usamos -u la primera vez que hacemos push?

- ¿Que pasaria si dos personas hacen push al mismo tiempo?

El flujo con GitHub tiene dos direcciones: puedes subir tus cambios locales al servidor (push) o descargar cambios del servidor a tu maquina (pull/fetch). Entender este flujo es fundamental para trabajar en equipo sin perder trabajo.

El flujo con GitHub agrega dos pasos al flujo local: push para subir tus commits al servidor, y pull para descargar los cambios de otros. La diferencia entre etch y pull es importante: etch descarga los cambios pero no los aplica, pull los descarga Y los aplica automaticamente.

## <span style="color: #2826DE;">Ramas, configuración y .gitignore (20 mins)

### <span style="color: #2772F2;">Gestión de ramas

```bash
# Ver ramas
git branch                       # Ramas locales
git branch -a                    # Todas las ramas

# Crear y cambiar de rama
git branch nueva-feature         # Crear rama
git checkout nueva-feature       # Cambiar a rama
git checkout -b otra-feature     # Crear y cambiar en un paso

# Fusionar ramas
git checkout main                # Volver a main
git merge nueva-feature          # Fusionar

# Etiquetas (versiones)
git tag v1.0                     # Crear tag
git push origin v1.0             # Subir tag
```

### Preguntas:

### <span style="color: #2772F2;">El archivo .gitignore

`.gitignore` le dice a Git qué archivos **no debe rastrear**.

**Ejemplo de `.gitignore` para proyectos de Data Science:**

```
# Entornos virtuales
.venv/
venv/
env/

# Python
__pycache__/
*.pyc
*.pyo
.ipynb_checkpoints/

# Datos (pueden ser muy grandes)
data/raw/
*.csv
*.xlsx

# Secretos
.env
secrets.json
credentials.json

# Sistema operativo
.DS_Store
Thumbs.db
```

⚠️ **Nunca hagas commit de contraseñas, API keys o datos sensibles.**

## <span style="color: #2826DE;">Tips y buenas practicas

- **Pull antes de push**: git pull antes de subir para evitar conflictos

- **Usa ramas para features**: no trabajes directamente en main

- **Escribe buenos mensajes de commit**: "Agregar analisis de outliers" no "cambios"

- **Sube todos tus proyectos del bootcamp** - es tu portafolio

- **Tu GitHub es tu CV tecnico** - los reclutadores lo revisan

- **Haz commits frecuentes** - es mejor muchos commits pequenos que uno grande

- **Nunca hagas commit de secretos** - usa .gitignore y .env

## <span style="color: #2826DE;">Cierre y Kahoot (5 mins)

Kahoot - Preguntas sugeridas

1️⃣  Que hace git push?

- Descarga cambios del repositorio remoto
- Crea una nueva rama
- Hace un commit de los cambios
- **Sube los commits locales al repositorio remoto** ✅

2️⃣ Cual es la diferencia entre git fetch y git pull?

- Son exactamente lo mismo
- Fetch es mas rapido que pull
- pull descarga pero no aplica los cambios
- **Fetch descarga sin aplicar, pull descarga y aplica** ✅

3️⃣ Que hace git checkout -b nueva-rama?

- Elimina la rama especificada
- Cambia a una rama existente
- Fusiona dos ramas
- **Crea una nueva rama y cambia a ella** ✅

4️⃣ Para que sirve .gitignore?

- Para hacer commits mas rapidos
- Para crear ramas automaticamente
- Para sincronizar con GitHub
- **Para excluir archivos del repositorio** ✅

5️⃣ Que hace git clone?

- Crea un repositorio vacio
- Hace un backup del repositorio
- Elimina el repositorio remoto
- **Descarga una copia completa de un repositorio remoto** ✅

6️⃣ Por que es importante el .gitignore para Data Science?

- Para que el repositorio sea mas pequeno
- Para ocultar el codigo fuente
- Para evitar conflictos de merge
- **Para no subir datos sensibles, credenciales o archivos grandes** ✅

7️⃣ Que hace git merge feature/analisis?

- Elimina la rama feature/analisis
- Crea una copia de la rama
- Sube la rama a GitHub
- **Fusiona los cambios de feature/analisis en la rama actual** ✅

8️⃣ Que es un tag en Git?

- Un tipo de rama especial
- Un mensaje de commit
- Un archivo de configuracion
- **Una etiqueta que marca un commit especifico como version** ✅
