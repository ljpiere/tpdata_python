### 7.1.5 Escribiendo tus primeras mini-funciones de limpieza de datos

Una mini-función de limpieza debería:

1. Recibir una entrada clara.
2. Aplicar reglas específicas.
3. Devolver un resultado consistente.
4. Ser fácil de reutilizar.

### Diseño de función
Antes de escribir código, define:

- **Nombre de la función:** `_____________________________________________`
- **Qué columna limpiará:** `_____________________________________________`
- **Qué regla aplicará:** `_____________________________________________`
- **Qué devolverá:** `_____________________________________________`


```python

# TODO:
# Completa una mini-función para limpiar edades.
# Puedes usar este esqueleto y ajustarlo.

def clean_age(df: pd.DataFrame, col: str = "age", min_age: int = 16, max_age: int = 80) -> pd.DataFrame:
    df = df.copy()
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df["age_clean"] = df[col].where(df[col].between(min_age, max_age), np.nan)
    return df

df_age = clean_age(df)
df_age[["age", "age_clean"]].head(10)
```

```python

# TODO:
# Completa una mini-función para normalizar country y city.
# Idea: quitar espacios, pasar a formato título y convertir vacíos a NaN.

def clean_country_city(df: pd.DataFrame, country_col: str = "country", city_col: str = "city") -> pd.DataFrame:
    df = df.copy()

    for col in [country_col, city_col]:
        df[col] = df[col].astype("string").str.strip()
        df[col] = df[col].replace({"": pd.NA})
        df[col] = df[col].str.title()

    return df

df_clean = clean_country_city(df_age)
df_clean[[ "country", "city" ]].head(10)
```
