## 10.1.1 · Introducción a las herramientas de BI

**BI (Business Intelligence)** son herramientas que te ayudan a:
- Conectar datos (CSV, Excel, bases de datos, APIs, etc.)
- Prepararlos (limpiar, transformar, combinar)
- Modelarlos (relaciones entre tablas)
- Visualizarlos (gráficos, KPIs, dashboards)
- Compartir resultados (publicar/exportar)

### ¿Cuándo usar Power BI o Tableau?
- Cuando quieres **comunicar hallazgos** a un público no técnico.
- Cuando necesitas un **dashboard recurrente**.
- Cuando quieres que otras personas puedan **explorar** la información con filtros y segmentación.

### Conceptos clave (para hoy)
- **KPI:** indicador clave de desempeño.
- **Dashboard:** conjunto de visualizaciones que responde preguntas de negocio.
- **Grano:** qué representa exactamente una fila.
- **Hecho:** tabla con eventos o transacciones.
- **Dimensión:** tabla con atributos para segmentar o describir.

```python
# ============================================================
# Generador de dataset sintético (para Power BI / Tableau Public)
# - No necesitas entender todo el código.
# - Solo ejecútalo para obtener los CSV.
# ============================================================

import os
import numpy as np
import pandas as pd

rng = np.random.default_rng(42)

out_dir = "dataset_bi_sprint10"
os.makedirs(out_dir, exist_ok=True)

# ---------- 1) Dimensiones (tablas pequeñas) ----------
n_customers = 2000
n_products = 60
n_channels = 12

regions = ["Andina","Caribe","Pacífica","Centroamérica","Cono Sur"]
segments = ["Básico","Plus","Premium"]
age_groups = ["18-24","25-34","35-44","45-54","55+"]

# Clientes
customer_id = np.arange(100000, 100000 + n_customers)
dim_customers = pd.DataFrame({
    "customer_id": customer_id,
    "segment": rng.choice(segments, size=n_customers, p=[0.55, 0.30, 0.15]),
    "age_group": rng.choice(age_groups, size=n_customers, p=[0.18, 0.32, 0.24, 0.16, 0.10]),
    "region": rng.choice(regions, size=n_customers, p=[0.28, 0.18, 0.18, 0.14, 0.22]),
})

# Productos
categories = ["Tecnología","Hogar","Moda","Deportes","Libros"]
product_id = np.arange(2000, 2000 + n_products)
dim_products = pd.DataFrame({
    "product_id": product_id,
    "category": rng.choice(categories, size=n_products),
    "product_name": [f"Producto {i}" for i in range(1, n_products+1)],
    "base_unit_price": np.round(rng.normal(70, 35, size=n_products).clip(5, 300), 2)
})

# Canales
channels = ["Orgánico","Búsqueda paga","Redes sociales","Email","Referidos","Afiliados"]
devices = ["Mobile","Desktop","Tablet"]
campaigns = ["Always-on","Lanzamiento","Seasonal","Retargeting","Influencers"]

channel_id = np.arange(10, 10 + n_channels)
dim_channels = pd.DataFrame({
    "channel_id": channel_id,
    "channel": rng.choice(channels, size=n_channels),
    "campaign": rng.choice(campaigns, size=n_channels),
    "device": rng.choice(devices, size=n_channels, p=[0.58, 0.36, 0.06]),
})

# ---------- 2) Hechos (tabla grande) ----------
n_rows = 20000  # ~20k filas (ideal para clase)
date_range = pd.date_range("2024-01-01", "2026-02-28", freq="D")

fact = pd.DataFrame({
    "order_item_id": np.arange(1, n_rows+1),
    "order_id": rng.integers(500000, 509000, size=n_rows),
    "order_date": rng.choice(date_range, size=n_rows),
    "customer_id": rng.choice(customer_id, size=n_rows),
    "product_id": rng.choice(product_id, size=n_rows),
    "channel_id": rng.choice(channel_id, size=n_rows),
    "quantity": rng.integers(1, 4, size=n_rows)
})

# Precios y descuentos (simples)
fact = fact.merge(dim_products[["product_id","base_unit_price","category"]], on="product_id", how="left")
fact["discount_pct"] = rng.choice([0, 0.05, 0.10, 0.15], size=n_rows, p=[0.35, 0.30, 0.25, 0.10])
fact["unit_price"] = np.round(fact["base_unit_price"] * (1 - fact["discount_pct"]) * (1 + rng.normal(0, 0.03, size=n_rows)), 2).clip(1, None)
fact["revenue"] = np.round(fact["unit_price"] * fact["quantity"], 2)

# Costos y utilidad
cost_ratio = rng.normal(0.60, 0.10, size=n_rows).clip(0.35, 0.90)
fact["cost"] = np.round(fact["revenue"] * cost_ratio, 2)
fact["profit"] = np.round(fact["revenue"] - fact["cost"], 2)

# Logística / calidad
fact["delivery_days"] = rng.integers(1, 15, size=n_rows)
fact["returned"] = (rng.random(n_rows) < 0.04).astype(int)  # ~4% devuelto
fact["rating"] = np.round((rng.normal(4.3, 0.6, size=n_rows) - 0.08*fact["delivery_days"] - 0.4*fact["returned"]).clip(1, 5), 1)

# Guardar fact sin columnas de apoyo
fact_sales = fact.drop(columns=["base_unit_price"])
fact_sales.to_csv(os.path.join(out_dir, "bi_fact_sales.csv"), index=False, encoding="utf-8")

# Guardar dimensiones
dim_customers.to_csv(os.path.join(out_dir, "bi_dim_customers.csv"), index=False, encoding="utf-8")
dim_products.drop(columns=["base_unit_price"]).to_csv(os.path.join(out_dir, "bi_dim_products.csv"), index=False, encoding="utf-8")
dim_channels.to_csv(os.path.join(out_dir, "bi_dim_channels.csv"), index=False, encoding="utf-8")

# Tabla plana (join) para Tableau / inicio rápido
sales_flat = (fact_sales
              .merge(dim_customers, on="customer_id", how="left")
              .merge(dim_products.drop(columns=["base_unit_price"]), on="product_id", how="left")
              .merge(dim_channels, on="channel_id", how="left"))

sales_flat["profit_margin"] = np.round(np.where(sales_flat["revenue"]>0, sales_flat["profit"]/sales_flat["revenue"], np.nan), 4)
sales_flat.to_csv(os.path.join(out_dir, "bi_sales_flat.csv"), index=False, encoding="utf-8")

print("✅ Archivos generados en:", os.path.abspath(out_dir))
print("   - bi_sales_flat.csv (recomendado para empezar)")
print("   - bi_fact_sales.csv + bi_dim_* (opcional para modelo estrella)")
```
