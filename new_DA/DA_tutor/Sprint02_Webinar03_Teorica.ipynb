## 1) Preguntas de negocio y mĂŠtricas (10 min)

Una pregunta de negocio no siempre estĂĄ lista para anĂĄlisis. Muchas veces empieza como una inquietud ambigua:

> âÂżCĂłmo vamos en ventas?â

### Estructura recomendada

```text

```

### Ejemplos

| Pregunta vaga | Pregunta analĂ­tica |

| Tipo | QuĂŠ indica | Ejemplos |
|---|---|---|
| *Leading* | SeĂąal temprana que puede anticipar resultados | visitas, leads, cotizaciones, unidades vendidas |
