### Fórmulas guía

Para quitar espacios:

```excel
=TRIM(A2)
```

Para convertir a número:

```excel
=VALUE(TRIM(A2))
```

Para crear semana:

```excel
=YEAR(C2)&"-"&TEXT(WEEKNUM(C2,2),"00")
```

Para normalizar feriado, define una regla sencilla:

| Valor original | Valor limpio |
|---|---|
| `TRUE`, `YES`, `1` | `1` |
| `FALSE`, `NO`, `0` | `0` |

## Parte 3 — KPIs principales (20 min)

Crea tres hojas:

```text

kpi_share
```

### KPI 1 — Eficiencia: ventas por m²

**Definición:**

```text
Sales_per_m2 = SUM(weekly_sales) / size_m2
```

**Pivot sugerida:**

| Área | Campo |
|---|---|
| Rows | `store_name` |
| Columns | `store_type` |
| Values | `SUM weekly_sales` |
| Helper | dividir por `size_m2` |

**Interpretación:** una tienda pequeña puede ser más eficiente que una tienda grande si genera más ventas por metro cuadrado.

### KPI 2 — Crecimiento semanal

**Definición:**

```text
Weekly_growth_pct = (sales_current_week - sales_previous_week) / sales_previous_week
```

**Pasos sugeridos:**

1. Crea una Pivot por `week_key` con suma de ventas.
2. Ordena semanas ascendentemente.
3. Agrega una columna de crecimiento.
4. Evita división por cero o semanas vacías.

### KPI 3 — Participación por departamento

**Definición:**

```text
Dept_share_pct = department_sales / total_sales
```

**Pivot sugerida:**

| Área | Campo |
|---|---|
| Rows | `dept_name` |
| Values | `SUM weekly_sales` |
| Show as | `% of grand total` |

Si la herramienta no permite porcentaje del total, calcula fuera de la Pivot:

```excel
=B2/SUM($B$2:$B$10)
```
