# Ejercicio 2 · Power BI: conexión, Power Query y KPIs (15 min)

## Parte A · Conectar archivo CSV

1. Abre **Power BI Desktop**.
2. Selecciona **Get Data / Obtener datos**.
3. Elige **Text/CSV**.
4. Carga `sprint10_proyecto_bi_ventas_multicanal.csv`.
5. Antes de cargar definitivamente, entra a **Transform Data / Transformar datos**.

## Parte B · Revisar tipos en Power Query

Valida estos tipos:

| Columna | Tipo esperado |
|---|---|
| `Fecha_pedido` | Date / Fecha |
| `Ingresos`, `Costo`, `Ganancia` | Decimal number |
| `Unidades`, `Devolucion`, `Dias_entrega` | Whole number |
| `Descuento_pct`, `Satisfaccion` | Decimal number |
| Dimensiones | Text |

### Fecha con locale
Si Power BI interpreta mal la fecha:

1. Selecciona `Fecha_pedido`.
2. Tipo de datos → **Using Locale / Usar configuración regional**.
3. Tipo: **Date / Fecha**.
4. Locale: **Spanish (Colombia)**, **Spanish (Peru)** o equivalente.

## Parte C · Crear columna condicional `Nivel_Venta`

En Power Query:

- Si `Ingresos` >= 300 → `Venta Alta`
- Si no → `Venta Baja`

Luego selecciona **Close & Apply / Cerrar y aplicar**.

## Parte D · Crear medidas DAX básicas

En Power BI: **Modeling → New measure**.

```DAX
Ingresos Total = SUM(Ventas[Ingresos])
```

```DAX
Ganancia Total = SUM(Ventas[Ganancia])
```

```DAX
Margen = DIVIDE([Ganancia Total], [Ingresos Total])
```

```DAX
Pedidos = DISTINCTCOUNT(Ventas[ID_Pedido])
```

```DAX
Clientes = DISTINCTCOUNT(Ventas[ID_Cliente])
```

```DAX
% Devoluciones = DIVIDE(SUM(Ventas[Devolucion]), [Pedidos])
```

> Si tu tabla no se llama `Ventas`, cambia `Ventas` por el nombre real de la tabla cargada.

### Narrativa ejecutiva SCQA

Usa esta estructura:

- **S — Situación:** qué muestra el desempeño general.
- **C — Complicación:** qué hallazgo requiere atención.
- **Q — Pregunta:** qué debe decidir el negocio.
- **A — Acción:** recomendación basada en evidencia.

### Ejemplo
**S:** Las ventas 2024–2025 muestran mayor concentración en Colombia y México.  
**C:** Algunas categorías generan ingresos altos, pero con menor margen o más devoluciones.  
**Q:** ¿Dónde debe enfocar la dirección comercial sus esfuerzos?  
**A:** Priorizar categorías con margen alto y revisar canales con mayor tasa de devolución.
