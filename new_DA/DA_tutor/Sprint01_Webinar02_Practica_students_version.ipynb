### Takeaways

1. Explorar no es lo mismo que limpiar.
2. Un buen diagnóstico evita arreglos improvisados.
3. Las copias protegen los datos originales.
4. Buscar y reemplazar debe aplicarse por rango específico.
5. Las funciones simples son suficientes para muchas limpiezas iniciales.
6. El análisis empieza a ser confiable cuando la limpieza está documentada.

> En estos notebooks usaremos ambos nombres cuando sea útil:  
> `TRIM/ESPACIOS`, `LOWER/MINUSC`, `UPPER/MAYUSC`, `PROPER/NOMPROPIO`, `COUNT/CONTAR`, `COUNTA/CONTARA`, `COUNTBLANK/CONTAR.BLANCO`, `AVERAGE/PROMEDIO`, `COUNTIF/CONTAR.SI`, `UNIQUE/UNICOS`, `SUMIF/SUMAR.SI`.
