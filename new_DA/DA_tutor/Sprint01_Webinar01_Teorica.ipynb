## 2) ¿Qué hace un/a Data Analyst? (10 min)

Un/a Data Analyst convierte preguntas de negocio en evidencia útil para tomar decisiones.

### 2.1 Entregables comunes

| Entregable | Para qué sirve |
|---|---|
| Dataset revisado | Garantizar que los datos son utilizables |
| Tabla resumen | Sintetizar métricas principales |
| Dashboard | Comunicar indicadores de forma visual |
| Informe | Explicar hallazgos, limitaciones y recomendaciones |
| Documentación | Dejar trazabilidad para que otro pueda revisar o reproducir |

### 2.2 Flujo mental del Data Analyst

```text
Pregunta de negocio
    ↓
Archivo o fuente de datos
    ↓
Exploración inicial
    ↓
Detección de problemas
    ↓
Limpieza mínima y documentada
    ↓
Cálculo de métricas
    ↓
Comunicación de hallazgos
```

### 2.3 Criterio profesional

Un/a Data Analyst no solo calcula. También decide si un dato es confiable, si una transformación está justificada y si el resultado puede ser explicado.

## 4) Dataset del Sprint 1: caso Taylor Swift (10 min)

El archivo `sp1_w1_taylor_swift_da_intro.xlsx` simula una situación común: recibimos datos útiles, pero no totalmente limpios.

### 4.1 Hojas principales

| Hoja | Unidad de análisis | Uso en clase |
|---|---|---|
| `raw_songs` | Una fila ≈ una canción | Inspección, problemas de formato y limpieza |

### 4.2 Problemas intencionales del dataset

En `raw_songs` aparecen problemas como:

En esta sesión los reconocemos y definimos cómo tratarlos. En la práctica los trabajaremos paso a paso.


## 5) Exploración y diagnóstico antes de limpiar (12 min)

Antes de corregir datos, se debe entender el archivo.

### 5.1 Preguntas de exploración inicial

Al abrir una hoja, responde:

- ¿Qué representa una fila?
- ¿Cuántas filas y columnas hay?
- ¿Qué columnas parecen numéricas?
- ¿Qué columnas parecen fechas?
- ¿Qué columnas son categóricas?
- ¿Qué columnas podrían tener nulos?
- ¿Qué columnas podrían tener duplicados?
- ¿Qué problemas parecen afectar cálculos?

### 5.2 Problemas que se documentan en `issues_log`

| Tipo de problema | Ejemplo | Riesgo |
|---|---|---|
| Espacios extra | `" LOVE STORY "` | Dificulta búsquedas exactas |
| Categoría inconsistente | `FOLKLORE`, `folklore`, ` Folklore ` | Agrupaciones incorrectas |
| Número como texto | `1.493,44` | Promedios o sumas incorrectas |
| Booleano no estándar | `Y`, `1`, `maybe` | Conteos incorrectos |
| Fecha ambigua | `12/09/08` | Error en análisis temporal |
| Duplicado | Canción repetida | Inflar métricas |

### 5.3 Estructura sugerida de `issues_log`

| Columna | Descripción |
|---|---|
| `id` | Número del hallazgo |
| `sheet` | Hoja donde aparece |
| `column` | Columna afectada |
| `issue_type` | Tipo de problema |
| `example_value` | Ejemplo observado |
| `business_risk` | Impacto del problema |
| `priority` | Alta, media o baja |
| `suggested_action` | Qué se recomienda hacer después |


## 6) Qué es limpieza mínima de datos (13 min)

La limpieza mínima busca corregir problemas que impiden una lectura o cálculo básico, sin sobreprocesar el archivo.

### 6.1 Sí es limpieza mínima

| Acción | Ejemplo |
|---|---|
| Quitar espacios extra | `" LOVE STORY "` → `"LOVE STORY"` |
| Estandarizar categorías | `FOLKLORE`, `folklore`, ` Folklore ` → `Folklore` |
| Corregir encabezados | `Spotify Streams (M)` → `spotify_streams_m` |
| Estandarizar booleanos | `Y`, `1`, `TRUE` → `TRUE` |
| Validar números como números | Revisar si Sheets puede promediar una columna |
| Documentar decisiones | Registrar cambio, hoja, antes y después |

### 6.2 No es limpieza mínima

| Acción | Por qué evitarla al inicio |
|---|---|
| Cambiar valores sin evidencia | Puede alterar el significado del dato |
| Eliminar filas duplicadas automáticamente | Puede borrar registros válidos |
| Interpretar fechas ambiguas sin regla | Puede crear errores temporales |
| Aplicar reemplazos a todo el archivo | Puede modificar columnas no relacionadas |
| Usar fórmulas complejas sin explicación | Dificulta auditoría y aprendizaje |

### 6.3 Regla central

Si no puedes explicar qué cambiaste, dónde lo cambiaste y por qué, todavía no deberías aplicar ese cambio al archivo final.


## 7) Buscar y reemplazar como herramienta de limpieza (7 min)

Buscar y reemplazar es útil, pero debe aplicarse con control.

### 7.1 Casos adecuados

| Caso | Ejemplo | Acción |
|---|---|---|
| Quitar símbolo repetido | `$1,200` | Buscar `$`, reemplazar vacío |
| Quitar espacios dobles | `Taylor  Swift` | Reemplazar doble espacio por uno |
| Estandarizar categoría | `yes`, `Y`, `1` | Convertir a una convención definida |
| Normalizar separador | `1,863.12` | Evaluar si la coma es miles o decimal |

### 7.2 Riesgos

Buscar `.` y reemplazar vacío puede dañar una fecha, un decimal o un texto.  
Buscar `1` y reemplazar `TRUE` en todo el archivo puede afectar valores numéricos.

### 7.3 Reglas de seguridad

1. Trabajar en una copia.
2. Aplicar el reemplazo a un rango específico.
3. Revisar ejemplos antes y después.
4. Documentar el cambio.
5. Verificar con una métrica simple.


## 8) Funciones simples de texto y verificación (10 min)

Las funciones simples permiten crear columnas auxiliares limpias.

### 8.1 Funciones de texto

| Función | Uso | Ejemplo conceptual |
|---|---|---|
| `TRIM/ESPACIOS` | Quitar espacios al inicio, final y repetidos | `" LOVE STORY "` → `"LOVE STORY"` |
| `LOWER/MINUSC` | Convertir a minúsculas | `"FOLKLORE"` → `"folklore"` |
| `UPPER/MAYUSC` | Convertir a mayúsculas | `"folklore"` → `"FOLKLORE"` |
| `PROPER/NOMPROPIO` | Convertir a nombre propio | `"lover"` → `"Lover"` |

### 8.2 Funciones de verificación

| Pregunta | Fórmula en inglés | Fórmula en español |
|---|---|---|
| ¿Cuántas celdas numéricas hay? | `=COUNT(rango)` | `=CONTAR(rango)` |
| ¿Cuántas celdas no vacías hay? | `=COUNTA(rango)` | `=CONTARA(rango)` |
| ¿Cuántas celdas vacías hay? | `=COUNTBLANK(rango)` | `=CONTAR.BLANCO(rango)` |
| ¿Cuántas veces aparece una categoría? | `=COUNTIF(rango,"x")` | `=CONTAR.SI(rango;"x")` |
| ¿Qué valores únicos existen? | `=UNIQUE(rango)` | `=UNICOS(rango)` |
| ¿Se puede calcular un promedio? | `=AVERAGE(rango)` | `=PROMEDIO(rango)` |

Si `AVERAGE/PROMEDIO` falla o ignora valores, probablemente algunos números siguen como texto.


## 9) Trazabilidad: `issues_log`, `change_log` y criterio profesional (6 min)

### 9.1 Diferencia entre `issues_log` y `change_log`

| Hoja | Cuándo se usa | Qué responde |
|---|---|---|
| `issues_log` | Durante la exploración | ¿Qué problema encontré y qué impacto tiene? |
| `change_log` | Durante o después de la limpieza | ¿Qué cambié, dónde y cómo quedó? |

### 9.2 Ejemplo de `change_log`

| step | who | when | change_description | sheet | before_example | after_example |
|---:|---|---|---|---|---|---|
| 1 | JP | 2026-06-06 | Aplicar `TRIM/ESPACIOS` en título de canción | `songs_clean_v1` | `" LOVE STORY "` | `"LOVE STORY"` |
| 2 | JP | 2026-06-06 | Estandarizar `Y` y `1` como `TRUE` en `explicit` | `songs_clean_v1` | `Y` | `TRUE` |

### 9.3 Cuándo detenerse

| Situación | Decisión recomendada |
|---|---|
| El problema impide calcular una métrica central | Limpiar ahora |
| El problema afecta pocas filas y no impacta el análisis actual | Documentar y dejar pendiente |
| La corrección requiere conocer una regla de negocio | Consultar antes de cambiar |
| El dato parece inválido pero no hay evidencia | Marcar para revisión |
| La limpieza puede dañar otros valores | Probar primero en una copia |
