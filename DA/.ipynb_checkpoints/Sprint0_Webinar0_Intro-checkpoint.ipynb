## 2) Introducción a la futura profesión en Data Analytics (10–25)
La profesión de **Data Analytics** integra habilidades **técnicas** (SQL, Excel, Power BI, Python), **analíticas** (estadística, diseño de métricas) y **de negocio** (KPIs, storytelling).
- Oportunidades en banca, salud, retail, logística, tecnología, sector público.
- Trayectorias de carrera: Analyst → Senior → Lead → Head/Director.
- Modalidades: remoto, híbrido, consultoría, producto.

## 3) Diferentes roles en Data y en la industria TI (25–40)
**Roles en datos**
- **Data Analyst**: KPIs, dashboards, insights accionables.
- **Data Engineer**: ingesta, ETL/ELT, calidad y modelos de datos, orquestación.
- **Data Scientist**: experimentos, ML, estadística, prototipos.
- **Machine Learning Engineer**: despliegue de modelos, MLOps, monitoreo.
- **BI Analyst/Developer**: semántica de métricas, cubos, autoservicio.
- **Analytics/AI Product Manager**: priorización de valor y roadmap.

**Roles TI relacionados**
- **Software Engineer/Backend** • **DBA** • **Cloud/DevOps** • **Security** • **Solutions Architect**.

**Colaboración típica**
Negocio plantea problema → Datos se preparan (DE) → Análisis/experimentos (DA/DS) → Producto de datos o modelo en prod (MLE/SE) → Monitoreo y mejora continua.

---
## 4) El ecosistema de los datos (40–55)
- **Fuentes**: OLTP, logs, APIs, IoT, redes sociales, encuestas.
- **Ingesta/Streaming**: batch vs streaming, colas (Kafka/Kinesis).
- **Almacenamiento**: DWH (BigQuery/Redshift/Snowflake), Data Lake (S3/GCS), Lakehouse.
- **Procesamiento**: ETL/ELT (dbt/Spark), orquestación (Airflow).
- **Analítica/Visualización**: SQL, Excel, Power BI, Tableau, Looker.
- **Gobernanza**: catálogo, linaje, calidad, privacidad, seguridad, FinOps.
- **MLOps**: experimentación, registro de modelos, despliegue, monitoreo.

> El analista **no** domina todo, pero comprende el flujo end-to-end y traduce datos a decisiones.

## 6) Hábitos y foco: Pomodoro + estudio online (80–95)
**Pomodoro:** 25 min foco + 5 min descanso; cada 4 bloques, pausa 15–20 min.

**Checklist**
- Horario fijo y entorno sin distracciones (modo “No molestar”).
- Notas activas (Cornell/Outline) + “hoja de una página” por sesión.
- Practicar inmediatamente tras la teoría (“aprender haciendo”).
- Técnica Feynman: explica en 60–120 s lo aprendido.
- Plan semanal: 6–8 pomodoros planeados y ejecutados + evidencia.
