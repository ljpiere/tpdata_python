# Errores frecuentes y cómo leerlos

| Error frecuente | Qué significa | Cómo corregir |
|---|---|---|
| `IO Error: No files found that match the pattern` | DuckDB no encontró el Parquet en la ruta indicada. | Abre la carpeta completa `bookstore_sql_vscode` y revisa que exista `data/`. |
| `Catalog Error: Table ... does not exist` | La vista no fue creada en esta conexión/base. | Ejecuta `sql/01_setup_views.sql`. |
| `Binder Error: column must appear in the GROUP BY` | Seleccionaste una columna no agregada que no está en el `GROUP BY`. | Agrégala al `GROUP BY` o usa una función agregada. |
| `Referenced table alias not found` | Usaste un alias que no existe. | Revisa el alias declarado en `FROM` o `JOIN`. |
| `Parser Error` | Problema de sintaxis. | Revisa comas, paréntesis, comillas y punto y coma. |
| Resultado duplicado o inflado | El `JOIN` multiplicó filas. | Revisa llaves de unión y usa `COUNT(DISTINCT ...)` cuando corresponda. |

La depuración también es parte del aprendizaje de SQL.
