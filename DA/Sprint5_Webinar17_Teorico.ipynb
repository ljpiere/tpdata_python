## 2. Dataset de licores de Sudamérica (demo)
Descripción del dataset...

```

import pandas as pd
import numpy as np

data = {
    "brand": [
        "Aguardiente Antioqueño", "Pisco Quebranta", "Cachaça do Rio",
        "Ron Caribeño", "Vino Andino", "Singani del Valle",
        "Aguardiente Cristal", "Pisco Acholado", "Cachaça da Serra",
        "Ron Pacífico", "Vino Patagónico", "Licor de Café",
    ],
    "country": [
        "Colombia", "Perú", "Brasil",
        "Colombia", "Chile", "Bolivia",
        "Colombia", "Peru", "Brasil",
        "Colombia", "Argentina", "",
    ],
    "liquor_type": [
        "Aguardiente", "Pisco", "Cachaça",
        "Ron", "Vino", "Singani",
        "Aguardiente", "Pisco", "Cachaça",
        "Ron", "Vino", "Licor",
    ],
    "category": [
        "estándar", "premium", "artesanal",
        "estándar", "premium", "artesanal",
        "N/A", "premium", "artesanal",
        "estándar", None, "estándar",
    ],
    "alcohol_abv": [
        29.0, 41.0, 38.0,
        37.5, 13.5, 40.0,
        29.0, np.nan, 38.0,
        35.0, 14.0, 28.0,
    ],
    "bottle_ml": [
        750, 700, 1000,
        750, 750, 700,
        750, 700, np.nan,
        1000, 750, 500,
    ],
    "price_usd": [
        12.5, 22.0, 18.0,
        15.0, 10.0, 25.0,
        11.0, np.nan, 17.5,
        16.0, 13.0, 0.0,
    ],
    "annual_sales_liters": [
        500000, 80000, 120000,
        200000, 150000, 30000,
        450000, 60000, 110000,
        210000, np.nan, -1,
    ],
    "export_flag": [
        "Yes", "Yes", "No",
        "No", "Yes", "No",
        "Yes", "Yes", "No",
        "No", "Yes", "N/A",
    ],
}

df = pd.DataFrame(data)
df
```

## Diagnóstico de valores faltantes
Uso de isna(), sum(), filtros...

```
df.isna().sum()
```

## Limpieza avanzada: Normalizar códigos especiales

```

df_clean = df.copy()
df_clean['country'] = df_clean['country'].replace({'': np.nan})
df_clean['category'] = df_clean['category'].replace({'N/A': np.nan})
df_clean['export_flag'] = df_clean['export_flag'].replace({'N/A': np.nan})
df_clean['price_usd'] = df_clean['price_usd'].replace({0.0: np.nan})
df_clean['annual_sales_liters'] = df_clean['annual_sales_liters'].replace({-1: np.nan})
df_clean
```

## Agrupación con groupby()

```
df_clean.groupby('country')['annual_sales_liters'].sum().reset_index()
```
