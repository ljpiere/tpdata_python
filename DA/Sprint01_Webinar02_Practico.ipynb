## Introducción

En el Webinar 1 aprendimos qué hace un/a Data Analyst y cómo abrir Google Sheets y Colab.

Ahora vamos a trabajar como si recibiéramos un archivo real de negocio:

**abrir → entender → inspeccionar → detectar problemas → documentar → preparar entrega al equipo**.

La práctica está diseñada para que cada estudiante toque el archivo y salga con una auditoría inicial clara.


## Objetivos de aprendizaje

Al finalizar la sesión podrás:

1. Importar el Excel del sprint en Google Sheets.
2. Identificar la unidad de análisis de cada hoja.
3. Activar filtros, congelar encabezados y duplicar hojas de trabajo.
4. Detectar problemas de nulos, duplicados, formatos y categorías.
5. Crear un `issues_log` para documentar problemas sin corregirlos todavía.
6. Preparar un resumen de hallazgos para una siguiente fase de limpieza.

## Ejercicio 2 — Auditoría de `raw_songs` / `songs_audit` (20 min)

### Objetivo

Identificar problemas frecuentes en una tabla de canciones.

### Descripción

Revisa `songs_audit` y documenta mínimo 6 hallazgos en `issues_log`.

Busca problemas en estas categorías:

| Categoría | Qué buscar | Columnas sugeridas |
|---|---|---|
| Espacios extra | Valores con espacios antes/después | `Song`, `Album`, `Tour Era` |
| Mayúsculas/minúsculas | Categorías iguales escritas distinto | `Album`, `Tour Era` |
| Fechas mixtas | Formatos diferentes de fecha | `Release Date`, `Date Added` |
| Números como texto | Comas, puntos o espacios inconsistentes | `Spotify Streams (M)`, `YouTube Views (M)` |
| Booleanos inconsistentes | `Y`, `N`, `1`, `0`, `TRUE`, `FALSE`, `maybe` | `Explicit` |
| Duplicados | Canciones repetidas | `Song`, `Album`, `ISRC` |
| Delimitadores mixtos | Separadores diferentes en una misma columna | `Writers` |

### Fórmulas simples permitidas para explorar

Puedes usar estas fórmulas en celdas auxiliares:

| Necesidad | Inglés | Español aproximado |
|---|---|---|
| Contar celdas con números | `=COUNT(rango)` | `=CONTAR(rango)` |
| Contar celdas no vacías | `=COUNTA(rango)` | `=CONTARA(rango)` |
| Contar vacíos | `=COUNTBLANK(rango)` | `=CONTAR.BLANCO(rango)` |
| Ver valores únicos | `=UNIQUE(rango)` | `=UNICOS(rango)` |
| Contar una categoría | `=COUNTIF(rango,"criterio")` | `=CONTAR.SI(rango;"criterio")` |

> No uses todavía fórmulas complejas. El objetivo es detectar, no solucionar.

### Pistas

- Filtra por celdas vacías.
- Ordena de A a Z para ver variaciones de texto juntas.
- Usa `UNIQUE/UNICOS` para categorías como `Explicit` o `Tour Era`.
- Mira ejemplos concretos, no solo “la columna está mal”.
