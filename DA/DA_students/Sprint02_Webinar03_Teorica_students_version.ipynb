## 1) Preguntas de negocio y métricas (10 min)

Una pregunta de negocio no siempre está lista para análisis. Muchas veces empieza como una inquietud ambigua:

> “¿Cómo vamos en ventas?”

Esa pregunta debe traducirse a algo medible:

> “¿Cómo evolucionaron las ventas mensuales de Laptop y Smartphone por región durante el último trimestre?”

### Estructura recomendada

```text
[Métrica] para [grupo o segmento] durante [periodo] con [filtro o condición]
```

### Ejemplos

| Pregunta vaga | Pregunta analítica |
|---|---|
| ¿Cómo van las ventas? | ¿Cuál fue el total de ventas por región durante el primer semestre de 2025? |
| ¿Qué productos funcionan mejor? | ¿Qué productos tienen mayor ingreso total y mayor ticket promedio? |
| ¿Dónde debemos invertir? | ¿Qué regiones tienen alto volumen de ventas pero bajo cumplimiento frente a la meta mensual? |
| ¿Qué canal es más fuerte? | ¿Qué canal aporta mayor venta total por segmento de cliente? |

### Leading vs. lagging indicators

| Tipo | Qué indica | Ejemplos |
|---|---|---|
| *Leading* | Señal temprana que puede anticipar resultados | visitas, leads, cotizaciones, unidades vendidas |
| *Lagging* | Resultado posterior de la operación | ingresos, margen, rentabilidad, cumplimiento de meta |

En análisis comercial, una buena pregunta suele combinar ambos: no solo mira cuánto se vendió, sino qué comportamiento podría explicar ese resultado.

## 3) Funciones condicionales para métricas segmentadas (13 min)

Las funciones condicionales permiten responder preguntas sin construir todavía una PivotTable.

### Funciones base

| Función | Uso |
|---|---|
| `SUMIF` | Sumar con una condición |
| `COUNTIF` | Contar con una condición |
| `AVERAGEIF` | Promediar con una condición |
| `SUMIFS` | Sumar con varias condiciones |
| `COUNTIFS` | Contar con varias condiciones |
| `AVERAGEIFS` | Promediar con varias condiciones |

### Ejemplo 1: ventas por región

```excel
=SUMIF(Sales_Raw!E:E, A2, Sales_Raw!M:M)
```

Donde `A2` contiene una región.

### Ejemplo 2: ventas por región y producto

```excel
=SUMIFS(Sales_Raw!M:M, Sales_Raw!E:E, A2, Sales_Raw!H:H, B1)
```

Donde `A2` contiene la región y `B1` contiene el producto.

## 4) PivotTables: resumir sin fórmulas extensas (14 min)

Las PivotTables permiten agrupar, filtrar y resumir rápidamente grandes tablas.

### Primera PivotTable

**Pregunta guía:** ¿qué regiones y categorías explican más `total_sales`?

Configuración sugerida:

| Área de la PivotTable | Campo |
|---|---|
| Rows / Filas | `region` |
| Columns / Columnas | `category` |
| Values / Valores | `SUM of total_sales` |
| Filter / Filtro | `quarter` o `channel` |

### Filtrar y ordenar

Después de crear la tabla dinámica:

1. Ordena regiones por ventas totales descendentes.
2. Filtra un trimestre específico.
3. Cambia columnas de `category` a `product` para mayor detalle.
4. Observa si la conclusión cambia.

### Diferentes cálculos

Una misma tabla puede resumir con:

- suma de ventas;
- conteo de ventas;
- promedio de ventas;
- promedio de unidades;
- participación porcentual del total.

### Riesgo común

Un promedio puede ocultar distribución. Si un producto tiene pocas ventas muy altas, puede parecer “mejor” que otro con ventas más frecuentes y estables. Por eso la interpretación debe revisar volumen y promedio.
