## 1. ¿Qué es Python?

Python es un **lenguaje de programación**. Un lenguaje permite darle instrucciones a un computador de forma estructurada.

En analítica de datos se usa porque permite:

- automatizar tareas repetitivas;
- limpiar y transformar datos;
- analizar tablas grandes con menos trabajo manual;
- crear visualizaciones;
- conectar diferentes fuentes de datos;
- documentar el proceso en notebooks.

Una forma simple de entenderlo:

```text
Datos + instrucciones en Python = resultados reproducibles
```

La palabra clave es **reproducible**: si guardamos el código, otra persona puede ejecutar los mismos pasos y verificar el análisis.

## 2. ¿Cómo funciona Python?

Python funciona interpretando instrucciones línea por línea. En una celda de Colab escribimos código; al ejecutarla, Python procesa esas instrucciones y devuelve un resultado.

Hay tres piezas importantes:

| Pieza | Qué significa | Ejemplo |
|---|---|---|
| Lenguaje | Reglas para escribir instrucciones | `precio * cantidad` |
| Intérprete | Programa que ejecuta Python | Runtime de Colab |
| Librería | Código ya creado para una tarea específica | `pandas` para tablas |

En este sprint no buscamos memorizar todo Python. Buscamos usarlo como herramienta de análisis.

````python
# Primeros ejemplos de Python puro
saludo = "Hola, estamos aprendiendo Python"
numero_estudiantes = 35
precio = 12.5
cantidad = 4

print(saludo)
print("Total:", precio * cantidad)
print("Tipo de dato de precio:", type(precio))
````

## 3. Google Colab: cómo interactuar con el entorno

Google Colab es un entorno en la nube para trabajar con notebooks. Un notebook combina:

- celdas de texto en Markdown;
- celdas de código en Python;
- resultados visibles debajo de cada celda;
- archivos cargados temporalmente;
- documentación del razonamiento del análisis.

### Acciones básicas en Colab

- Ejecutar una celda: `Shift + Enter` o botón ▶️.
- Crear celda de código: menú `+ Código`.
- Crear celda de texto: menú `+ Texto`.
- Reiniciar ejecución: `Entorno de ejecución > Reiniciar entorno de ejecución`.
- Ejecutar todo: `Entorno de ejecución > Ejecutar todo`.
- Ver archivos: panel izquierdo > ícono de carpeta.

### Colab vs Python local

| Aspecto | Google Colab | Python local |
|---|---|---|
| Instalación | No requiere instalación inicial | Requiere instalar Python y librerías |
| Archivos | Se cargan al runtime y pueden perderse | Permanecen en el computador |
| Recursos | Usa una máquina temporal en la nube | Usa recursos del computador propio |
| Ideal para | Aprender, prototipar y compartir | Proyectos más estables o productivos |
| Riesgo común | Olvidar subir el dataset cada sesión | Problemas de instalación o versiones |

Para clase usaremos Colab porque reduce fricción técnica y permite concentrarnos en el análisis.

## 5. Tipos de datos básicos en Python

| `str` | `"Colombia"` | Texto |

## 6. pandas y DataFrames

`pandas` es una librería de Python para trabajar con datos tabulares.

Su estructura principal es el **DataFrame**. Un DataFrame se parece a una tabla de Excel o Google Sheets:

- tiene filas;
- tiene columnas;
- cada columna tiene un nombre;
- cada columna puede tener un tipo de dato;
- permite filtrar, agrupar, resumir y transformar datos con código.

La diferencia principal es que en pandas dejamos un registro reproducible de cada paso.

````python
import pandas as pd

pd.set_option("display.max_columns", 30)
pd.set_option("display.precision", 2)

print("Versión de pandas:", pd.__version__)
````
