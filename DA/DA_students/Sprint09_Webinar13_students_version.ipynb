# 1. Introducción matemática corta y directa

Antes de usar Python, fijemos la lógica estadística de la clase.

## 1.1 Muestra vs población

Normalmente no observamos a todos los usuarios posibles. Observamos una **muestra**.

- **Población:** todos los usuarios que podrían visitar la landing.
- **Muestra:** usuarios que participaron en este experimento.

La estadística ayuda a responder:

> Si veo una diferencia en la muestra, ¿esa diferencia parece real o podría ser ruido?

## 1.2 Media

La media resume una variable numérica:

\[
\bar{x} = \frac{x_1 + x_2 + \cdots + x_n}{n}
\]

En este proyecto, la usaremos para comparar el **gasto promedio** entre A y B.

## 1.3 Proporción

Una proporción mide una parte del total:

\[
p = \frac{\text{casos favorables}}{\text{casos totales}}
\]

En este proyecto, la usaremos para comparar la **tasa de conversión**.

## 1.4 Hipótesis

Una prueba estadística compara dos explicaciones:

- **H0 — hipótesis nula:** no hay diferencia real entre A y B.
- **H1 — hipótesis alternativa:** sí hay diferencia real entre A y B.

## 1.5 p-value

El **p-value** responde, de forma simplificada:

> Si H0 fuera cierta, ¿qué tan raro sería observar una diferencia como la que vimos?

Regla práctica de esta clase:

- Si `p-value < 0.05`: hay evidencia estadística para rechazar H0.
- Si `p-value >= 0.05`: no hay evidencia suficiente para rechazar H0.

Importante: un p-value pequeño no significa que el efecto sea grande. Por eso también revisamos el **tamaño del efecto**.

# Takeaways

- La estadística aplicada empieza con una buena pregunta de negocio.
- Una hipótesis se prueba sobre una métrica concreta y grupos bien definidos.
- Antes de aplicar pruebas, hay que revisar calidad de datos.
- Para proporciones como conversión usamos tablas de contingencia y Chi-cuadrado.
- Para promedios como gasto usamos t-test, preferiblemente Welch si no asumimos varianzas iguales.
- El p-value no reemplaza al criterio de negocio: también importan tamaño del efecto, costos, riesgos y limitaciones.
- Una buena recomendación combina métricas, evidencia estadística, visualizaciones y una conclusión clara.
