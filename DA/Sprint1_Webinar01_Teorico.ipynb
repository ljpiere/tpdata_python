## Introducción

En esta primera sesión vamos a aterrizar **qué hace un/a Data Analyst (DA)** y cómo se traduce ese rol a un flujo de trabajo real **usando Google Sheets** como herramienta principal.

> Idea clave: en la vida real, gran parte del trabajo inicial de análisis ocurre en herramientas “ligeras” (Sheets/Excel) para **entender el dato rápido**, validar supuestos y preparar una primera entrega.

**¿Por qué Google Sheets en este sprint?**
- Es accesible (no requiere instalar nada) y permite colaborar en tiempo real.
- Tiene lo necesario para: **limpiar, explorar, resumir y comunicar** hallazgos.

Durante la clase trabajaremos con un archivo realista (Taylor Swift) para practicar el “día 1” del DA: **abrir un archivo, entenderlo, detectar problemas y construir un resumen**.

## 1) Rol del Data Analyst (DA)

Un/a **Data Analyst** transforma datos en decisiones. No es “hacer gráficos por hacerlos”, sino responder preguntas con evidencia.

### 1.1 ¿Qué entregables produce un DA?
- **Resumen ejecutivo**: qué está pasando y por qué importa.
- **Tablas / dashboards**: indicadores y segmentaciones que se actualizan.
- **Insights accionables**: recomendaciones con respaldo.
- **Documentación**: definiciones, supuestos, limitaciones y fuentes.

### 1.2 El DA como puente (negocio ↔ datos)
En la práctica, el DA:
1. Aclara la pregunta (qué decisión se quiere tomar).
2. Identifica la fuente de datos y valida calidad.
3. Analiza (agrega, segmenta, compara, detecta outliers).
4. Comunica resultados (claro, visual, verificable).
5. Itera con stakeholders (feedback → ajuste).

### 1.3 ¿Cómo se ve este rol usando Google Sheets?
Google Sheets es un “laboratorio” perfecto para el DA porque permite:
- **Auditar datos** (nulos, duplicados, formatos).
- **Explorar rápido** con filtros, ordenamiento y tablas dinámicas.
- **Construir un resumen** sin depender de ingeniería ni código.
- **Dejar evidencia** (pestañas, notas, fórmulas simples) para que otro pueda revisar.


## 2) Un mundo lleno de herramientas (y por qué empezamos con Google Sheets)

En analítica existen muchas herramientas: BI (Power BI/Tableau/Looker), bases de datos (SQL), notebooks (Python/R), plataformas de producto, etc.  
Pero antes de “subir de nivel”, necesitas dominar los **fundamentos**.

### 2.1 ¿Qué es Google Sheets?
Google Sheets es una hoja de cálculo en la nube que te permite:
- Trabajar con **tablas** (filas/columnas) y aplicar **filtros y ordenamientos**.
- Crear **tablas dinámicas (Pivot Tables)** para resumir y segmentar.
- Usar **fórmulas** para limpieza y cálculo.
- Compartir y colaborar (comentarios, historial, control de cambios).

### 2.2 ¿Cuándo es ideal usar Sheets?
- Exploración inicial y validación rápida del dato.
- Análisis descriptivo: conteos, promedios, distribuciones simples.
- Prototipado: “primera versión” de indicadores y reportes.
- Colaboración con negocio: revisiones rápidas.

### 2.3 ¿Cuáles son sus límites?
- Manejo de datos muy grandes (millones de filas).
- Transformaciones complejas o automatización avanzada.
- Control fino de versiones (aunque tiene historial).

> Regla práctica: Sheets para **entender y comunicar**; SQL/Python para **escalar y automatizar**.


## 3) DA en diferentes tipos de organizaciones

- **Startup:** rol generalista (limpieza → insight → slide). Velocidad > formalización.
- **Mid-size:** especialización moderada; DA trabaja con BI/DE; procesos básicos de gobierno.
- **Enterprise:** clara división de funciones, compliance, estándares estrictos, *data governance*.

**Implicaciones para el DA**
- En entornos pequeños, el DA arma el flujo end-to-end.
- En entornos grandes, el foco es claridad, trazabilidad y *hand-off* impecable.

---

## 4) Tipos de Seniority

- **Junior:** limpieza y métricas básicas; documentación guiada; foco en *best practices*.
- **Mid:** autonomía en limpieza y comunicación; propone KPIs y flujos de trabajo.
- **Senior:** define estándares, lidera entregas complejas y coordinación con otros roles.
- **Lead/Manager:** visión, priorización de iniciativas, relación con negocio.

**Autoevaluación rápida**
- ¿Puedo explicar una decisión metodológica a un stakeholder no técnico?
- ¿Mis archivos se entienden sin mí (nombres, notas, `change_log`)?
- ¿Sé cuándo escalar a BI/DE/DS?

---

## 5) Work Process (de la pregunta a la entrega)

Un flujo práctico y repetible para el DA:

1. **Descubrir el problema**
   - ¿Qué decisión se tomará? ¿Qué KPI importa? ¿Para quién?
2. **Auditar fuentes**
   - ¿Qué tablas/hojas hay? ¿Qué columnas? ¿Qué representa cada campo?
3. **Limpieza y estandarización**
   - Tipos correctos, duplicados, nulos, formatos consistentes.
4. **Cálculo & EDA**
   - SUM/COUNT/AVG, segmentación por categorías, tendencias simples.
5. **Visualización & narrativa**
   - Gráfico + interpretación + recomendación.
6. **Handoff & documentación**
   - Archivo ordenado + `change_log` con pasos reproducibles.

> Un DA no solo entrega “un número”, entrega **confianza**: datos limpios + reglas + evidencia.

### 6.2 Checklist de calidad (lo que un DA busca al abrir el archivo)

Al inspeccionar `raw_songs` y `raw_albums`, un DA típicamente verifica:

- **Fechas mixtas** (p. ej. texto vs fecha real).
- **Booleanos inconsistentes** (Yes/No, 1/0, TRUE/FALSE).
- **Números con símbolos** (%, $, comas de miles, espacios).
- **Separadores inconsistentes** en listas (p. ej., `Writers` con `,` y `;` mezclados).
- **Duplicados** (mismo Song/Album con variaciones mínimas).
- **Nulos** o valores fuera de rango.

Esto conecta con el work process: antes de analizar, **estandarizar**.

### 6.3 Inspección rápida del dataset en Google Sheets (sin código)


**Objetivo:** abrir el archivo y hacer un diagnóstico inicial (5–10 min) como lo haría un DA.

#### Paso A — Importar el archivo
1. Abre Google Drive → **Nuevo → Hojas de cálculo de Google**.
2. En Sheets: **Archivo → Importar → Subir** y selecciona `sp1_w1_taylor_swift_da_intro.xlsx`.
3. Importa como **“Insertar nuevas hojas”** (para conservar las pestañas originales).

#### Paso B — Entender las pestañas (qué mirar en 60 segundos)
- **`raw_songs`**: tabla principal (cada fila ≈ una canción con atributos).
- **`raw_albums`**: tabla de referencia (metadatos del álbum).
- **`data_dictionary`**: diccionario (definición de columnas; tu “contrato” de datos).
- **`change_log`**: registro de cambios (qué se modificó y por qué).
- **`summary_template`**: estructura guía para el reporte/entrega.

#### Paso C — “Data Audit” mínimo (checklist en Sheets)
En `raw_songs` y `raw_albums`:

1. **Congelar encabezado:** Ver → Congelar → 1 fila.  
2. **Activar filtros:** Datos → Crear un filtro.  
3. **Revisar tipos/formato visual:**
   - ¿Hay números guardados como texto? (alineación rara, comillas, símbolos)
   - ¿Fechas bien parseadas o vienen como texto?
4. **Buscar nulos / vacíos:**
   - Filtra por “(Vacío)” en columnas críticas (ej.: `track_name`, `album`, `release_date`).
5. **Duplicados:**
   - Datos → **Limpieza de datos → Quitar duplicados** (primero prueba en una copia).
   - Alternativa: Formato → **Formato condicional** para resaltar duplicados.
6. **Valores fuera de rango / inconsistentes:**
   - Ordena asc/desc columnas numéricas (ej.: `popularity`, `duration_ms`) y detecta extremos.
7. **Consistencia de categorías:**
   - Crea una columna auxiliar con `=LOWER(TRIM(A2))` para normalizar textos (si aplica).
   - Usa `=UNIQUE(rango)` para listar categorías y ver “typos”.

#### Paso D — Evidencia (cómo documentar como DA)
- Crea una pestaña `audit_notes` y registra:
  - columnas con vacíos,
  - columnas con formatos inconsistentes,
  - supuestos (qué consideras “llave” / granularidad),
  - decisiones (qué limpiarás y qué NO).


**Interpretación:** si aparecen columnas “sospechosas”, probablemente son métricas que deberían ser numéricas pero llegaron como texto con formato (p. ej., `$1,234` o `45%`).  
En Excel/Sheets esto se traduce en: *no puedes sumar/promediar correctamente hasta limpiar*.

---

### 6.4 Cómo trabajarlo en Excel/Google Sheets (lo que haremos en clase)

**Paso A — Inspección y mapeo**
- Abre `data_dictionary` y define **tipos esperados**.
- Haz una lista de **riesgos** en `change_log`.

**Paso B — Reglas de limpieza (definir antes de tocar datos)**
- Fechas → `YYYY-MM-DD`.
- Booleanos → `{TRUE, FALSE}` únicamente.
- Números → sin símbolos, sin espacios; separador consistente.
- `Writers` → separador único (p. ej., `;`).
- Duplicados → criterio explícito (p. ej., `Song + Album + Release Date`).

**Paso C — Documentación (change_log)**
- Cada transformación debe quedar registrada:
  - qué cambiaste
  - por qué
  - ejemplo antes/después
  - fecha y responsable
