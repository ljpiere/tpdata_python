## 1) ¿Qué hace un/a Data Analyst? (15 min)

Un/a **Data Analyst** transforma datos en evidencia útil para tomar decisiones.  
No se trata de “hacer gráficos bonitos”, sino de responder preguntas con datos confiables.

### 1.1 Entregables comunes

| Entregable | Ejemplo | Para qué sirve |
|---|---|---|
| Tabla limpia | Archivo con columnas consistentes | Permite calcular métricas sin errores |
| Reporte exploratorio | Resumen de hallazgos iniciales | Ayuda a decidir próximos pasos |
| Dashboard | Vista con KPIs y filtros | Monitorea resultados de forma recurrente |
| Presentación ejecutiva | Recomendaciones con evidencia | Comunica decisiones a negocio |
| Documentación | `change_log`, diccionario, notas | Permite reproducir y auditar el trabajo |

### 1.2 Flujo mental del DA

Antes de calcular, un DA pregunta:

- ¿Cuál es la pregunta de negocio?
- ¿Qué representa cada fila?
- ¿Qué columnas son confiables?
- ¿Qué columnas necesitan limpieza?
- ¿Qué métricas se pueden calcular ahora y cuáles no?
- ¿Qué limitaciones debo comunicar?

## 2) Por qué empezamos con Google Sheets (15 min)

Google Sheets es útil al inicio porque permite:

- Ver los datos inmediatamente.
- Compartir el archivo con otras personas.
- Detectar errores visualmente.
- Aplicar limpiezas simples.
- Documentar cambios.
- Construir métricas rápidas sin programar.

### 2.1 Cuándo usar Sheets

Usa Google Sheets cuando:

- El dataset es pequeño o mediano.
- Necesitas colaboración rápida.
- Estás inspeccionando datos por primera vez.
- Quieres explicar hallazgos a personas no técnicas.
- El objetivo es aprender fundamentos de análisis.

### 2.2 Cuándo NO basta Sheets

Sheets empieza a quedarse corto cuando:

- Hay millones de filas.
- Necesitas automatización robusta.
- El proceso debe correr cada día.
- Hay múltiples fuentes conectadas.
- Se requiere control estricto de versiones.

En esos casos aparecen herramientas como SQL, Python, BI o pipelines de datos.

## 5) Dataset del Sprint 1: caso Taylor Swift (15 min)

El archivo `sp1_w1_taylor_swift_da_intro.xlsx` simula una situación común: recibimos un archivo con datos útiles, pero no totalmente limpios.

### 5.1 Hojas principales

| Hoja | Unidad de análisis | Uso en clase |
|---|---|---|
| `raw_songs` | Una fila ≈ una canción | Inspección, problemas de formato y limpieza |
| `raw_albums` | Una fila = un álbum | Validación y métricas de álbum |
| `data_dictionary` | Una fila = una columna documentada | Entender tipos esperados |
| `change_log` | Una fila = un cambio realizado | Trazabilidad |
| `summary_template` | Una fila = una métrica | Resumen posterior a la limpieza |
| `README` | Una fila = una instrucción | Contexto del archivo |

### 5.2 Problemas intencionales del dataset

En `raw_songs` aparecen problemas como:

- Nombres con espacios extra: `" LOVE STORY "`.
- Mayúsculas y minúsculas inconsistentes: `FOLKLORE`, `folklore`, ` Lover `.
- Fechas en formatos diferentes: `12/09/08`, `Oct 05, 2015`, `18-08-2014`.
- Números como texto: `1,863.12`, `349,76`, `1.493,44`.
- Valores booleanos inconsistentes: `Y`, `N`, `1`, `0`, `TRUE`, `FALSE`, `maybe`, vacíos.
- Duplicados: por ejemplo, una canción puede aparecer repetida.
- Delimitadores diferentes en texto: coma, slash, barra vertical o punto y coma en `Writers`.

En el Webinar 1 solo los reconocemos. En los Webinars 2, 3 y 4 los trabajaremos paso a paso.
