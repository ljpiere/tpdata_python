## Ejercicio 2 — Buscar y reemplazar en números y categorías (15 min)

### Objetivo

Corregir problemas repetitivos sin fórmulas complejas.

### Parte A — Revisar columnas numéricas

En `songs_clean_v1`, revisa:

- `spotify_streams_m`
- `youtube_views_m`

Busca casos como:

- `1,863.12`
- `349,76`
- `1.493,44`
- espacios antes o después

### Acción sugerida

1. Filtra la columna.
2. Revisa varios ejemplos.
3. Decide si la coma representa miles o decimales.
4. Aplica Buscar y reemplazar solo en el rango de la columna.
5. Verifica con `COUNT/CONTAR`.

### Parte B — Revisar `explicit`

Valores posibles en el archivo:

- `Y`
- `N`
- `1`
- `0`
- `TRUE`
- `FALSE`
- `Yes`
- `maybe`
- vacío

Convención sugerida:

| Valor original | Valor limpio |
|---|---|
| `Y`, `Yes`, `1`, `TRUE` | `TRUE` |
| `N`, `No`, `0`, `FALSE` | `FALSE` |
| `maybe`, vacío | dejar pendiente / revisar |

### Pistas

- No reemplaces `1` en todo el archivo.
- Aplica reemplazos solo sobre la columna `explicit`.
- Documenta cada grupo de reemplazos.

## Ejercicio 4 — Verificación con métricas rápidas (11 min)

### Objetivo

Demostrar que la limpieza permite calcular métricas básicas.

### Descripción

En `summary_template`, completa al menos 4 métricas:

| Métrica | Fórmula sugerida |
|---|---|
| Total Tracks | `=COUNTA(songs_clean_v1!A2:A)` / `=CONTARA(...)` |
| Unique Albums | `=COUNTUNIQUE(songs_clean_v1!B2:B)` o lista con `UNIQUE/UNICOS` |
| Average Spotify Streams (M) | `=AVERAGE(songs_clean_v1!E2:E)` / `=PROMEDIO(...)` |
| Average YouTube Views (M) | `=AVERAGE(songs_clean_v1!F2:F)` / `=PROMEDIO(...)` |
| Tracks Marked Explicit | `=COUNTIF(songs_clean_v1!G2:G,"TRUE")` / `=CONTAR.SI(...)` |

### Pistas

- Si `AVERAGE/PROMEDIO` no calcula bien, revisa si los números siguen como texto.
- Si `COUNTUNIQUE` no está disponible en tu idioma, usa `UNIQUE/UNICOS` y cuenta manualmente.
- Compara el total de filas antes y después.

### Takeaways

- Limpiar datos no es improvisar: es decidir, aplicar, verificar y documentar.
- Las copias protegen los datos originales.
- Buscar y reemplazar debe aplicarse por rango.
- Las funciones simples son suficientes para muchas limpiezas iniciales.
- El análisis empieza a ser confiable cuando la limpieza está documentada.
