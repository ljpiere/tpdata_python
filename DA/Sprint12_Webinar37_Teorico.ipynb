# Bloque 4 · Paso 2: Análisis de rentabilidad con Python (5 min) 💰

## Petición del jefe
> "Necesito entender si el negocio realmente está generando ganancias."

## ¿Qué deben hacer los estudiantes?

Usar los datasets limpios del Paso 1 para calcular **KPIs de negocio**:

### Parte 1 — Rentabilidad general
| KPI | Cómo calcularlo |
|---|---|
| **Revenue total** | Suma de `monto_total` en orders |
| **Costo total de productos** | Unir orders con catalog por producto, calcular `cantidad × costo_unitario` |
| **Inversión en marketing** | Suma de `gasto` en marketing_spend |
| **Ganancia bruta** | Revenue − Costo de productos |
| **Ganancia neta** | Revenue − Costo de productos − Marketing |
| **Margen de ganancia** | Ganancia neta / Revenue × 100 |

### Parte 2 — Segmentación
- ¿Qué **categorías** de producto son más rentables?
- ¿Qué **países** generan más revenue?
- ¿Qué **canales de marketing** son más eficientes? (Revenue por dólar invertido)

## Tips para el Paso 2

> 💡 **El merge es clave aquí.** Para calcular costos, necesitan unir `orders` con `catalog`
> usando `nombre_producto` (o la columna equivalente). Recuérdales:
> ```python
> orders_with_cost = orders.merge(catalog[['nombre_producto', 'costo_unitario']], on='nombre_producto', how='left')
> ```

> ⚠️ **Error común:** Calcular el costo total sin multiplicar por cantidad.
> El costo de un pedido es `cantidad × costo_unitario`, no solo `costo_unitario`.

> 🧠 **No basta con calcular un número.** Deben interpretar: ¿el negocio es rentable?
> ¿Hay categorías que generan pérdida? ¿Algún país tiene margen negativo?

> 💡 **Conexión con el dashboard:** Los KPIs calculados aquí son los que van
> a visualizar en el Paso 6. Si los calculan bien, el dashboard será más fácil.

# Bloque 5 · Paso 3: Funnel de conversión con SQL (5 min) 🔄

## Petición del jefe
> "Quiero entender en qué parte del proceso los usuarios dejan de comprar."

## ¿Qué deben hacer los estudiantes?

Trabajar con la tabla `events` (base de datos SQL) para construir un **funnel de conversión**.

### Parte 1 — Construcción del funnel
- Contar usuarios únicos por cada tipo de evento (`nombre_evento`)
- Los eventos típicos de un funnel serían: `first_visit` → `product_page` → `add_to_cart` → `purchase`
- Ordenar de mayor a menor cantidad de usuarios

### Parte 2 — Análisis de conversiones
- Calcular la **tasa de conversión** entre cada etapa
- Identificar el **mayor punto de abandono** (drop-off)

## Ejemplo de estructura SQL esperada
```sql
SELECT
    nombre_evento,
    COUNT(DISTINCT id_usuario) AS usuarios
FROM events
GROUP BY nombre_evento
ORDER BY usuarios DESC;
```

## Tips para el Paso 3

> 💡 **Primero explorar, luego construir.** Antes de escribir el funnel,
> que ejecuten un `SELECT DISTINCT nombre_evento FROM events;` para ver qué eventos existen.

> ⚠️ **Usar `COUNT(DISTINCT id_usuario)`**, no `COUNT(*)`.
> Un usuario puede generar múltiples eventos del mismo tipo. Queremos usuarios únicos por etapa.

> 🧠 **La conversión se calcula entre etapas consecutivas:**
> ```
> Tasa conversión = Usuarios etapa N / Usuarios etapa N-1 × 100
> ```
> También es útil calcular la conversión total: Últimos / Primeros.

> 💡 **Recuérdales que la conexión SQL ya está lista** en el notebook.
> Solo necesitan ejecutar la celda de conexión y luego escribir sus queries.

# Bloque 6 · Paso 4: Retención por cohortes con SQL (5 min) 📊

## Petición del jefe
> "Necesito saber si los usuarios regresan o abandonan la plataforma."

## ¿Qué deben hacer los estudiantes?

Usar las tablas `users` y `user_activity` para construir un **análisis de cohortes**.

### Concepto clave: ¿Qué es una cohorte?
Una cohorte agrupa usuarios por su **mes de registro**. Luego se rastrea cuántos
de esos usuarios siguen activos en semanas posteriores.

### Pasos lógicos
1. Identificar la cohorte de cada usuario (mes de `fecha_registro` en `users`)
2. Unir con `user_activity` para ver actividad posterior
3. Calcular `semanas_desde_registro` = diferencia entre `fecha_actividad` y `fecha_registro`
4. Contar usuarios activos por cohorte y semana
5. Calcular la **tasa de retención** = activos en semana N / total de la cohorte × 100

## Tips para el Paso 4

> 💡 **La parte más desafiante es el SQL para cohortes.** Recuérdales que pueden:
> - Usar `DATE_TRUNC('month', fecha_registro)` para obtener el mes de cohorte
> - Calcular semanas con: `(fecha_actividad - fecha_registro) / 7`
> - Usar `JOIN` entre `users` y `user_activity` por `id_usuario`

> ⚠️ **Error común:** No filtrar por `activo = 1` en `user_activity`.
> La columna `activo` indica si el usuario estuvo realmente activo (1) o no (0).

> 🧠 **¿Qué buscar en los resultados?**
> - ¿La retención cae drásticamente después de la semana 1? (Problema de onboarding)
> - ¿Se estabiliza en algún punto? (Usuarios comprometidos)
> - ¿Hay cohortes que retienen mejor que otras? (¿Qué cambió?)

> 💡 **Este análisis conecta con el Paso 5:** si encuentran baja retención,
> el test A/B del siguiente paso podría ser una forma de mejorarla.

# Bloque 7 · Paso 5: Experimentación A/B con Python (5 min) 🧪

## Petición del jefe
> "Hemos hecho cambios en el producto, necesito saber si realmente tuvieron impacto."

## ¿Qué deben hacer los estudiantes?

Trabajar con el dataset `experiment_checkout_ui.csv` para evaluar un **test A/B**.

### Contexto del experimento
Se modificó la interfaz (UI) del checkout. Se asignaron usuarios aleatoriamente a:
- **Control:** UI original
- **Tratamiento:** UI modificada

La métrica principal es **`convirtio`** (1 = compró, 0 = no compró).

### Pasos esperados
1. Cargar el dataset
2. Calcular la tasa de conversión por grupo (control vs. tratamiento)
3. Plantear hipótesis:
   - **H₀:** No hay diferencia en conversión entre los grupos
   - **H₁:** Hay diferencia significativa en conversión
4. Elegir y aplicar un test estadístico adecuado
5. Interpretar el resultado (p-value vs. alpha)
6. Dar una recomendación de negocio

## Tips para el Paso 5

> 💡 **¿Qué test usar?** Al comparar proporciones (tasas de conversión) entre dos grupos,
> el test más adecuado es el **test Z de proporciones** o un **chi-cuadrado**.
> En Python:
> ```python
> from scipy.stats import proportions_ztest
> # o
> from scipy.stats import chi2_contingency
> ```

> ⚠️ **Error común:** Usar un test t de Student para comparar proporciones.
> El test t es para medias de variables continuas, no para tasas de conversión.

> 🧠 **No solo reportar el p-value.** Deben interpretar:
> - Si p < alpha → rechazamos H₀, hay diferencia significativa
> - Si p ≥ alpha → no rechazamos H₀, no hay evidencia de diferencia
> - ¿Cuál es la **recomendación de negocio**? ¿Implementar el cambio o no?

> 💡 **Alpha estándar:** 0.05 (95% de confianza). Déjalos decidir, pero si preguntan,
> este es el valor más común en la industria.
