## 1) Qué es limpieza mínima de datos (13 min)

La limpieza mínima busca que el dataset sea **usable** para análisis básico sin rediseñarlo completamente.

### 1.1 Sí es limpieza mínima

| Problema | Acción razonable |
|---|---|
| Espacios extra | Aplicar `TRIM/ESPACIOS` |
| Mayúsculas inconsistentes | Aplicar `LOWER/MINUSC`, `UPPER/MAYUSC` o `PROPER/NOMPROPIO` |
| Separadores numéricos inconsistentes | Usar Buscar y reemplazar con cuidado |
| Booleanos variados | Mapear a `TRUE/FALSE` o `Sí/No` |
| Encabezados con espacios | Renombrar a formato consistente |
| Duplicados evidentes | Marcar o eliminar después de validar |

### 1.2 No es limpieza mínima

| Situación | Por qué no se hace ahora |
|---|---|
| Crear un modelo de datos completo | Es una tarea posterior |
| Inferir datos faltantes sin regla | Puede introducir sesgo |
| Eliminar registros sin validar | Puede borrar información válida |
| Automatizar todo con scripts | No corresponde al nivel del sprint |
| Corregir fechas ambiguas a ojo | Puede cambiar el significado del dato |

## 2) Buscar y reemplazar como herramienta de limpieza (15 min)

**Buscar y reemplazar** es útil cuando el problema es repetitivo y fácil de describir.

Ruta en Google Sheets:

**Editar → Buscar y reemplazar**

### 2.1 Casos adecuados

| Caso | Ejemplo | Acción |
|---|---|---|
| Quitar símbolo repetido | `$1,200` | Buscar `$`, reemplazar vacío |
| Normalizar separador | `1,863.12` | Evaluar si la coma es miles o decimal |
| Estandarizar categoría | `yes`, `Y`, `1` | Convertir a una convención definida |
| Quitar espacios dobles | `Taylor  Swift` | Reemplazar doble espacio por uno |

### 2.2 Riesgos

Buscar y reemplazar puede dañar datos si se aplica sin revisar.

Ejemplo:

- Buscar `.` y reemplazar vacío puede convertir una fecha, un decimal o un texto de forma incorrecta.
- Buscar `1` y reemplazar `TRUE` puede afectar valores numéricos.

### 2.3 Reglas de seguridad

1. Trabajar en una copia.
2. Aplicar a un rango específico, no a todo el archivo.
3. Revisar ejemplos antes y después.
4. Documentar el cambio.
5. Verificar con una métrica simple.

## 3) Funciones simples de texto (15 min)

Las funciones de texto permiten crear columnas auxiliares limpias.

### 3.1 `TRIM/ESPACIOS`

Quita espacios al inicio, al final y espacios repetidos.

Ejemplo conceptual:

```text
" LOVE STORY " → "LOVE STORY"
```

Uso típico:

```text
=TRIM(A2)
=ESPACIOS(A2)
```

### 3.2 `LOWER/MINUSC`

Convierte texto a minúsculas.

```text
"FOLKLORE" → "folklore"
```

### 3.3 `UPPER/MAYUSC`

Convierte texto a mayúsculas.

```text
"folklore" → "FOLKLORE"
```

### 3.4 `PROPER/NOMPROPIO`

Convierte texto a formato nombre propio.

```text
"the tortured poets department" → "The Tortured Poets Department"
```

### 3.5 Decisión de formato

| Columna | Formato sugerido | Motivo |
|---|---|---|
| `song` | Nombre propio o formato original revisado | Legibilidad |
| `album` | Nombre propio con cuidado | Agrupaciones consistentes |
| `tour_era` | Minúsculas o nombre propio | Categorías comparables |
| `writers` | Mantener texto, limpiar espacios | Puede tener varios autores |

## 4) Verificación rápida con funciones básicas (12 min)

Después de limpiar, se debe comprobar si el cambio funcionó.

### 4.1 Conteos

| Pregunta | Fórmula en inglés | Fórmula en español |
|---|---|---|
| ¿Cuántas celdas numéricas hay? | `=COUNT(rango)` | `=CONTAR(rango)` |
| ¿Cuántas celdas no vacías hay? | `=COUNTA(rango)` | `=CONTARA(rango)` |
| ¿Cuántas celdas vacías hay? | `=COUNTBLANK(rango)` | `=CONTAR.BLANCO(rango)` |
| ¿Cuántas veces aparece una categoría? | `=COUNTIF(rango,"x")` | `=CONTAR.SI(rango;"x")` |

### 4.2 Valores únicos

Para ver categorías distintas:

```text
=UNIQUE(rango)
=UNICOS(rango)
```

Esto es útil para columnas como:

- `Explicit`
- `Tour Era`
- `Album`

### 4.3 Promedios

Una columna solo debe promediarse si Google Sheets la reconoce como numérica.

```text
=AVERAGE(rango)
=PROMEDIO(rango)
```

Si el promedio falla o ignora valores, probablemente algunos números siguen como texto.

## 5) Limpieza guiada por tipo de columna (12 min)

### 5.1 Columnas de texto

Ejemplos:

- `Song`
- `Album`
- `Writers`
- `Tour Era`

Rutina sugerida:

1. Crear columna auxiliar.
2. Aplicar `TRIM/ESPACIOS`.
3. Aplicar una convención de mayúsculas/minúsculas si es necesario.
4. Revisar `UNIQUE/UNICOS`.
5. Reemplazar la columna original solo si el resultado fue validado.

### 5.2 Columnas numéricas

Ejemplos:

- `Spotify Streams (M)`
- `YouTube Views (M)`

Rutina sugerida:

1. Revisar ejemplos.
2. Identificar separador decimal y separador de miles.
3. Usar Buscar y reemplazar en copia.
4. Verificar con `COUNT/CONTAR`.
5. Calcular promedio con `AVERAGE/PROMEDIO`.

### 5.3 Columnas booleanas

Ejemplo:

- `Explicit`

Posible convención final:

| Original | Limpio |
|---|---|
| `Y`, `Yes`, `1`, `TRUE` | `TRUE` |
| `N`, `No`, `0`, `FALSE` | `FALSE` |
| `maybe`, vacío | revisar o dejar como pendiente |

### 5.4 Fechas

Ejemplos:

- `Release Date`
- `Date Added`

En nivel principiante, las fechas se revisan con cuidado.  
No se debe “adivinar” una fecha ambigua sin regla de negocio.

Ejemplo de ambigüedad:

```text
12/09/08
```

Puede interpretarse como:

- 12 de septiembre de 2008, o
- 9 de diciembre de 2008.

Cuando una fecha es ambigua, se documenta y se consulta la regla.

## 7) Criterio profesional: cuándo detenerse (7 min)

Un error común en limpieza es intentar dejar el archivo “perfecto”.

En análisis real, se busca un balance entre:

- tiempo disponible,
- impacto del problema,
- riesgo de modificar datos,
- necesidad de negocio,
- trazabilidad.

### Matriz simple de decisión

| Situación | Decisión recomendada |
|---|---|
| El problema impide calcular una métrica central | Limpiar ahora |
| El problema afecta pocas filas y no impacta el análisis actual | Documentar y dejar pendiente |
| La corrección requiere conocer una regla de negocio | Consultar antes de cambiar |
| El dato parece inválido pero no hay evidencia | Marcar para revisión |
| La limpieza puede dañar otros valores | Probar primero en una copia |
