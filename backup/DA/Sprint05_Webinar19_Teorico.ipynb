```python

# TODO:
# Crea una variable derivada o segmentación útil para el análisis.
# Ejemplo sugerido: un segmento de precio.
# Puedes completar el código o proponer otra lógica.

bins = [0, 10, 25, 100]
labels = ["económico", "medio", "premium"]

# Completa aquí
df_licores["price_segment"] = pd.cut(
    df_licores["price_usd"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

df_licores[["price_usd", "price_segment"]].head()
```

```python

import matplotlib.pyplot as plt

# TODO:
# Construye una tabla resumen o un gráfico simple que responda
# una pregunta de negocio del ejercicio 3.

# Sugerencia 1: exportaciones totales por país
# Sugerencia 2: precio promedio por tipo de licor
# Sugerencia 3: boxplot de price_usd por país

# Escribe tu solución debajo
resumen = (
    df_licores
    .groupby("country")["export_volume_liters"]
    .sum()
    .sort_values(ascending=False)
)

resumen.plot(kind="bar", figsize=(10, 5))
plt.title("Volumen total de exportación por país")
plt.xlabel("País")
plt.ylabel("Litros exportados")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
