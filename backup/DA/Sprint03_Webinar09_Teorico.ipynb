# Bloque 1 · SQL en contexto

SQL significa **Structured Query Language**. Es un lenguaje declarativo para consultar, transformar y resumir datos organizados en tablas.

La palabra clave es **declarativo**: no le dices al motor cada operación física paso a paso. Le indicas qué resultado quieres obtener. El motor decide cómo leer, filtrar, unir y calcular internamente.

### Preguntas de negocio que SQL puede responder

Completa con un ejemplo propio:

| Tipo de pregunta | Ejemplo |
|---|---|
| Conteo | ¿Cuántas órdenes se hicieron...? |
| Segmentación | ¿Qué tipo de cliente...? |
| Ranking | ¿Cuáles productos...? |
| Rentabilidad | ¿Qué categoría...? |
| Comparación | ¿Qué pasa con y sin...? |

### Flujo mental

```text
Datos en tablas o archivos → Consulta SQL → Resultado tabular → Interpretación de negocio
```

# Bloque 2 · Modelo mental de una consulta SQL

Aunque una consulta se escribe empezando por `SELECT`, conviene pensarla en este orden lógico:

```text
FROM      → de dónde salen los datos
WHERE     → qué filas se conservan
GROUP BY  → cómo se agrupan las filas
HAVING    → qué grupos se conservan
SELECT    → qué columnas o cálculos se muestran
ORDER BY  → cómo se ordena el resultado
LIMIT     → cuántas filas se muestran
```
