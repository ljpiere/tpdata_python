```python
# Crear una copia para trabajar sin alterar el dataset original
df_clean = df.copy()

# Ejemplo de renombrado de columnas
# Revisa si estos nombres existen exactamente en tu dataset.
df_clean = df_clean.rename(columns={
    "Title": "title",
    "Album": "album",
    "Release Year": "release_year",
    "Duration (mins)": "duration_mins",
    "Genre": "genre",
    "Spotify Streams (Millions)": "spotify_streams_millions"
})

df_clean.head()
```

## 5. Filtrado de datos y slicing con Pandas

### Ideas clave para completar

### Traducción de negocio a código

Completa el patrón:

- **Pregunta de negocio:** “Quiero ver ______________________________”
- **Columnas necesarias:** __________________________________________
- **Filtro o condición:** ____________________________________________


```python
# Ejemplos de filtrado
# Observa, modifica y vuelve a ejecutar.

# Canciones Pop
df_pop = df_clean[df_clean["genre"] == "Pop"]
df_pop.head()

# Canciones Pop con muchos streams
df_pop_hits = df_clean[
    (df_clean["genre"] == "Pop") &
    (df_clean["spotify_streams_millions"] > 1500)
]

df_pop_hits
```

```python
# Slicing por posición
df_clean[0:3]

# Selección por posición con .iloc
df_clean.iloc[0:5, 0:4]
```

```python
# Selección de columnas específicas
columnas_clave = ["title", "album", "genre", "spotify_streams_millions"]
df_clave = df_clean[columnas_clave]
df_clave.head()
```
