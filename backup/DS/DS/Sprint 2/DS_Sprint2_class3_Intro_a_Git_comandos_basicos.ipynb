## <span style="color: #2826DE;">¿Qué es Git y control de versiones? (20 mins)

### <span style="color: #2772F2;">El problema sin control de versiones

¿Alguna vez has tenido archivos como estos?

```
analisis_final.py
analisis_final_v2.py
analisis_final_v2_corregido.py
analisis_final_ESTE_SI.py
analisis_final_ESTE_SI_de_verdad.py
```

**Git resuelve este problema** guardando el historial completo de cambios de forma organizada.

### Preguntas:

- ¿Has tenido este problema de versiones de archivos antes? ¿Cómo lo resolviste?

- ¿Qué información perderías si solo guardas la última versión de un archivo?

### <span style="color: #2772F2;">¿Qué es el control de versiones?

Es un sistema que **registra los cambios** en archivos a lo largo del tiempo, permitiendo:

- **Volver a versiones anteriores** si algo sale mal
- **Colaborar** con otros sin sobrescribir el trabajo de nadie
- **Ver quién cambió qué** y cuándo
- **Experimentar** en ramas separadas sin afectar el código principal

### <span style="color: #2772F2;">Git vs otros sistemas

| Sistema | Tipo | Estado |
|---|---|---|
| **Git** | Distribuido | ✅ Estándar de la industria |
| SVN | Centralizado | Obsoleto |
| Mercurial | Distribuido | Poco usado |

**Git ganó** porque es rápido, distribuido (cada copia es un backup completo) y tiene GitHub.

### <span style="color: #2772F2;">Por qué Git es fundamental para DA/DS

- **Reproducibilidad**: puedes volver exactamente al estado del código cuando generaste un resultado
- **Colaboración**: trabajar en equipo sin conflictos
- **Portafolio**: GitHub es tu CV técnico
- **Entrevistas**: si no sabes Git, pierdes puntos

> **"Un Data Scientist sin Git es como un chef sin recetas escritas."**

## <span style="color: #2826DE;">Conceptos clave de Git (20 mins)

### <span style="color: #2772F2;">Glosario esencial

| Concepto | Definición | Analogía |
|---|---|---|
| **Repositorio (repo)** | Carpeta con historial de cambios | Carpeta de proyecto con memoria |
| **Commit** | Foto del estado del proyecto en un momento | Guardar partida en un videojuego |
| **Rama (branch)** | Línea paralela de desarrollo | Borrador de un documento |
| **Remoto (remote)** | Copia del repo en un servidor | Backup en la nube |
| **Origin** | Nombre por defecto del remoto principal | El servidor de GitHub |
| **Tag** | Etiqueta en un commit específico | Marcar una versión como v1.0 |

### <span style="color: #2772F2;">Los 3 estados de un archivo en Git

```
Working Directory  →  Staging Area  →  Local Repository
   (modificado)         (preparado)        (guardado)
       git add →              git commit →
```

## <span style="color: #2826DE;">Comandos básicos de Git (30 mins)

### <span style="color: #2772F2;">Configuración inicial (solo una vez)

```bash
# Configurar tu identidad
git config --global user.name "Tu Nombre"
git config --global user.email "tu@email.com"

# Verificar configuración
git config --list
```

### <span style="color: #2772F2;">Flujo básico de trabajo

```bash
# 1. Inicializar un repositorio
git init

# 2. Ver el estado actual
git status

# 3. Agregar archivos al staging area
git add archivo.py          # Un archivo específico
git add .                   # Todos los archivos

# 4. Hacer un commit
git commit -m "Descripción clara de los cambios"

# 5. Ver el historial
git log
git log --oneline           # Versión compacta
```

### <span style="color: #2772F2;">Demo en vivo: crear tu primer repositorio

Sigue estos pasos en tu terminal:

```bash
# Crear una carpeta de proyecto
mkdir mi_proyecto_ds
cd mi_proyecto_ds

# Inicializar Git
git init

# Crear un archivo
echo "# Mi Proyecto de Data Science" > README.md

# Ver el estado
git status

# Agregar el archivo
git add README.md

# Hacer el primer commit
git commit -m "Primer commit: agregar README"

# Ver el historial
git log --oneline
```

## <span style="color: #2826DE;">Tips y buenas prácticas (5 mins)

- **Haz commits frecuentes** — cada vez que termines una tarea pequeña

- **Escribe mensajes descriptivos**: `"Agregar función de limpieza de datos"` no `"cambios"`

- **Nunca hagas commit de secretos** (contraseñas, API keys) — usa `.gitignore`

- **Usa `.gitignore`** para excluir archivos innecesarios (`.venv/`, `__pycache__/`, `*.pyc`)

- **Sube todos tus proyectos del bootcamp** a GitHub — es tu portafolio

## <span style="color: #2826DE;">Cierre y Kahoot (5 mins)

Kahoot – Preguntas sugeridas

1️⃣ ¿Qué hace `git init`?

- Sube el repositorio a GitHub
- Hace un commit con todos los archivos
- Descarga un repositorio remoto
- **Inicializa un repositorio Git local** ✅


2️⃣ ¿Qué hace `git add .`?

- Hace un commit de todos los archivos
- Sube los archivos a GitHub
- Descarga cambios del remoto
- **Agrega todos los archivos al staging area** ✅


3️⃣ ¿Cuál es la diferencia entre Git y GitHub?

- Son exactamente lo mismo
- GitHub es más antiguo que Git
- Git es solo para proyectos grandes
- **Git es la herramienta local, GitHub es el servicio web** ✅


4️⃣ ¿Qué es un commit?

- Un error en el código
- Una rama del repositorio
- Un archivo de configuración
- **Una foto del estado del proyecto en un momento** ✅


5️⃣ ¿Qué comando muestra el historial de commits?

- git status
- git add
- git push
- **git log** ✅


6️⃣ ¿Para qué sirve `.gitignore`?

- Para hacer commits más rápidos
- Para crear ramas automáticamente
- Para sincronizar con GitHub
- **Para excluir archivos del repositorio** ✅


7️⃣ ¿Cuáles son los 3 estados de un archivo en Git?

- Creado, Modificado, Eliminado
- Local, Remoto, Archivado
- **Working Directory, Staging Area, Repository** ✅
- Nuevo, Guardado, Publicado


8️⃣ ¿Qué hace `git push -u origin main`?

- Descarga cambios del repositorio remoto
- Crea una nueva rama llamada main
- Elimina el repositorio remoto
- **Sube los commits locales a GitHub por primera vez** ✅
